# 섹션1. 인프라 구축 이해

## 인프라 보안의 필요성

- 현대 사회는 비즈니스와 IT 연관성이 매우 밀접하여 보안과 비즈니스 방법론 연관성도 강화.
- 시간이 지날수록 IT 기반의 비즈니스가 많아지고 이러한 서비스를 공격하려는 시도가 많아짐.
- 민감한 데이터와 함께 사진이 많이 수집이 되어 개인의 사생활침해 등과 같은 문제를 피하기 위해 인프라 보안을 철저히 해야 함.

### 그래서, 왜 필요한가?

- 모의해킹이든 침해대응 분석이든 인프라에 대한 이해가 필요하다. 인프라에 엄청나게 많은 양의 자원들이 모여있는데, 이 인프라가 대략적으로 어떻게 생겼는지 그리고 어떤 역할을 하는지 모르면 원하는 정보를 찾을 수 없게 된다.
- 예를 들어, 공격을 당했을 때 분석관 또는 분석가들이 분석을 하게 되는데 ‘이 로그를 보니 공격자가 어떻게 해서 어떤 정보들을 빼앗을 거다’라는 것을 그릴 수 있어야 한다. 그러나, 이런 인프라에 대한 이해가 없으면 연관된 것들을 이해하거나 단서를 찾기가 굉장히 어렵게 된다.

### 인프라 보안 설계

- 실제 업무를 하는 경우 직접 구축하는 경우보다는 이미 구축되어 있는 환경을 관리하는 경우가 많다. 그러기 때문에 실습을 진행하지 않아 기억에 잘 남지 않는 경우가 많아 실제로 테스트하지 못해도 오픈소스 기반으로 경험을 쌓을 수 있다. 이외에도 트러블 슈팅 과정 등 다양한 장점이 존재한다.
- 기본적으로 공용망이 있고 중간 지점(DMZ)이 있고 실제 업무를 하는 업무망을 따로 설계한다. 그리고 중간지점을 통해 외부에 서비스를 제공한다.
