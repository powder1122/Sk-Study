## VMware ESXI

- 맞춤형 베어메탈 하이퍼바이저로, 물리적 서버에 바로 설치되는 강력한 베어메탈 하이퍼바이저이다.
- 베어메탈 하이퍼바이저는 OS 없이 하드웨어 위해 직접 설치되어 실행되는 하이퍼바이저이다. 하드웨어 요소를 직접 제어할 수 있어 네이티브에 가까운 성능을 제공한다.
- VM을 사용하면 스냅샷을 통해 공격을 당하거나 문제가 발생해도 쉽게 복구가 가능하여 확장이나 유지보수가 훨씬 편하다.

## ESXI 설치

- 해당 강의 최소 설치 사양
    - CPU 2 core
    - RAM 4GB
    - HDD(SSD) 500GB
- VMware vSphere Hypervisor 6.7 version 사용. iso 파일 제공

### 순서

1. VMware 설치: VMware Workstation(VT옵션 활성화. 바이오스 화면에서 Virtualization enable로 설정)
2. VMware 위에 ESXI 설치
    1. Hypervisor 삭제 후 진행할 것.
    2. 라이센스 없이 진행
    3. VMware-VMvisor-installer-6.7.0 사용(제공됨)
    4. new virtual machine 클릭하여 새로운 가상머신 설치
    5. Typical Recommended 진행
    6. ISO 파일: vSphere Hypervisor로 세팅
    7. 호스트 이름 설정
    8. 디스크 용량 최소 200GB
    9. store virtual disk as a single file 선택(빠른 액세스)
    10. CPU 4~8, Memory 16~32GB 설정하기.
        1. 2-1. 하이퍼바이저 오류 발생 시
        2. error: Virtualized Intel VT-x/EPT is not supported on this platform. 
        3. 해당 오류 발생 시 관리자 모드로 파워쉘을 열고 bcdedit을 쳤을 때 Hypervisor가 auto로 되어 있는 경우에 이 에러가 발생하는데, 옵션을 off로 변경하면 해결된다. (이 후 reboot)
    
    ```powershell
    bcdedit /set hypervisorlaunchtype off
    ```
    
3.  기본 설정
    1.  약관 동의
    2. 200GB 확인
    3. 키보드: usdefault
    4. ID : root  /  PW : rhkswp123!(=관제123!)
    5. 설치 완료 후 리부트
    6. 받은 주소로 접속하여 GUI 환경으로 ESXI 컨트롤 가능

    4. ISO 파일 선택

# 가상 스위치를 사용한 네트워크 분리
## 기본 지식

- NIC: 가상 네트워크 어댑터 드라이버
- 표준 가상 스위치: 가상 머신들의 NIC를 연결하는 스위치
- 업링크: 외부와의 연결 상태(업링크가 없는 경우 외부와 단절)
- 포트 그룹: 가상 스위치 내에 생성하는 그룹, 보안과 트래픽 단위 조절, 네트워크 이중화 같은 속성을 지정하는데 사용. **그룹 단위**로 관리할 수 있다.

### ESXI에서 네트워킹 설정

- 네트워크 탭에서 포트 그룹을 생성해 각각 분리된 네트워크를 구성
- **총 7개의 네트워크 구성**
    1. route-wan-net(VM Network, vSwitch0, 가상 스위치)
    2. route-att-net(공격자)
    3. route-fw-net(방화벽)
    4. intra-net(서버존)
    5. dmz-net(서비스)
    6. soc-net(보안부서)
    7. dev-net(개발부서)
- 외부 네트워크 대역에서 방화벽으로 들어오기 전에 라우터 설치
- 공격자도 이 라우터와 연결되어 있음
- 각 구간마다 스위치가 필요하고, 스위치에는 적어도 한 개 이상의 포트 그룹이 존재해야 한다.
- **업링크 하나**인 상태. 그림과 완벽히 같지 않다.
- 비규칙 모드
    - **IDS 장비**를 사용할 때 트래픽을 미러링 하여 보기 때문에 TAP 또는 미러링 장비(스위치 등)를 사용해 트래픽을 복사해서 본다. 트래픽 저하를 최소화하기 위해 사용하는 방법으로, 이를 구현하기 위해 **Promiscuos mode**를 사용한다.
    - 보안 설정에서 비규칙 모드 동의로 변경하기

## 가상 스위치 생성과 가상 머신 연결

### 네트워크 연결하기

1. ESXI → 네트워킹 → 가상 스위치 → 표준 가상 스위치 추가 - 새 스위치
2. 설정하기
    1. vm network(wan-net. 기본)
    2. route-att-net(라우터와 어태커를 연결)
    3. route-fw-net(라우터와 방화벽 연결)
    4. dmz-net
    5. intra-net
    6. soc-net
    7. dev-net
3. 업링크 설정하기
    1. 각각의 스위치에 **포트 그룹 추가**해야함.
    2. 이름에 맞는 포트 그룹 추가해주기
    3. 잘못 만들었을 시 삭제 후 다시 생성. (수정 불가능)
4. kali에 가상 스위치 연결해보기 → route-att-net 연결
5. mint에 soc 연결 → soc-net
6. 인터넷 연결은 뒤에서 작업.
