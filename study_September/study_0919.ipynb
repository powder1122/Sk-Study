# 네트워크 보호
## 목차

- 주요 네트워크 공격 기법(스캐닝,스니핑&스푸핑, 도스&디도스)

## 네트워크 보호의 필요성
- 중국 해킹그룹 샤오치잉의 해킹 사례
    - 국내 12개 학술 기관이 해킹당함
    - 공개되어서는 안되는 웹진 신청 이메일 주소 목록 60건의 유출 피해 의심

- 이러한 피해를 막기 위해 필요하다.

- 대부분의 침해사고는 네트워크를 통해 발생한다.
- 침해사고 발생 시 원인을 규명하고 복구하기 위한 노력이 많이 필요하다.
- 침해사고로 이어지기 전, 공격이 들어오는 경계 또는 통로에서 사이버 공격과 침해 시도를 탐지하여 차단하는 것이 중요

## OSI 참조 모형
- 정보시스템과 네트워크
    - 정보시스템: 정보를 수집,가공,저장,검색하거나 수신,송신 및 배포하도록 설계,조직화된 시스템
    - 과거에는 스탠드얼론 형태로 사용했음
        - 스탠드얼론(SA): 네트워크 연결 없이 독립적인 단일 컴퓨터로만 구성된 형태
    - 네트워크: 2대 이상의 컴퓨터를 통신매체만으로 상호 연결하거나 통신장치(네트워크 장비)를 두고 연결,구성된 형태로 오늘날 가장 보편적인 형태이다.

- 네트워크가 일상적인 환경으로 자리잡은 이후 스탠드얼론 형태는 잘 사용되지 않음

- 네트워크의 구성 단위 구분
    - 근거리통신망(LAN): 네트워크의 최소 단위
    - 광역통신망(WAN): 여러 개의 LAM이 넓은 지역에 걸쳐 구성

### OSI 참조 모형의 이해
- 네트워크를 통해 자료/정보를 전달함에 있어 각 통신장치 개발사들이 참조할 수 있도록 자료/정보의 송싱 및 수신 구조를 국제표준화기구(ISO)에서 표준 모형으로 정한 것

## 주요 네트워크 공격 기법
### 스캐닝
- 공격자가 공격대상의 네트워크나 시스템 정보를 탐색하는 행위
    - 라이브 포트 스캐닝은 법적 문제 소지가 있다.
    - 활성 호스트 스캐닝
    - 포트 스캐닝
    - 취약점 스캐닝

1. 활성 호스트 스캐닝
    - 현재 동작 중인 컴퓨팅 장치(호스트)를 찾기 위한 스캐닝 방법
    - ICMP를 이용한 ping 통신을 통해 동작 여부 확인
    - 직접적인 공격 영향 없음
    - 종류
        - ICMP Echo ping
        - ARPing
        - TCP/UDP ping
    - 대표적인 도구
        - Nmap
        - ARPing

2. 포트 스캐닝
    - 호스트에서 실행 줒ㅇ인 서비스를 찾기 위한 스캐닝 기법
    - TCP Port Scanning
    - UDP Port Scanning
    - 종류
        - TCP Connect Scanning: TCP 연결을 맺어서 포트의 활성화 여부를 판단하는 방법, 포트가 열려있으면 로그가 남음(하수)
        - SYN Stealth(Half Open) Scanning: TCP 연결을 완전히 하지 않고 포트의 활성화 여부를 판단. 로그가 남지 않음
        - FIN Scanning: FIN 플래그를 설정하여 전송하여 확인하는 방법, 포트가 열려있으면 응답하지 않음
        - X-MAS Scanning: TCP 헤더 내 일부 또는 모든 플래그를 설정하여 전송하고, 응답으로 판단하는 방법. 열린 경우 무응답.
        - UDP Port Scanning: ICMP Unreachable 메시지를 이용하여 판단하는 방법, 열린 경우 UDP응답이 오거나 응답 없음(정책 차단), 닫힌 경우 ICMP메시지 응답(type3)

3. 취약점 스캐닝
    - Nmap을 이용한 취약점 스캐닝




### 스니핑과 스푸핑
- 스니핑은 네트워크 트래픽을 도청하는 공격
- 허브 환경과는 달리 스위치 환경에서 스니핑을 하기 위해서는 스푸핑 등을 유발시키는 선행 공격 필요
- 스위치 재밍
- ARP 스푸핑
- ARP 리다이렉트
- ICMP 리다이렉트

### DoS/DDoS
- DoS: 시스템을 악의적으로 공격해 해당 시스템의 자원을 고갈시키는 공격
    - TCP SYN Flooding: SYN 요청을 지속적으로 보내는 네트워크 계층 공격. 반개방 공격이라고 불리며, TCP 연결 과정을 이용한 공격이다.
    - TCP Connection Flooding: TCP 세션을 무수히 연결하여 자원을 고갈시키는 공격
    - DNS Query Flooding: DNS Query 요청 패킷을 대량으로 서버에 전송하여 정상적인 서비스 방해
    - Slow HTTP Header Dos: HTTP 헤더 정보를 비정상적으로 조작하여 웹서버가 온전한 Header 정보가 올 때까지 기다리도록 함. 헤더의 끝을 \r\n이라는 문자로 구분하게 되는데, 이 개행 문자를 보내지 않는다.
    

# 방화벽(침입탐지시스템)
## 목차
- 방화벽
- 방화벽 정책/규칙 작성



## 방화벽
- 네트워크 간 전송되는 패킷들을 정해진 규칙에 따라 통과시키거나 차단시켜, 공격자로부터 내부 정보자산을 보호하기 위한 목적으로 사용되는 가장 기본적인 정보보호시스템
- 보통 화이트리스트 방식으로 운영

### 방화벽 정책/규칙의 이해
- 차단 규칙을 잘못 구성하느 경우
    - 모든 트래픽이 허용되어 악성 트래픽이 유입될 가능성
    - 모든 트래픽이 차단되어 어떠한 통신도 불가할 가능성
    

### 방화벽 차단 규칙 구성 시 주의사항
- 모듀 허용 규칙 금지: 방화벽이 없는 것과 다름없음
- 과도한 허용 규칙 지양: 특정 IP 또는 대역으로 허용 범위 한정
- 테스트용 임시 등록 규칙 테스트 종료 시 삭제: 테스트 목적과 유효기간 명시
- 실효성 없는 장기 미사용 규칙 삭제
- 불필요한 중복 규칙 삭제

## 방화벽 정책/규칙 작성
### 실습
- Network Firewall Visualization
    - 미 공군사관학교 소속 장교가 개발한 교육용 실습 도구
    - 애니메이션 효과를 통해 직관적인 사용자 환경 제공
    - 방화벽 유무에 따른 네트워크 환경별로 감염되는 과정을 볼 수 있음

- 방화벽 규칙을 작성해보기
- SPI(Statefule Packet Inspection)는 IN,OUT 구분 없이 ALLOW 하나로 관리 가능하다.
- 간단한 환경이기 때문에 기능에 한계가 존재한다.
- 또한, 실제로 스캔 같은 공격이 아예 들어오지 않는다면 오히려 규칙에 문제가 있을 수 있다.

# IDPS(침입탐지/방지시스템)
## 목차
- IDPS의 이해
- 정규표현식과 기본 문법
- IDPS 정책/규칙 국성
- PCRE 기반 IDPS 규칙 작성


## IDPS의 이해
### 방화벽의 한계
- 방화벽은 전통적인 정보보호시스템으로 1차 방어선을 구축해주지만, 한계가 있다.
    - 본문 내용(컨텐츠)까지 확인하지 않는다.
        - 공격자의 행위가 발생되는 영역까지 확인하기 위해서는 본문 데이터까지 확인이 필요하다.
    - 해킹메일의 피해 --> IP와 포트만으로 누가 정상 사용자인지 알 수 없다.
    - 첨부파일 오픈 시 악성코드 감염

- 여전히 1차 방어선으로 필요한 기술이다!

### IDPS
- IDS: 네트워크 간 전송되는 패킷을 수집하고 내용을 분석하여 침해 시도르 **탐지**하는 정보보호시스템
- 침입 탐지 시스템으로, 실시간으로 탐지하지만 **사후분석 대응 방식**으로 활용한다.
- 방화벽이 구성된 네트워크 경계 보호를 강화하기 위한 목적으로 사용한다.

- IPS: 네트워크 간 전송되는 패킷을 수집하고 내용을 분석하여 침해 시도를 **탐지하고 차단하는** 정보보호시스템
- IDS 기능에 차단 기능이 추가된 시스템
- **실시간 대응 방식**으로 활용 가능
- 근래에는 IDS보다 IPS를 사용하는 경향
- 방화벽이 있는 상태에서 설치



### IDS 설치 위치
- IDS는 TAP 장치를 이용하여 미러링한 트래픽을 탐지한다.
- 직접적으로 설치하는 것이 아니라, 미러링한 트래픽을 설피할 수 있도록 아웃라인으로 설치
- 네트워크에 대한 영향을 최소화할 수 있으나 실시간 대응은 불가
- 미러링 모드: 네트워크의 모든 트래픽을 복제하여 복제된 트래픽이 정보보호시스템에 도달하도록 구성하는 형태

### IPS 설치 위치
- IDS와는 다르게 인라인 방식으로 설치한다.
- 인라인 모드: 네트워크의 모든 트래픽이 반드시 정보보호시스템을 경유하도록 구성하는 형태
- 규칙에 따라 패킷을 허용하거나 거부(차단) 가능
- 모든 패킷에 대한 판단이 요구되기 때문에 부하 발생 가능
    - 장애 발생 시 우회할 수 있게 구성하거나 이중화 구성 필요


### IPS로의 전환
- 컴퓨터와 네트워크 성능의 발달로, IDS를 인라인으로 설치하여 실시간 차단 기능을 제공하더라도 네트워크 성능을 과도하게 저하시키지 않을 수 있게 됨
- 사후적인 방식으로 대응하기에는 공격의 파급력이 지나치게 커짐
- 확산력이 빠른 공격이 증가


### 주요 기능
- 침입 탐지
    - 오용탐지: 
        - 알려진 공격 패턴을 미리 규칙으로 등록해두고 일치 여부를 탐지하는 방식
        - 알려진 공격 이외에는 탐지 불가능
        - 대량의 데이터 분석에 적합하지 않고, 공격 순서를 추론하기 어려움
    - 상태전이:
        - 공격 상황에 대한 시나리오를 작성하여 각 상태에 따른 공격을 분석하는 방식
        - 결과가 직관적이지만, 세밀한 시나리오 제작을 위해 고도의 기술과 노하우가 요구됨
        - 추론 엔진이 포함되어 성능에 영향을 줄 수 있음
    - 이상탐지:
        - 급격한 변화의 발생이나 발생 확률이 낮은 이벤트의 발생 여부를 탐지하는 방식으로, 알려지지 않은 공격도 탐지할 수 있음
        - 인공지능 수준이 낮을 경우 오탐률이 높고, 인공지능 학습이 필요하다.

- 책임 추적성과 알림
    - 침입탐지 사실을 관제사들이 쉽게 인지할 수 있도록 실시간으로 표시하고 알리는 기능 제공
    - 기간별, 시간대별, 공격 유형별 등 필터링 및 분석 기능 제공
    - 침입 방지(IPS): 탐지하고 알리는 데 그치지 않고 차단을 통해 능동적인 대응 수행


- 집에서 IDS를 구현할 수 있는 방법
    - ZEEK를 통해 직접 홈IDS 구축 가능
    - GS102 스위치를 통해 TAP(미러링) 장치로 동작시킬 수 있다.


## 정규표현식과 기본 문법
- 정규표현식: 어떤 문자열 내에서 '특정한 형태나 규칙을 가진 문자열'을 찾기 위해 그 형태나 규칙을 나타내는 패턴을 정의하는 식

- POSIX 정규식
- vim 정규식
- PCRE

### 정규표현식의 문법
- 시작과 끝은 같은 구분자를 사용
- 일반적으로 슬래시를 많이 사용하지만, 정규표현식 내 슬래시가 포함되는 경우는 다른 문자를 사용할 수도 있음

- 리터럴
    - 문자 그대로의 값을 의미
    - 이를테면 apple을 찾는 경우 apple을 그대로 기재.(/apple/패턴변경자)

- 메타문자
    - /패턴/패턴변경자
- 수량한정자
    - ?, *, +, {n}, {n,m}
- 탐색
    - ?=, ?<=
- 패턴변경자
    - i, m, s, x, g

## IDPS 정책/규칙 구성
- IDPS의 탐지/차단 원리
    - IDPS의 특성 상 블랙리스트 방식으로 동작
    - 규칙을 등록하지 않으면 설치하지 않은 것과 동일한 상태
    - 천차만별인 본문 데이터에서 어떤 내용을 검사할지 정의해줘야 함
        - 규칙성을 갖고 변화하는 경우도 있음
    - 내용을 검사하는 것은 컴퓨팅 자원을 많이 활용하여 네트워크 성능에 영향이 있으므로, 빠르고 정확하게 검사할 수 있는 기술과 엔진이 요구됨

- 대부분의 IDPS는 PCRE 정규표현식 규격을 적용한 오픈소스 IDS인 Snort를 널리 활용하고 있기 때문에 IDPS 탐지/차단 규칙을 Snort 규칙(Snort Rule)이라고도 함

- Snort Rule: 동작 - 프로토콜 - 출발지IP - 출발지 포트 - 방향 - 목적지IP 목적지 포트 - 내용

- 동작
    - alert: 규칙과 일치하는 경우 경고 발생, 로그 파일에 기록
    - log: 로그 파일에 기록
    - pass: 패킷 무시
    - drop: 패킷을 차단하고 로그에 기록
    - reject: 패킷을 차단하고 로그 파일에 기록하며 프로토콜에 따라 대응
        - TCP: RST 응답
        - UDP: ICMP Unreachable 패킷 응답
    - sdrop: 패킷을 차단하되, 로그 파일에 기록하지 않음

- 프로토콜
    - TCP
    - UDP
    - ICMP
    - IP

- IP주소
    - any: 모두
    - 지정하는 경우 'IP주소/비트수' 형태로 표시하고, 여러 주소를 지정하는 경우 []를 사용하며 ,로 구분

- 포트 번호
    - n:m --> n~m
    - :n --> n 이하
    - n: --> n 이상
    - !n:m -> n~m을 제외한 나머지
    - [] -> 여러 포트 번호들을 지정, 콤마로 구분

- 방향
    - -> : 출발징제서 목적지까지의 방향
    - <> : 출발지와 목적지 사이의 모든 방향

- 내용(주요 옵션들)
    - 괄호로 둘러싸며, 각 항목 사이를 세미콜론으로 구분
    - msg: 액션이 alert인 경우 표시할 내용
    - sid: 규칙 ID(1,000,001 ~ : 사용자 정의)
    - rev: 규칙 수정 횟수
    - dsize: 상한 범위 또는 하한 범위(바이트 단위)
    - content: 정적 문자열 또는 HEX 값
    - offset: 검색을 건너뛸 바이트 수
    - depth: offset을 기점으로 검색할 바이트 수
    - nocase: 대소문자 구별하지 않음
    - flags: TCP 제어 플래그
    - pcre: 정규표현식
    - threshold: 발생량 기반 탐지
        - type limit(임계값 기준)/threshold(패킷량 기준)/both(IP기준)
        - track by_src(송신지 IP 기준 추적) / by_dst(목적지 IP 기준 추적)
        - count n: 횟수
        - seconds m: 초

- 사용 예
    - LDAP 서버에 대한 LDAP Injection 탐지
    - 웹 서버에 대한 SQL Injection 탐지
    - 악성코드 감염 PC에서 발생하는 통신 탐지



- 보통 벤더사에서 제공해주는 룰셋이 있다.
