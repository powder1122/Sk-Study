## Splunk 구성과 메타 데이터
- Single Server의 경우
    - Search Header
    - Indexer
    - Deployment
    - Forwarder

### 구성 요소별 기능
- Search Header: 
    - 인덱서를 통해 병렬수집된 검색된 부분 결과를 통합하여 사용자에게 제공(데이터 분석/시각화). 
    - 인덱서에서 수집된 데이터를 통계처리 및 정렬
- Indexer: 
    - 스트리밍+비스트리밍 명령어 처리 (데이터 검색을 실제 수행)
    - 인덱스 생성,관리 및 데이터 인덱싱
    - 원시 데이터와 인덱싱된 데이터를 저장 및 검색
- Deployment: 로컬 및 분산 인스턴스 관리
- Forwarder: 
    - 다양한 OS 지원
    - 데이터를 수집하는 구셩요소
    - 데이터 로드밸런싱 수행
    - 인덱서에 데이터 전송 --> 인덱서가 파싱

### Splunk Metadata
- 내부 필드
    - Splunk 이벤트에 대한 일반적인 정보를 포함하는 필드
- 메타데이터 필드
    - 인덱싱 되고 필드 메뉴에 기본적으로 추가됨
    - 이벤트가 시작된 저장소, 위치해 있는 저장소, 이벤트 유형, 포함된 줄 수, 발생 시간에 대한 정보가 포함된 필드 
- 시간 필드
    - 타임스탬프에 더 세부적으로 추가된 정보 필드
    - data_*필드는 각 시스템에서 생성된 타임스탬프 정보가 있는 이벤트만 생성됨

### 이상치 탐지 실습
- secure.log가 유닉스/리눅스 인증 기반 로그이다.
- Failed password는 인증에 실패한 것
- 특정 시간 동안 로그인 실패가 5번이 넘는다! --> 공격 의심
- 전체 기간에 대해서 접속 횟수를 분석해도 되지만, 특정 시간에 집중적인 접속 시도가 있었는지 살펴볼 필요도 있다.

- source=*secure.log" "Failed password" | rex field=_raw "Failed password for( invalid user )? (?<user>[^\s]+) from (?<src_ip>\d+\.\d+\.\d+\.\d+)" | bin _time span=10m | stats count as fail_count by _time, src_ip | where fail_count >= 5 | sort - fail_count | head 10 

    - bin과 span을 사용하여 10분 단위로 끊기

- source="*access.log" | eval error=if(status>=400, 1, 0) | stats count as total, sum(error) as error_count by url_path | eval error_rate=round(((error_count / total) * 100),2) | where error_rate >= 15 | sort - error_rate

- access log를 토대로 400~ 오류(클라이언트의 잘못된 요청)와 500~ 오류(서버 오류)의 비율을 계산
- 500~ 오류가 발생되는 비율이 더 높은 이벤트들을 uri_path 기준으로 정렬



```python
#코드 작성
source="*access.log" 
| eval error400=if( Status >= 400 and Status < 500, 1, 0) 
| eval error500=if(Status >= 500, 1, 0) 
|stats sum(error400) as error_count400,sum(error500) as error_count500 by url_path 
| eval error_rate=round(error_count500 / error_count400, 2) 
| where error_rate >=1
```
