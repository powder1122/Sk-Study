## Malware 분석의 중요성
- 하루에 56만개의 새로운 멀웨어가 생성된다.
- 비슷한 유형을 가지고 약간의 수정을 거쳐 배포되는 경우가 많다.
    - 1바이트만 바꿔도 해시값이 바뀌기 때문에 새로운 유형의 파일로 착각을 한다.
    - 기존의 해시값은 이미 악성으로 분류가 되어 사용할 수 없기 때문
- 피해 규모가 크다.
- 실질적으로 파일 유출, 금전적 손해는 악성 파일에 의해서 일어나는 경우가 많다.
    - 파일 관리가 중요
    - 평균 탐지 기간이 207일
- 많은 사람을 투입해서 지속적인 분석이 필요.







### 분석 - Safety First
- 가상 환경이 기본적으로 사용되어야 한다.
- 고립된 환경의 VM에서 분석을 진행해야 한다. 실제 PC로 넘어가지 않도록 주의해야 한다.
- 스냅샷 찍기
- VM을 항상 스냅샷을 찍어두고 파일이 들어왔을 때 VM에서 실행해보고 악성이면 VM을 삭제 후 다시 생성하는 방식을 사용한다.

- 통합 툴 설치

### Static Analysis 
- What we can find:
    - File structure
    - Embedded strings
    - Import/Export functions
    - Packing indicators
    - Digital signatures

- Advantages:
    - Safe - no execution
    - Any file type
    - No evasion

- packing: 악성파일이 있는데, 패킹을 통해 악성 부분을 숨겨 분석을 어렵게 한다. 대부분의 악성 파일은 패킹 과정을 거친다.
- 해킹된 파일은 파일 구조가 다르기 때문에 패킹을 판단하는 것은 어렵지 않다. 그러나 패킹됐다고 반드시 악성은 아니다.

## PE File Structure
- Portable Executable Format
- MZ 패턴이 항상 맨 앞에 온다. 예: 0x5A4D
- 헤더는 메타 데이터에 해당하기 때문에 분석에 사용되지는 않는다.
- 섹션에서 실질적인 데이터가 실행된다.

- import API table (IAT)
- export table
- section


### Section Headers
- Packet Signs: UPX0, .aspack, .themida
- text --> code
- data --> data
- rdata --> read-only data
- rscr --> resources
- reloc --> relocation

### Import Table(IAT)
- 여러 API 관련 정보들이 IAT에 들어가는데, 이것만 분석해도 악성코드 유출을 알 수 있다고 본다.
- 문서 뷰어가 크리에이트 파일에 대한 권한이 있는 등의 경우 의심할 여지가 있다.
- 과도한 권한 주의
- 프로세스의 경우 갑자기 자식 프로세스를 생성하는 경우 찾기 어려워 주의가 필요하다.

- Crypt: 랜섬웨어의 핵심이라고도 볼 수 있는 개념. 파일이 잠기게 된다. (암호화 --> 파일 수정이 가능하다는 뜻.)
    - 100개의 파일을 정확하게 동시에 암호화하는 것은 불가능하다.
    - 따라서, 예전의 랜섬웨어는 오름차순으로, .으로 시작하는 파일부터 암호화를 진행했었다.

2. Dynamic

- 두 방식의 가장 큰 차이점
    - 정적 분석은 실행을 하지 않는 것이다. 
    - 여러 가지 툴을 써서 이 파일이 안전한지 아닌지를 분석가가 분석을 하는데, 분석했을 때 악성이라고 판단했을 때 100% 악성이라 볼 수 없고, 정상이라 판단했는데 100% 정상이라 판단할 수 없다.
    - 정적 분석으로 결과가 애매하면 동적 분석을 실시한다. 
    - 어떤 분석이든 악성이라고 정확하게 판단이 나오면, 다음 작업은 Hash Caclulation이 되어야 한다.
    


### Hash Caclulation
- 해시값 만들기 - 여러가지 방법이 있는데(윈도우 키트, 파워쉘 등), 파이썬 정도는 다룰 수 있어야 한다.

- 해시는 컨텐츠에 대해서 값을 매기기 때문에 이름을 바꾼다고 값이 달라지지 않는다. 그러나, 내용을 수정하면 완전히 달라진다.
