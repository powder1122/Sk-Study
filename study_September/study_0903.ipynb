## Malware 분석의 중요성
- 하루에 56만개의 새로운 멀웨어가 생성된다.
- 비슷한 유형을 가지고 약간의 수정을 거쳐 배포되는 경우가 많다.
    - 1바이트만 바꿔도 해시값이 바뀌기 때문에 새로운 유형의 파일로 착각을 한다.
    - 기존의 해시값은 이미 악성으로 분류가 되어 사용할 수 없기 때문
- 피해 규모가 크다.
- 실질적으로 파일 유출, 금전적 손해는 악성 파일에 의해서 일어나는 경우가 많다.

    - 평균 탐지 기간이 207일

## PE File Structure
- Portable Executable Format
- MZ 패턴이 항상 맨 앞에 온다. 예: 0x5A4D
- 헤더는 메타 데이터에 해당하기 때문에 분석에 사용되지는 않는다.
- 섹션에서 실질적인 데이터가 실행된다.

- import API table (IAT)
- export table
- section


### Section Headers
- Packet Signs: UPX0, .aspack, .themida
- text --> code
- data --> data
- rdata --> read-only data
- rscr --> resources
- reloc --> relocation

### Import Table(IAT)
- 여러 API 관련 정보들이 IAT에 들어가는데, 이것만 분석해도 악성코드 유출을 알 수 있다고 본다.
- 문서 뷰어가 크리에이트 파일에 대한 권한이 있는 등의 경우 의심할 여지가 있다.
- 과도한 권한 주의
- 프로세스의 경우 갑자기 자식 프로세스를 생성하는 경우 찾기 어려워 주의가 필요하다.

- Crypt: 랜섬웨어의 핵심이라고도 볼 수 있는 개념. 파일이 잠기게 된다. (암호화 --> 파일 수정이 가능하다는 뜻.)
    - 100개의 파일을 정확하게 동시에 암호화하는 것은 불가능하다.
    - 따라서, 예전의 랜섬웨어는 오름차순으로, .으로 시작하는 파일부터 암호화를 진행했었다.

### RVA to File Offset
- Critical Calculation
    - 메모리에 올라간 후에는 손쓸 수 없다.
    - 그냥 실행해서 메모리에 적재하면 OS가 어디에서 실행시킬 지 모른다.
    - RVA를 이용해서 실제로 베이스 어드레스 내에 어디에 들어가는지 찾을 수 있다. 파일 오프셋으로 확인 --> 구체화하는 툴이 따로 존재.
    - OS상에서 메모리 위치를 찾기 위한 도구!(RVA)

### Rich Header
- 일반적이지 않은 값이 들어간 헤더
- DB 조회를 통해 똑같은 헤더를 비교
- 새로운 컴파일러, 패킹 횟수 등을 찾을 수 있다.

### Digital Signatures
- 명령어로 시그니처 정보를 받을 수 있다.
- 멀웨어에서 어떻게 활용하는가?
    - 유출된 내역인가 --> 멀웨어에 해당
    - 날짜(만료일)이 지난 경우
    - 자체인증서를 사용하는 경우(CLI에서 직접 생성한 것)
    - 보통 파트가 4개로 구성되는 시그니처, 만약 5개 이상의 파트라면 의심의 여지가 있다.
    - 위 조건에 해당하면 멀웨어로 의심


### TLS Callback

### Resource Extraction
- 파일 자체도 하나의 압축 파일이라고 할 수 있다.
    - 파일 확장자만 zip으로 바꿔도 정보를 확인할 수 있다.
- 어떤 PE 파일이 있고, 그 안의 리소스가 있을 때, 그 리소스를 뽑아내는 방법이 있다.
- 직관적인 분석이 가능해진다.
- PE 파일 분석에는 시간이 걸리고 모호한데, 리소스 추출을 이용하면 이미지 등을 통해 직관적인 분석이 가능해진다.

### API Monitor
- 결국 PE파일을 사용해도 윈도우API를 실행해야 한다.
- 중요 API를 모니터할 수 있는 파이썬 스크립트 등을 통해 모니터링 할 수 있다.
- PC 내에서 발생하는 액션들을 서버로보내 한 번에 감시하여 악성 행위를 확인할수 있다.(EDR 솔루션)
- 개발이 들어가 정적인 분석과는 거리가 좀 있다.




### Anti-Analysis Tricks
- 공격이 나오면 안티 바이러스 벤더에서 방어하고, 다시 공격이 나오기 때문에 그때 그때 테크닉이 바뀐다.
- 악성 코드가 분석가의 컴퓨터나 가상환경, 디버거에서 실행될 때 이를 감지하여 실제 동작을 멈추거나 다른 동작을 수행하게 만들어 분석을 어렵게 한다.

1. 디버거가 존재. 실제 감염을 일으키는 행위를 하지 말아라.
2. 샌드박스 환경일 때, 샌드박스 유저면 실행하지 말아라.
3. 의도적인 지연을 주어라.(안심할 수 있도록)
4. VM 툴즈 등이 설치되어 있으면 돌아가지 않도록 해라.(VM Detection)
    5. CPUID를 보고 VM인지 판단할 수 있다.
    6. 우회가 성공하면 화이트리스트에 들어가기 때문에 위험하다.

### IOC Extraction
- 스팸 메일을 설정하는 것이 블랙리스트고, 그 리스트가 IOC이다.
- 파일의 해시를 가지고 악성 여부를 1차적으로 판단.
    - 바이러스 토탈 등을 사용
- 악성 결과가 없으면 검사 단계를 진행
- 와이어샤크 필요
    - 파일 실행 후 트래픽에 필터를 걸어 확인
- 요소 4개가 많이 사용(정확도 높음)
    - Hash
    - IP
    - Domain
    - URL
- 요소 3개는 컨텐츠가 바뀔 가능성이 있음
    - Registy
    - Mutex
    - File
    - 컨텐츠를 봐야 하기 때문에 있는 그대로 활용하기 어려울 뿐 의미 있는 요소이다.

- 무턱대고 큰 파일을 유출하는 것이 아니라 파일을 쪼개던지 조금씩 보내면서 트래픽을 조절하여 감춘다.

- 패턴을 찾기 위해선 IOC가 중요

- 패킷이 나가는 것이 아니라 들어올 때
    - C&C에서 들어오는 패킷 차단 필요
    - 인바운드에서 공격 툴이 들어오는 경우 주의
    - 공격자는 침투 후 스캔을 통해 취약점을 발견하고 C&C 공격을 실행
    - 잘 찾아주는 것이 중요.





### Literal Movement
- Network Spread
- C&C에서 파일이 감염되면 파일을 실행할 수 있음
- 공격자는 그 후 포트 스캔을 진행하는데, 그 경우 회사의 자산이 노출될 수 있다.
- 각각의 서버마다 방화벽을 설치하면 예방할 수 있지만, 보통 네트워크 단위로 설치하기 때문에 같은 네트워크 대역에 대한 공격에 취약하다.
- 서버 정보들이 스캐닝이 되면 직접 공격에 대한 전략을 자세하게 수립할 수 있어 빨리 찾아야 한다.
- 최선은 모니터링을 통해 빠르게 차단하는 것이다.
- 차단에 실패해도 관제파트에서 인지할 수 있어 포렌식을 진행한다.
    - C&C가 어디서 시작됐는지 포렌식을 진행하여 역으로 추적







### Sandbox Detection
- 샌드박스라는 시스템을 통해 동적 분석을 진행하는 것.
- 동적 분석은 사람이 직접 하지만, 샌드박스 시스템을 이용해서 자동화할 수 있다.
- 이메일에 첨부파일이 있는 경우 안전한지 확인하기 위해 분석 후 문제가 없을 경우 넘겨주려면 1시간이 넘게 걸린다.
- 샌드박스를 사용하면 자동으로 첨부파일을 샌드박스에서 실행하여 결과를 확인할 수 있는데, 1분정도의 시간으로 단축시킬 수 있어 효율적이다.
- 공격자는 샌드박스를 우회하기 위한 방법을 고도화하는 중이다.
    - 이전처럼 엄청 높은 확률로 차단하지는 못하지만 자동화에 더불어 효과적인 방법으로 사용 중이다.
    - 화이트리스트는 명확하게 화이트리스트라고 판단할 수 있는 것들만 넣어야 이후의 보안도 지킬 수 있어 주의가 필요하다.

- nodejs, python 등의 경우 패키지 매니지먼트가 오픈되어 있어 해커들이 악성 코드 등을 많이 올린다. 사용에 주의가 필요

### Powershell Malware

- PE보다는 중요도가 떨어지고, 중요한 커맨드들이 암호화가 되지 않아 분석 시 확인할 수 있다.
- base64로 인코딩하여 유입되는 경우도 있다.
    - 디코딩이 어렵진 않다.
- 다양하게 활용이 될 수 있다.(파일 다운, 실행, 인코딩 등)

- 파워쉘 자체에 대한 보안 검증이 가능!

- VBA 매크로 설정이 되어 있는 경우 악성 코드가 실행될 수 있다.

PDF Malware
