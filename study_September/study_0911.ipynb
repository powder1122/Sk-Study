## 호스트 보호의 필요성
- 보안관제는 호스트 부분은 잘 보지 않다가 EDR이 나오면서 SIEM을 이용해서 호스트에서 어떤 유의한 이벤트가 나왔는지, 이상 징후가 있는지 보기 시작함.
- Rule과 같은 경우 호스트와 관련이 있어 알아야 함.

- 물리 호스트: 하드웨어와 소프트웨어로 구성된 전통적인 컴퓨팅 장치
    - 서버, 단말기, 특정목적 기기, 센서/스위치
- 가상 호스트: 가상화 기술을 기반으로 동작하는 가상 머신

- 공격자의 최종 목표는 호스트, 탐지를 회피하기 위해 끊임없이 변종을 제작함.

- 네트워크 보호가 1차 필터의 역할이라면 호스트 보호는 2차 필터의 역할
- 침해사고로 이어질 가능성 감소




## CPU와 기억장치의 작동 원리, 컴퓨터 운영체제
- 특정 목적을 위해서만 사용되는 전용 컴퓨터 - IoT센서, 산업기기, 의료기기 등
- 다양한 목적 및 용도로 두루 사용할 수 있는 컴퓨터
    - 개인 컴퓨터, 서버, 스마트폰, PDA 등
- 컴퓨터 = 하드웨어 + 소프트웨어

### 하드웨어
- 컴퓨터를 구성하는 물리적 요소
- 중앙처리장치(CPU): 사람의 두뇌에 대응되며 가산,감산 등 연산을 수행
- 기억장치: 연산을 수행하기 위해 필요한 자료를 저장하거나 파일 등을 보관하는 장치
- 입력장치: 컴퓨터가 처리해야 할 자료를 입력하거나 제공하는 장치
- 출력장치: 컴퓨터에 의해 처리된 정보를 표시하거나 출력하는 장치

### 소프트웨어
- 컴퓨터를 구성하는 논리적 요소
- 컴퓨터 프로그램 및 관련된 문서(설계서, 소스코드, 메뉴얼 등)를 통칭하는 것이지만, 현실 세계에서는 컴퓨터 프로그램만을 지칭하는 경우가 대부분
    - 시스템 소프트웨어: 컴퓨터를 제어하고 자원을 관리하는 목적(운영체제, 컴파일러, 시스템 유틸리티 등)
    - 응용 소프트웨어: 사용자가 요구하는 특정 기능을 수행/처리하기 위한 목적의 컴퓨터 프로그램(워드, 엑셀, 게임 등)



### 컴퓨터 프로그램의 실행
- PE(Portable Executable): 윈도우 환경에서 실행가능한 파일의 포맷
- ELF(Executable and Linkable Format): 리눅스 환경에서 실행 가능하며 링크 가능한 파일의 포맷

- 프로그램(보조기억장치) --> (실행) --> 프로세스(주기억장치) --> (명령코드 전달) --> 프로세서(CPU)

- 레지스터: CPU가 요청을 처리하는 데 필요한 데이터를 일시적으로 저장하는 공간
- 범용 레지스터, 인덱스 레지스터, 포인터 레지스터, 세그먼트 레지스터, 플래그 레지스터
    - x64의 경우, 세그먼트 레지스터를 사용하지 않으나 32bit 호환용으로 존재




### ELF
- ELF파일 포맷의 메모리 매핑 순서와 구조

1. ELF 헤더 읽기
2. Program 헤더 읽기
3. Section 헤더 읽기
4. Segment 로드
5. EP(Entry Point)로 점프 --> 낮은 주소의 코드 부분에 적재

## 시스템 취약점과 익스플로잇
- 취약점
    - 시스템 또는 소프트웨어 상에 존재하는 보안상의 약점
    - 시스템, 소프트웨어를 만드는데 사람의 실수가 있을 수 있기에 취약점 발생

- CVE
    - 공개적으로 알려진 컴퓨터 보안 결함 목록
    - 취약점을 가진 다른 도구, 저장소 및 서비스간에 데이터를 공유하는 것이 목적

- CVSS
    - 소프트웨어 취약점이 사이버 보안에 미치는 영향과 위험도를 나타내는 수치
    - 소프트웨어 취약점의 특성과 심각도를 파악하여 대응의 우선순위를 정할 수 있도록 지원
    - FIRST에서 관리

- CWE
    - 소프트웨어와 하드웨어의 보안약점을 나열한 공식 리스트
        - 취약점으로 이어질 수 있는 오류, 결함, 버그 등을 의미
    - 보안 약점의 유형을 분류화하고 목록화 하고 있음


### 익스플로잇
- 시스템 또는 소프트웨어의 취약점을 이용하여 공격자의 의도된 동작을 수행하도록 하는 절차나 명령 등을 의미함
    - 취약점 탐색, 분석, 익스플로잇 개발, 실행, 정보 탈취 및 추가 공격을 포함하기도 함


### 역공학
- 장치 또는 시스템의 기술적인 원리를 구조분석을 통해 발견하는 과정
- 소프트웨어를 분석하는 방법
    - 정적 분석: 프로그램을 실행하지 않고 파일 포맷에 따라 저장된 인코딩 값을 분석하는 방법
    - 동적 분석: 프로그램을 실행하며 발생한 이벤트(행위)를 분석하는 방법

- 정적 분석
    - **소스 코드의 실행 없이** 정적으로 프로그램을 분석하는 방법
    - 파일의 종류, 크기, 헤더 정보 등 파일의 내용을 확인할 수 있음
    - IDA, Ghidra와 같은 Disassembler를 이용해 원본 코드와 구조를 확인할 수 잇음
    - 장점: 
        - 전체구조를 파악하기 쉬움
        - 분석 환경의 제약에서 비교적 자유로움
    - 단점:
        - 난독화 해제가 매우 어려움
        - 다양한 동적 요소를 고려하기 어려움

- 동적 분석
    - 프로그램을 직접 실행 시키면서 분석하는 방법
    - 디버깅 도구를 이용하여 코드 흐름과 메모리 상태 등을 살펴볼 수 있음
    - 장점:
        - 프로그램의 개략적인 행위를 파악할 수 있음
        - 프로그램의 복잡한 동작을 더 명확히 이해하고 식별할 수 있음
    - 단점:
        - 분석 환경 구축의 어려움
        - 실행흐름에 따라 분석되어 구조 파악이 어려움

### Buffer Overflow
- 사용자 입력을 받을 때 크기에 대한 검증 없이 입력받아 메모리의 다른 영역을 오염시키는 취약점
- 할당된 버퍼의 크기보다 큰 값을 입력하는 경우, 메모리의 영역을 넘어서서 return 주소를 덮어쓰는 것을 확인할 수 있음



### Out of Boundary
- 사용자 입력에 대한 검증이 없어 범위를 벗어나는 곳의 데이터를 접근할 수 있는 취약점
- 본래 의도를 벗어나는 요청을 차단하지 않아 발생


### off-By-One
- 경계 검사 시 하나 이상의 오차가 있을 때 발생하는 취약점
- 버퍼의 경계 계산 혹은 반복문의 횟수를 제대로 고려하지 못해 발생
- 예를 들어, 반복문의 경우 인덱스 0번부터 시작하는데 이를 고려하지 않아 한 번 더 수행하는 경우가 있음


# 안티 바이러스

## 바이러스와 악성코드
### 컴퓨터 바이러스
- 컴퓨터 내 다른 프로그램이나 기억장치에 자기 자신 또는 자기 자신의 변형을 복제 및 삽입해가며 감염(확산)시키는 컴퓨터 프로그램
- 정보통신망 환경에서는 네트워크 범위로까지의 확산 가능하며, 시대가 지나면서 바이러스의 기능도 점차 진화하는 경향
- 컴퓨터 바이러스의 세대
    - 1세대: 원시형 바이러스
        - 부트 바이러스
        - 파일 바이러스
    - 2세대: 암호형 바이러스
        - 슬로우, 원더러, 캐스케이드, 버글러 등
        - 바이러스 코드를 암호화하고 복호화 논리와 키를 같이 붙이는 방식으로 감염
    - 3세대: 은폐형 바이러스
        - 일정 기간 잠복기를 갖는 바이러스
        - 브레인, 조쉬, 512 바이러스, 4096 바이러스 등
    - 4세대: 다형성 바이러스
        - 바이러스가 자신의 코드를 지속적으로 변형해가며 전파시키는 바이러스
        - 바이러스 고유의 패턴을 변경하여 탐지 원리를 우회
    - 5세대: 매크로 바이러스
        - 마이크로소프트 오피스 제품군에 있는 자동화를 위한 편의 기능인 매크로 기능을 악용하는 바이러스
        - 액셀-라룩스, 워드 콘셉트, 와쭈, 멜리사 바이러스 등



### 악성코드(Malware)
- 전통적인 바이러스의 특성을 갖지 않으면서도 악성 기능을 수행하는 컴퓨터 프로그램이 등장.
- 종래의 바이러스 개념을 포함하며, 악의적 목적을 가진 모든 종류의 실행 가능한 컴퓨터 프로그램을 총칭하는 개념
- 종류
    - 웜
        - 자체적으로 실행되면서 네트워크를 통해 다른 컴퓨터로 전파가 가능한 악성코드
    - 트로이목마
        - 정상적인 프로그램처럼 위장한 악성코드
        - 사회공학적 방법으로 유포되는 경향이 있음
        - 불법 소프트웨어나 불법 파일에 포함되는 경우도 상당수
        - 백도어 조성이 주 목적으로 사용됨
    - 스파이웨어
        - 개인이나 기업의 정보를 몰래 수집하는 정보 절취를 목적으로 사용되는 악성코드
        - 키로거 유형: 사용자의 키 입력이나 비밀번호 입력을 절취
        - 원격 제어 유형: 컴퓨터 내 파일을 절취하거나 웹 캠이나 마이크를 활성화하는 등 컴퓨터 원격 제어
    - 애드웨어
        - 귀찮을 정도로 광고를 보여주는 등의 행위를 하는 악성코드
        - 불필요한 파일 및 서비스가 발생되면서 컴퓨터의 성능 저하 원인으로도 작용
    - 랜섬웨어
        - 컴퓨터 내 파일들을 암호화하거나 화면을 잠그고 이를 풀어주는 대가로 금전을 요구하는 악성코드
        - 금전 목적의 공격 다수
    - 루트킷
        - 대부분의 악성코드는 사용자 모드에서 동작하여 안티 바이러스에 의한 탐지가 용이함.
        - 루트킷은 커널 모드에서 동작하여 컴퓨터의 모든 권한을 장악하여 안티 바이러스에 의한 탐지를 회피하여 치명적인 영향을 초래할 가능성
            - 고도의 기술이 요구됨

## 안티 바이러스의 이해
### 안티 바이러스
- 컴퓨터 바이러스와 악성코드를 탐지하고 방어하는 호스트 기반 정보보호시스템

- 설치 위치
    - 안티 바이러스의 핵심 기능은 커널 모드에서 동작하며, 사용자가 조작 가능한 환경은 사용자 모드에서 제공

- 커널 모드
    - 드라이버 설치
    - AV에게 운영체제에 준하는 강력한 권한을 부여하기 위함
    - 루트킷 등의 바이러스 및 악성코드까지 탐지하기 위함
    - 공격자가 AV를 강제 종료하기 전에 우선권을 선점하기 위함
- 사용자 모드
    - UI 제공
    - 과거에는 커널 모드에서 실행되도록 했으나, 바이러스/악성코드가 악용하는 사례가 빈번해지면서 사용자 모드로 분리

- 주요 기능
    - 실시간 감지
        - 사용자가 필요할 때마다 컴퓨터 전체 영역 또는 사용자가 선택한 영역에 대해 검사 기능을 실행하는 방법으로 진단.
        - 운영체제에서 발생되는 모든 행위를 모니터링
    - 자가보호
        - 안티 바이러스의 실시간 감시 기능에 안티 바이러스 스스로를 보호하는 자가 보호 기능을 같이 포함하는 경향
        - 바이러스가 AV를 강제종료 또는 삭제하는 것을 방지하기 위함
    - 진단
        - 패턴(시그처 기반 탐지)
            - 알려진 바이러스/악성코드가 갖고 있는 고유한 패턴을 시그니처 데이터베이스에 등록
            - 알려진 바이러스/악성코드의 변종이나 알려지지 않은 바이러스/악성코드를 탐지하지 못함
            - 문자열
            - 식별자
            - 파일 해시 값
        - 휴리스틱 기반 탐지
            - 대부분의 바이러스/악성코드가 가질 수 있는 파일의 구조를 규칙화
            - 규칙화 된 휴리스틱 데이터베이스와 검사 대상 파일의 구조를 비교하며 확인
            - 도입 초기
                - 오탐/오진율이 높았음
            - 현대
                - 머신러닝 기술의 적용으로 오탐/오진율이 낮아짐
        - 루트킷 탐지
            - 운영체제의 커널에서 불러온 드라이버 목록이나 발생된 이벤트 목록을 분석하는 방식으로 루트킷의 진단 가능
            - 드라이버 목록 중 특이한 이름이나 비정상적인 위치에서 불러온 드라이버를 식별
    - 치료
        - 진단한 결과를 토대로 컴퓨터를 감염 이전으로 복구하는 기능
        - 바이러스/악성코드의 원인 파일을 제거
        - 바이러스에 감염되어 손상된 파일들을 원래 상태로 복구
        - 바이러스/악성코드로 인해 생성된 부산물을 제거




## 바이러스/악성코드의 탐지 원리
- 컴퓨터 내 파일 확인
    - 알려진 악성코드가 갖고 있는 고유한 패턴(문자열, 식별자, 해시 값)을 비교해가며 컴퓨터 내 동일한 패턴을 갖는 파일의 존재 여부 확인
    - 바이러스에 감염된 파일은 원래의 파일에 변형이 발생된 것이므로 이를 추적/검사하는 방식으로 감염 사실 확인

- 운영체제 내 설정값 확인
    - 운영체제의 특정 설정 팡리이나 설정 값을 변형하는 경우가 있으므로 이를 확인
    - Windows 운영체제의 경우 레지스트리, 시작 프로그램, 서비스 목록에 특정 값을 생성하는 경우가 있으므로 이를 확인



### 바이러스/악성코드 감염 여부 확인 방법
- 운영체제의 네트워크 상태 확인
    - 컴퓨터는 외부와의 통신이 필요한 경우가 발생하면 네트워크 포트를 개방
    - 특정 악성코드는 알려진 포트 이외 자신을 특정할 수 있는 번호의 포트 번호를 사용하는 경향이 있어 이러한 포트 번호가 개방된 경우 악성코드 감염을 의심
    - 알려진 포트 번호라 하더라도 자의에 의해 개방한 사실이 없는 포트 번호라면 악성코드 감염을 의심

- 운영체에의 실행 프로세스 확인
    - 컴퓨터 프로그램이 실행되면 운영체제에 해당 프로그램에 대한 프로세스 생성
    - 악성코드도 프로그램이므로 프로세스 생성
        - 정상 프로세스와 유사한 이름의 프로세스도 존재
    

# 사이버위협정보(CTI)
## 목차
- 정보와 사이버위협정보
- 내부 위협 헌팅
- 외부 위협 헌팅


## 정보와 사이버위협정보
### 정보의 이해
- 정보기관과 정보
    - 최근 사이버보안 업무는 국가 차원의 사이버안보 업무로 그 범위가 확장되면서 정보기관에서 주도적으로 관여하는 양상
    - 정책적 측면이 반영된 거버넌스 문제로 접근

- 사이버위협정보(CTI)의 개념은 전직 정보기관 요원들이 정보기관에서 일했던 방식을 상업화하면서 등장

- 자료
    - 평가되거나 가공되지 않은 단순한 사실
    - 날 것 그대로의 생 자료(Raw Data)를 의미
- 첩보
    - 목적성을 갖고 의도적으로 수집되었으나 평가되거나 가공되지 않은 자료 또는 자료들의 집합
    - 전산학/컴퓨터공학 및 정보공학에서는 정보(Information)로 저장
- 정보
    - 목적성을 갖고 의도적으로 수집되어 평가, 분석, 검증되어 의사결정에 활용 가능하며 비밀성을 가진 지식
    - 첩보를 처리/가공하여 만들어진 의사결정에 필요한 새로운 지식으로, 정보사용자/소비자의 지식수준, 관심, 시간, 시력 등을 감안한 '맞춤' 지식

- 정보의 효용과 한계
    - 정보기관은 NCND(Neither Confirm Nor Deny)원칙에 따라 정보성공 사례를 공연하게 홍보하지 않는 경향이 있어 알려진 효용 사례가 많지 않음
    - NCND: 어떠한 사안에 대해 긍정도 부정도 하지 않는다는 원칙
    - 정보기관은 헌법에 따라 분립된 기관이 아닌 행정부에 귀속되는 기관적 특성 탓에 정책결정권자에 의해 악용되어, 이로 인한 정보실패 야기 가능성이 존재
        - 정보실패
        - 정보의 정치화
        - 집단사고
        


### 사이버위협정보
- 협의의 정의: 사이버정보 중 사이버보안에 관한 기술정보 및 기술분석정보로 한정되는 범위
- 광의의 정의: 사이버위협정보의 협의의 정의에 더해 인물, 단체, 동향 등 비기술정보를 포함하는 범위

- 오늘날 사이버위협이 APT 형태로 지능화/지속화/고도화되면서 단편적인 이벤트만을 들여다보는 것만으로는 대응하기에 곤란
- 공격자와 관련된 단서들을 확보하면 공겨자를 추적/식별하는 프로파일링도 가능


### 사이버 킬 체인
- 사이버 공격을 적극적으로 방어하기 위한 대응 개념을 구조화한 것
- 공격 절차: 정찰 --> 무기화 --> 유포 --> 악용/권한탈취 --> 설치 --> 명령 및 제어 --> 목적 달성
- 방어 모델: 탐지/거부/교란/약화/기만/파괴



## 내부 위협 헌팅
### 위협의 이해
- 보안 위험과 구성요소

- **위험**
    - 내/외부로부터의 위협이 보안상 허점인 취약성을 악용하여 보유하고 있는 자산에 손실을 초래할 수 있는 잠재적 가능성
    - 위협: 자산에 손실을 초래할 수 있는 잠재적인 원인이 되는 행위 또는 행위자
    - 취약성(취약점): 위협에 의해 악용될 수 있는 미흡한 속성 또는 허점으로, 취약점이라고도 지칭
    - 자산: 조직/체계에서 보호해야 할 가치가 있는 대상으로, 일반적으로는 정보자산에 국한
- 위험은 완전히 제거될 수 없으며, 위험관리를 통해 수용 가능한 수준으로 감소/유지


### 위협 헌팅
- 알려지지 않은 위협을 탐색하기 위한 방법
    - 내부 위협 헌팅
    - 외부 위협 헌팅

- 위협 헌팅의 효용성
    - 최근 보안관제 트렌드는 수동적인 모니터링 수준을 탈피하여 능동적인 대응 능력을 확충하려는 추세
    - 위험은 완전히 제거될 수 없으므로 위험을 완화하기 위한 방법으로 가능한 한 위협을 음지에서 양지로 올려 가시화할 필요
    - 
- 위협 헌팅의 한계
    - 고도의 기술 수준이 요구되는 만큼, 충분한 수준의 인프라와 전문인력이 필요
    - 상당한 비용이 발생하는데, 국내 기업은 인색한 경향
    - 국가기관 이외에는 대기업이라 하더라도 자체 위협 헌팅 조직을 구비하는 경우가 매우 드문 실정
    - 위협 헌팅 전문 업체를 통해 사이버위협정보를 구매할 수 있으나, 정보의 특성상 품질 수준을 정량적으로 측정하기 어렵기 때문에 가격과 품질이 천차만별이다.
    - 내부 위협 헌팅의 경우 NDR, EDR등의 정보보호체계를 활용하면서 개인정보 논란이 발생될 가능성
        - 국가기관의 경우 외부 위협 헌팅에 집중하는 경향
    


### 내부 위협 헌팅
- 내부 위협 헌팅의 필요성
    - 내부 위협 헌팅을 통해 조직에서 소관하는 정보통신망 및 정보시스템 내 정보자산들을 대상으로 위협을 탐색
    - 정보자산 내 존재할 수 있는 위협을 조기에 식별하고 가시화함으로써 선제적인 대응 능력과 침해사고를 사전에 예방할 가능성을 향상하는 효과

- 주요 기법
    - 호스트 기반 헌팅 기법
    - 통신채널 헌팅 기법

- 호스트 기반 헌팅 기법
    - 침해지표(IoC): 침해사고 흔적을 탐지,분석하기 위해 사용되는 지표
    - 주요 표준: IODE, CybOX, OpenIOC

    - 주요 수집항목: 프로세스 이벤트, 서비스 이벤트, 파일 기록 이벤트, 레지스트리 이벤트, DNS 통신 이벤트, 네트워크 이벤트

    - 다양한 도구를 활용할 수 있다.
    - 정보보호시스템 중 EDR은 호스트 내 침해지표를 자동화된 방식으로 수집하는 기능 탑재!

- 통신채널 헌팅 기법
    - 공격자는 자신이 만든 악성코듣가 탐지되는 것을 가능한 한 회피하기 위해 악성코드 내 많은 기능을 포함하지 않고 단계적으로 공격
        1. 사회공학적 방법을 통해 문서 파일 등으로 위장한 트로이목마 배포
        2. 트로이목마가 조성한 백도어를 통해 추가 악성코드 배포
        3. 추가 악성코드에 의한 자료유출 또는 시스템 손상 발생
    - 공격자는 지속적인 통신을 위해 백도어를 통한 통신채널 구성
        - 정상 데이터를 가장한 악성 데이터 전송
        - 악성 데이터를 최대한 작게 쪼개어 전송(합치지 않으면 탐지하기 어려움)
        - 통신 프로토콜의 구조적 특성을 악용하여 악성 데이터 전송
    - 공격자의 통신채널 구성 예
        - XOR 연산, Base64 등을 활용한 부호화된 악성 데이터 전송
            - 대부분의 정보보호체계가 부호화된 데이터를 복호화하지 않는다는 특성을 악용하는 방식
        - SSL을 이용한 암호화된 악성 데이터 전송
            - 암호화된 데이터는 복호화 키가 없으면 확인할 수 없다는 특성을 이용한 방식

        

## 외부 위협 헌팅
- 전출처 정보: 모든 출처의 정보가 통합된 정보

- 공개출처정보
    - OSINT
    - 인터넷, 언론매체, 관보/공시 등 합법적으로 접근 가능한 공개된 환경에서 수집한 정보
- 인간출처정보
    - HUMINT
    - 협조자 등 사람을 통해 수집한 정보
- 기술출처정보
    - TECHINT
    - 누구나 접근하기 곤란한 폐쇄적인 환경에 대해 기술 수단ㅇ르 통해 수집한 정보
        - 신호정보: SIGINT, 통신 도/감청 및 암호해독,해킹 등으로 수집
        - 영상정보: IMINT, 사진촬영/항공정찰/위성영상 등으로 수집

- 공개출처정보(OSINT)와 외부 위협 헌팅
    - 외부 위협 헌팅은 법에 의한 단서가 전제되지 않은 경우 출처 범위를 공개출처로 한정
    - 외부 위협 헌팅을 통해 획득되는 정보는 사이버정보(CYINT)와 사이버위협정보(CTI)에 해당

- 필요성
    - 공개된 출처에 존재하는 위협을 탐색하거나 사이버위협정보 공유체계에 공유된 위협을 인지
    - 조직 내 유입되지는 않았지만 유입 가능성이 높은 위협을 조기에 식별하고 파악함으로써 선제적인 대응 능력과 침해사고를 사전에 예방할 가능성을 향상하는 효과

### 인터넷 검색엔진 기반 헌팅 기법
- 크롤링: 수집
- 스크래핑: 가공/처리
- Google: 범용 검색엔진(웹 사이트 관리자 페이지 노출, 개인정보 목록 파일 노출 등)
- Shodan: 전용 검색엔진(라우터 관리자 로그인 노출, CCTV 영상 화면 노출 등)

### 허니넷 기반 헌팅 기법
- 허니팟: 비정상적인 접근을 탐지하기 위해 의도적으로 설치해둔 덫
- 허니넷: 다수의 허니팟들을 정보통신망으로 구성한 형태
- 공격자 유인 및 기만

### 사용자 행위 모사 헌팅 기법
- 허니넷 기반 헌팅 기법과는 반대로, 실제 사용자의 행위를 모사하는 호스트들을 활용하여 인터넷 공간을 탐색하면서 의도적인 악성코드 감염을 유도하는 기법
- 검색엔진과 유사하게 크롤링/스크래핑 기술을 활용
- 사용자의 행위를 모사하기 위해 가상화 기술, 행위 모사 기술 등을 추가적으로 활용
### 다크웹 헌팅 기법
- 익명성이 보장되는 비밀 네트워크 영역으로 특정 브라우저를 통해 접근
- 불법 콘텐츠가 대부분
- 크롤링과 스크래핑 기술을 활용
- 접속 자체는 불법이 아니다.
- 익명성이 완전히 보장된다고 볼 수 없고, 불법 행위는 지속적으로 모니터링 중
- 획득할 수 있는 외부 위협 유형
    - 악성코드 샘플
    - 취약점 유발 코드
    - 유출된 기업 내부정보, 개인정보 등


### 사이버위협정보의 공유
- 국가 주도 정보공유시스템(국가정보원)
    - NCTI
    - 운영주체: 국가정보원
    - 국가/공공 분야 보안운영센터에서 정보공유를 목적으로 사용하는 시스템
    - 국가사이터보안센터를 중심으로, 각급기관 보안운영센터에 전용회선으로 설치 및 운영 중

- 인터넷 기반 정보공유시스템
    - KCTI
    - 운영주체: 국가정보원
    - NCTI는 국가/공공 전용이기 때문에 민간 사용은 제한
    - 민간의 핵심기반시설 및 방산업체 등과의 정보공유를 위한 인터넷 기반 KCTI 구축/운영 중

- 국가 주고 정보공유시스템(한국인터넷진흥원)
    - C-TAS
    - KISA
    - 민간 분야 사이버위협정보 공유를 위해 한국인터넷진흥원 주도로 구성한 사이버위협정보 공유체계
    - 공유형 C-TAS: API를 활용하여 KISA와 각 회원사 간 양방향으로 실시간 정보공유
    - 개방형 C-TAS: 누구나 가입할 수 있으며 KISA에서 제공하는 웹사이트를 통해 취약점 정보 등 제공

- CTI의 생산 과정
    - CTI를 자체적으로 생산하는 조직은 정보기관의 정보순환주기를 준용하여 각 조직의 실정에 맞는 정보순환주기 구성
    - CTI의 개념이 태동할 당시는 정보수집 자체가 중요하게 여겨졌으나, 최근에는 민간 주도 정보시장이 활성화되면서 정보분석 및 생산 역량을 강조하는 추세
    - 정보시장을 통해 구입한 정보에만 의존할 수 없다는 한계 인식

- CTI 순환주기
    - 기획: 수집 대상 정보를 정의하고 요구사항을 구체화
    - 수집: 외부 위협 헌팅, 내부 위협 헌팅 등의 방법으로 수집하거나 정보시장을 통해 구매
    - 분석: 수집된 정보를 다양한 방식과 기법을 활용하여 분석
    - 생산: 분석된 내용을 사이버위협정보 표준, 사이버위협정보 보고서 등의 형태로 생산
    - 공유: 생산된 결과물을 사이버위협정보 공유체계를 통해 배포

## 인공지능과 머신러밍의 이해
- 인공지능: 컴퓨터로 구현한 지능 또는 이와 관련한 전산학의 연구 분야
    - 문제해결능력
    - 학습능력
    - 범용성

- 범위
    - 인공지능 > 머신러닝 > 딥러닝 > 생성형 인공지능(LLM, 비지도학습GAN, 변분 오토인코더VAE)


### 머신러닝의 이해
- 머신러닝 프로세스
    - 구체적인 방법에 대한 프로그래밍 없이 학습을 통해 컴퓨터 스스로가 학습할 수 있게 하는 방법
    - 지도학습 --> 정/오탐 예측이 목적(예측)
    - 비지도학습 --> 이상행위 탐지가 목적(이상치)

- 머신러닝 프로세스 
1. 데이터 수집
    - 대상: 침입방지시스템(IPS), WAF, TMS 등
    - 관제 요원들이 이벤트를 분석할 때 무엇을 보고 판단할까
2. 학습데이터 생성
    - 어떤 데이터가 어떤 결과를 내는지 Labeling 필요
    - 패턴을 알려주는 과정 - 머신러닝은 그에 따른 결과를 도출
    - 탐지된 보안 이벤트에 대해 정/오탐 여부를 라벨링
    - 공격 의도, 유효성, 사고 이관 등 각 관제 환경을 고려하여 라벨링
3. 데이터 전처리
    - 제대로 이용하기 위해 데이터 가공
    - 원본 데이터를 머신러닝에 적용하기 좋게 하는 과정
4. 피처 추출 및 선택
    - 데이터에서 줒ㅇ요한 특징을 추출하는 과정
    - 원하는 결과에 따라 피처에 대한 선택은 달라진다.
    - 보안솔루션마다 이벤트 결과 형식, 탐지 방식, 결과에 대한 의도 등의 특징을 표현
    - 피처 추출 방법은 데이터에 대한 통계적 기술, 도메인 지식에 근거한 방법을 통해 생성
    - 전처리된 보안 데이터는 적절한 알고리즘 사용
    - 전처리된 WEB 문자형(URL) 데이터는 CNN을 이용하여 Text에서 Token 단위에 대해 부분적인 특징 추출
5. 모델 평가
    - 검증된 학습 기법 수행
    - 보안 분석가 피드백 반영을 통한 지속적인 모델 개선

## SOAR와 XDR
- 4세대 보안관제: 인공지능보안관제
### SOAR의 이해
- SOAR: 보안 오케스트레이션/자동화/대응
- 인공지능 기술을 기반으로 다양한 사이버위협에 대한 대응 수준을 자동으로 분류하고 표준화된 업무 프로세스에 따라 유기적인 협력을 지원하는 보안관제시스템

### 도입 시 선결조건
- SIEM의 도입 및 운용. SIEM에 SOAR를 Add-On으로 활용
- 보안운영 거버넌스 정립
- 보안운영 프로세스의 표준화
    - NIST CSF 등 표준 프레임워크에 기반을 둔 포괄적인 접근방식 채택
- 제대로 도입하기 위해 선결조건이 반드시 충족되어야 함


### SOAR 구성요소
- SOA: 보안 오케스트레이션 및 사동화
    - 다양한 형태의 컴퓨팅 장치 및 소프트웨어 등 도구와 도구 사이의 업무 흐름을 조율하고 자동화
    - 특히, 보안운영에 관한 조직 내 표준화된 업무 프로세스를 플레이불으로 구성하여 단순 반복 업무 등을 자동화
- SIRP: 보안사고 대응 플랫폼
    - 보안사고 대응에 관한 조직 내 표준화된 업무 프로세스를 자동화
    - 침해시도/침해사고 발생 시 사고 유형별 정해진 프로세스에 따라 통합관리 및 처리 수행
- TIP: 사이버위협정보 플랫폼
    - 사이버위협정보(CTI)를 수집/분석/공유하고, 이를 토대로 분석된 위협정보를 정보보호체계 등과 연계하여 사전 대응능력 향상
    - 타 조직들과의 보안관제 공동 대응체계 마련에 도움

### XDR의 이해
- SOAR의 한계
    - 수집되는 보안 정보 및 이벤트 기록의 수준에 변함이 없다면 SOAR의 개념을 실제 실현하는 것은 이상적인 이야기에 불과

- XDR(확장된 탐지 및 대응)
    - 탐지/차단되지 못한 이벤트라 하더라도 유의한 침해사고/침해시도였을 가능성이 존재하는 만큼, 모든 유형의 위협에 대한 가시성을 확보하는 것이 중요
    - 총체적인 보안 위험에 대한 가시성을 확보하여 능동적으로 대응하려는 개념

### XDR의 구성요소
- NDR
    - 네트워크 탐지 및 대응
    - 네트워크 상에서의 보안 위협을 탐지하고 대응하기 위해 가시성을 확보하는 기술이자 이를 토대로 한 정보보호체계
    - 모든 네트워크 트래픽을 수집/보관하며 심층 분석할 수 있게 지원
- EDR
    - 사용자 단말기 상에서의 보안 위협을 탐지하고 대응하기 위해 가시성을 확보하는 기술이자 이를 토대로 한 정보보호체계
    - 사용자 단말기의 모든 이벤트를 수집/보관하며 심층 분석할 수 있게 지원
- SIEM
    - SIM + SEM, 보안 정보 및 이벤트 관리
- UEBA
    - NDR과 EDR을 비롯하여 다양한 정보보호체계를 통해 수집된 보안 정보 및 이벤트 기록을 토대로 기계학습(머신러닝)을 적용하여 사용자의 정상적인 활동과 이상 행위를 비교분석하는 기술
    - 로그인 시도, 파일 생성/접근 패턴, 네트워크 사용량 등
    - NDR, EDR이 갖고 있는 자체 분석 기능은 대부분 개별 패킷 또는 개별 이벤트/행위에 초점을 두는 경향
    - UEBA를 위해 이외에도 NAC, DRM, DLP 등 다양한 정보보호체계를 추가로 통합하는 방식으로 구성
    - UEBA의 구현을 위해 SIEM을 활용

- 플레이북 기반 자동화 워크플로우/실시간 정보공유로 대응 시간이 빠르게 단축되었다.

## 머신러닝의 보안관제 적용
- 현 보안에도 인공지능이 많이 적용
- 지능화 고도화 됨에 따라 시그니쳐 기반의 침입탐지 한계
- 머신러닝 기반의 침입탐지 적용으로 정탐 증가 및 오탐 감소

- Payload 분석을 통한 정/오탐 이벤트 식별

- 보안과 관련한 다양한 이해관계자들이 필요
- 보안 이벤트에 대한 최적의 ML 결과를 얻기 위해 데이터사이언스 방법론에 따라 예측 모델을 테스트 함

### 현 시점에서 인공지능 적용을 방해하는 요인
- 학습 데이터 부족 및 품질 이슈
- 또한, 머신러닝을 사이버 보안에 적용할 때 가장 큰 이슈는 학습 데이터 부족
    - 데이터를 만드는데 너무나 많은 시간과 전문 인력 필요
    - 보안 데이터의 특성 상 기록되어 있지 않거나 오픈 되지 않은 경우가 많음

## 이상행위 탐지를 위한 보안 데이터 수집 및 분석
- 오용탐지와 이상탐지
    - 오용탐지: 시그니처 기반 탐지
    - 이상탐지: 정상 행동 패턴을 벗어난 활동 탐지

1. 보안 데이터 수집
2. 비정상 행위 데이터 추출
3. 비정상 행위 데이터 분석
4. 파이썬 환경 설정
5. CSIC
6. CSIC 데이터 분석-정상 및 비정상 요청 분류
