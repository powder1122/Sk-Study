## 2. 보안관제의 이해

### 2-1. 보안관제의 개념
- 보안관제란 조직의 정보기술 자원 및 보안 시스템을 안전하게 운영하기 위하여 사이버 공격 정보를 탐지 및 분석하여 대응하는 일련의 업무와 사전예방 및 관제시스템 운영에 관한 업무를 말한다.
- 보안관제 : Security Monitoring, Security Monitoring & Control
- 관제 : 국가나 공항 따위에서 필요에 따라 강제적으로 관리하여 통제하는 일
- Monitoring : 컴퓨터의 프로그램 수행 중 일어날 수 있는 여러가지 오류에 대비하기 위한 감시활동
- 1999년 민간보안관제서비스 이 후 현재까지 민·관·군에서도 ‘보안관제’라는 용어를 도입하여 사용 중
- 협의의 개념 : 모니터링, 즉 사이버공격을 탐지하는 활동
- 광의의 개념 : 모니터링과 콘트롤, 즉 탐지·분석·대응까지 포함하는 일련의 활동

### 보안관제 기본 원칙
- 무중단의 원칙
    - 사이버 공격을 실시간으로 신속하게 탐지 / 차단 하기 위해서 24시간 365일 중단 없이 보안관제 업무를 수행

- 전문성의 원칙
    - 보안관제업무 수행을 위해서는 사이버공격 탐지시스템 등 보안관제에 필요한 시스템과 함께 정보시스템 및 네트워크 이론을 포함한 프로그램 분석, 포렌식, 해킹기술 등 다양한 방면에 전문지식과 경험, 노하우를 가진 전문 인력이 매우 중요

- 정보공유의 원칙
    - 범 국가차원에서 사이버 공격을 철저하게 탐지 / 차단 / 대응 하기 위해서는 관계 법령에 위배되지 않는 범위 내에서 보안관제 관련 정보가 신속하게 공유

- 탐지된 이벤트의 정오탐 유무 판정
    - 오탐의 가능성? → 탐지되는 이벤트에 대한 이해와 위협행위에 대한 이해
    - 확실한 정탐인가? → 이벤트가 탐지되었다면 보안솔루션에서 차단되어 확인 불가
    - 정오탐 판단의 근거는 무엇인가?
        - 단순 문자열 일치, IP 대역 일치, 정상 서비스
    - 탐지된 보안 이벤트의 정탐과 오탐을 결정하는 것은 보안 관제 근무자!

- 초동분석
- 이벤트, 트래픽, 홈페이지 등 모니터링 중 탐지된 비정상 건에 대해 1차 분석을 통해 침해사고 여부를 판별하는 업무
1) 로그 분석 – 탐지된 비정상 이벤트 관련 로그를 검색하여 연관성을 분석함
2) 패킷 분석 – 비정상 이벤트, 트래픽 관련 네트워크 패킷을 수집·분석함
3) 과거이력 조회 – 비정상으로 탐지된 건 관련 대응 이력을 검색하여 담당자 문의 등 후속 Task의 수행 여부를 결정함

- 정책 관리 필요
    - 정책관리
        - 모니터링을 위한 탐지정책을 Category 기준을 고려하여 분류하고, 정의하는 등 정책 전반의 체계를 관리하는 업무
    - 정책체계 관리 – 관제 모니터링을 위한 정책체계를 관리함

    - 탐지정책 관리
        - 사고 분석 결과, 신규 취약점 등을 탐지정책으로 개발, 정탐율 향상을 위한 수정, 불필요한 정책 삭제 등 모니터링을 위한 탐지정책을 관리하는 업무
        - 탐지정책 개발, 수정, 삭제 등 정보보안환경에 최적화하여 적용함
    - 차단정책 관리
        - 모니터링, 사고 분석 등을 통해 차단 설정이 필요한 경우 단위 보안장비 등에 차단 정책을 적용하는 업무
        - 차단정책 관리 – 보안장비 별 적용할 유해정보(블랙IP/URL/메일주소)를 추출하여 차단 적용함

### 침해예방
- 모의훈련
    - 침해사고 발생시 대응력 향상을 위해 사고 상황을 계획하고 발생시켜 상황별 대응하고, 개선점을 도출하여 보완하는 일련의 활동

- 정보공유
    - 국내외 보안동향 등 정보보안 분야 정보를 수집하여 공유하는 업무

- 침해예방 점검 – 정보노출/감염
    - 고의성 유무와 무관하게 외부에 노출되면 안되는 기관 정보가 노출되어 있는지 점검하여 발견된 정보의 삭제를 목적으로 수행하는 업무

- 침해예방 점검 – 웹 서버
    - 고의성 유무와 무관하게 기관 내부시스템에 비인가된 파일이나 신규 취약점 존재 여부를 점검하여 제거 조치하는 것을 목적으로 수행하는 업무

### 보안운영
- 보안운영은 운영대상 정보시스템(주로 정보보호시스템)의 가용성 유지를 위해 자산을 관리하고 운영하는 업무
- 세부적으로 자산관리, 장애관리, 작업관리, 설정관리, 예방점검, 백업관리, 기술지원 등을 수행

- 장애관리 – 장애대응/재발방지
    - 운영대상 정보시스템의 장애 발생시 대응 및 재발방지대책을 수립하여 가용성을 유지하는 업무

- 작업관리 – 교체/변경/패치/기타작업
    - 보안운영 중 가용성 및 성능유지를 위해 작업이 필요한 경우 계획을 수립하여 작업을 수행하는 업무

- 성능관리 - 가용성 모니터링
    - 운영대상 정보시스템의 가용성 이상유무를 인지하기 위해 모니터링을 실시하고 이상징후 발생 시 대응하는 업무

- 성능관리 – 정기/긴급점검
    - 안정적인 보안운영을 위해 정기점검을 실시하고, 필요시 긴급점검을 수행하고, 이상징후 발견 또는 예방을 위해 추가작업을 수행하는 업무

- 재해복구관리 – 백업관리 
    - 재해, 장애 등의 사유로 운영대상 정보시스템 중단발생시 신속한 정상화를 위한 시스템/데이터 백업을 수행하고, 주기적으로 관리하는 업무

- 보안운영 - 장애관리
    - 보안시스템 의 효율적인 장애 관리 및 ESM 연동대상 장비에 대한 형상 관리로 시스템의 안정성을 도모하고 정규화된 운영 프로세스 및 Issue Lists를 통해 보안운영 최적화

## 3. 보안관제 기술
### 3-1. 보안관제 탐지/방어기술
- 패턴기반 탐지기술
    - 탐지속도가 빠르고 정확성 높음
    - 사전에 분석된 공격형태만 탐지 가능
    - IDS/IPS
    - WAF
    - DDoS
    - Anti-Virus
- 행위기반 탐지기술
    - 통계기반의 비정상 행위 탐지에 효과적
    - 통계적 특징이 불명확한 공격 탐지 불가능
    - IDS/IPS
    - WAF
- 상관분석 기술
    - 공격 시나리오 기반 탐지가 가능하고 탐지의 정확성 향상
    - 상관분석 역량 및 시스템 성능 부족
    - ESM
    - SIEM
    - TMS
- 세 가지 기술을 유기적으로 결합시키는 것이 중요

- 패턴기반방어: 사전에 정의한 패턴과 동일한 태래픽을 차단하는 기술
- 규칙기반방어: 취해야 할 특정 행위를 기준으로 차단하는 기술

### 3-2. 탐지패턴 활용
- 특정 바이러스나 악성코드 등 특정 패킷이 네트워크를 통과 할 때 탐지하기 위해 제작한 일종의 문구(시그니처)이며, Rule 이라고 부름

- Snort Rule: 보통 alert 혹은 drop을 많이 사용

- 일반적인 탐지패턴
    - 특정 탐지 패턴을 입력하면 해당 문자가 있으면 모두 탐지
- HEXA CODE 방식
    - 헥사코드 16진수로 나타내는 인코딩 방식 문자열 탐지패턴이란 점은 동일하지만, 패턴으로 작성된 글자가 숫자로 표현

## 4. 보안관제 시스템
### 4-1. 보안솔루션 분류
- step1~4로 나누어져 있고, 각 단계에 맞는 솔루션 분류
    - step1: 소규모 조직, 조직을 안전하게 운영하기 위한 최소한의 보안솔루션
    - step2: 중소기업, 중소 규모의 조직에서 효과적인 보안 체계를 갖추기 위한 보안 솔루션
    - step3: 대기업, 대규모 조직에서 효율적으로 관리 및 통제하기 위한 보안 솔루션
    - step4: 기밀정보를 다루는 주요 조직, 군사, 주요 정부기관, 핵심 사업부 등 고도의 보안 수준을 요구하는 조직을 위한 보안 솔루션

### 4-2. 방화벽
- 모든 규칙은 위에서 아래 순으로 적용, 허용된 규칙만 통과시킨다.
- 방화벽 정책을 운영하는 환경은 다양하나, 기본적인 운영은 큰 차이가 없다.
- **방화벽 정책 설정 시 최소 3번 이상 확인할 것!**
    - 규칙이 너무 많아 충돌이 날 경우 인터넷이 막힐 수 있음

- 방어기능
    - 애플리케이션 게이트웨이
    - 통신을 중계하는 프록시 서버를 이용해 사내 네트워크와 인터넷 사이에 직접 통신 차단
    - NAT 기능을 통해 어느 정보의 보안 기능을 지원

- 방화벽을 이용한 로그 분석 및 관제
    - 패턴으로 탐지하지 못해도 방화벽에서 어떤 IP에서 어떤 포트로 접속했는지 세션 정보가 확인됨
    - 방화벽의 로그를 이용하면 시그니처 기반으로 탐지 못한 이상행위 발견 가능성

- 한계
    - 네트워크 트래픽이 흐르는 경로 내부에 위치하기 때문에 방화벽 자체가 공격의 대상이 되어 대용량의 트래픽을 견디지 못하고 다운되는 경우가 많음
    - 전체 네트워크 마비가 발생할 위험이 상당히 높음
        - 최근에는 대용량 트래픽을 견디는 장비들이 있어 그나마 나아짐
    - 정상적인 서비스 HTTP, DNS 같은 일반적인 인터넷 프로토콜을 이용한 공격은 방어가 어려움

### IDS/IPS
- IPS는 탐지에 차단까지 수행
- 패턴 기반 탐지
- 임계치 기반으로 트래픽을 탐지해 차단 가능
- In-Line구조로 설치

- IDS는 침입 탐지만 하고 차단은 수행하지 않음
- TAP 방식이나 Mirror 방식으로 설치
    - TAP: 트래픽 손실이나 변조 없이 복사해주는 장비로 트래픽 검사 장비
    - 최초 설치 시 네트워크 단절 필요
    - 장비 장애 발생 시 서비스 중단 추가 비용 발생
    
    - 스위치 미러링: 일반적으로 EnterPrise급 스위치에는 Analyzer의 연결을 위한 Mirroring 기능을 지원

- IDS / IPS 트래픽 분석 구성
