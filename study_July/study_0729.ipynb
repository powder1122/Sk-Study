## AWS 로그 수집
- 로그 수집의 기본 원칙
    - 최대한 오래, 많이 로그 수집이 기본
    - 문제에 대한 원인 파악하여, 개선 자료로 활용
    - 리뷰 : CloudTrail

- ### VPC 흐름 로그
    - VPC 내 로그를 수집하여 장애나 문제 발생 시 원인 파악, 분석 등의 용도로 활용
    - Subnet, ENI 등을 대상으로 활성화 가능
    - 로그 게시를 S3나 CloudWatch logs로 선택 가능
    - Account-id, src_addr, dst_addr, src_port, dst_port, packets,bytes, action(**ACCEPT or REJECT**)
    - 활성화 필요

- ### CloudWatch Logs
    - 로그 수집 및 모니터링, 저장
        - 기존 시스템(서비스), 애플리케이션(EC2) 및 사용자 지정 로그
        - 거의 실시간 모니터링 가능
    - 예: EC2 애플리케이션 로그, VPC 흐름로그, CloudTrail 등
    - 특정 구문, 값 또는 패턴에 대한 로그 모니터링
    - CloudWatch 경보와 연계하여, 알림 가능
    - S3에 저장할 수 있는데 CloudWatch logs에 게시하는 이유는?
        - Logs에는 로그에 나오는 패턴을 경보로 지정하여 알람을 보낼 수 있다.(SNS)
        - S3에 저장하는 것은 주로 나중에 분석하기 위한 용도이다.
        - CloudWatch logs는 실시간으로 도움을 받을 수 있다.
    
- ### S3 액세스 로그
    - S3 버킷에 수행된 요청에 대한 상세 내용 저장
    - 고객 기반 이해 및 보안 및 액세스 감사 유용
    - 활성화 필요
    - 로그 필드
        - 버킷, 버킷 소유자, 요청시간, 원격 IP, 요청자, 작업, 객체키(객체 이름),총시간(요청수신후 응답완료시간) 등

- ### ELB 액세스 로그
    - 로드 밸런서에 전송된 요청에 대한 자세한 정보 저장
    - 트래픽 패턴 분석 문제 해결 유용
    - 활성화 필요
    - 로그 필드
        - 요청 받은 시간, 클라이언트 IP, 지연시간, 요청 경로 등

### AWS Config
- **리소스 구성 변경을 지속적으로 모니터링, 변경사항 관리**
- Rule 기반 점검
- 규정 준수, 미준수 적용 
- 특정 포트만 파악하는 것 등 룰에 대해서 준수했는지, 미준수했는지에 대해 **리포트**를 보여준다.

- Config Rule
    - AWS 관리형 규칙 : 가장 일반적인 사용 사례 중 미리 정의된 구성 상태를 평가
    - 사용자 지정 람다 규칙 : 사용자 지정 논리를 정의하여 리소스의 구성 상태 평가. Python, java 같은 언어로 작성 (고객 생성 룰)
    - Guard 사용한 사용자 지정 규칙 : Guard Custom 정책을 사용하여 사용자규칙을 만들어 리소스 구성 평가

    - 예: S3는 퍼블릭 읽기가 금지되는데, 이를 지키면 규정 준수, 아니면 미준수로 판단한다.
    - 그런 경우, 후속 조치로 로그를 확인 후 권한이 부여되어 있는 유저의 정책에 대해서 컨디션(Condition = True)을 걸어줄 수 있다. (태그를 달아야 해당 리소스 생성을 허용하는 정책 생성)
    - 다양한 룰이 존재한다. AWS Config Rules 문서에서 확인 가능(https://docs.aws.amazon.com/config/latest/developerguide/evaluate-config_use-managed-rules.html)


- 규정 미준수에 대한 보정 : 수동 & 자동
    - Config도 이벤트를 지원하는 서비스이다.
    - Config 결과가 noncompliant가 나온다면 이벤트 룰을 지정하여 타깃에게 메일을 보내거나 람다를 사용하여 조치를 취할 수 있도록 지원한다.
    - Config 안에서 교정하는 메뉴가 존재한다. Config --> remediation을 선택하면 Systems Manager 서비스 안의 Automation 문서 중 **S3를 공개에서 비공개로 수정**하는 기능을 사용할 수 있다. (대상 EC2에게 자동으로 실행할 수 있도록 하는 서비스)

### DDoS 공격 완화
- DDoS(Distributed Denial of Service) 공격
    - 서비스 중단을 목적으로 대상 서버, 서비스 혹은 네트워크에 대량의 인터넷 트래픽을 시도하는 악의적인 공격
    - 대상이 해야하는 합법적인 서비스들을 짧은 시간(1시간 이내)에 요청을 폭주시켜 리소스를 고갈시켜 마비시키는 공격이다.

- 규모 기반, 상태 고갈, 애플리케이션 레이어 DDoS 공격이 있다.

### DDoS 공격 완화를 위한 대응 가이드
- Route53, CloudFront를 구성하여 endpoint를 노출시키지 않으면서 애플리케이션의 DDoS 공격을 완화할 수 있다. (엣지 서비스, API GW와 비슷한 역할) 또한 캐쉬 우회 공격도 예방할 수 있다. 
- CloudFront에서 발급받은 오리진ID를 S3 Principal로 지정하면 CloudFront를 통해서만 접근이 가능해지기 때문에 공격을 막을 수 있다.
- 앞단의 보안 그룹을 거칠 수 있도록 sg-체인 다이어그램을 활용하여 CloudFront를 설정하여 이를 거쳐야만 들어올 수 있도록 구성할 수 있다.
    - 캐쉬 우회 공격: 오리진 서버에 직접적으로 연결하는 공격.

- 확장이 가능한 서비스 구조 : ELB + AutoScaling
    - 확장이 필요할 시 자동으로 확장이 가능한 서비스로, 온프레미스 환경에 비해 유연한 구성이 가능하다. 앞부분이 다 뚫려도 최종단에서 완화시켜줄 수 있다.

- 공개적으로 노출되는 엔드포인트 노출 최소화
    - 직접적으로 접근 불가

    - 필터링

### AWS Shield
- AWS에서 DDoS 공격을 완화하기 위해 지원하는 서비스
- DDoS 공격을 방어하기 위해 개발된 AWS 서비스
- AWS 서비스에 내장된 구조로 아키텍처의 큰 변경 없이 DDoS 완화를 빠르게 활용
- Shield Standard(무료)와 Shield Advanced(유료)로 구분
    - 유료 버전을 쓰면 애플리케이션 계층 DDoS 공격까지 보호할 뿐만 아니라 24시간 전담팀 지원, 추가 비용 없이 WAF 사용까지 지원한다.
    - 기본적으로 3,4계층 DDoS 공격에서 보호한다.

### AWS WAF(Web Application Firewall)
- AWS 웹 애플리케이션 방화벽(7계층 공격 방화벽)
- ALB, CloudFront, API Gateway에 적용
- WebACL을 구성하여 규칙(rule)이나 특정 패턴에 따른 허용, 거부 설정 가능
- 규칙(rule) 예
    - IP 주소, HTTP 헤더와 본문, URI 문자열
    - SQL Injection, XSS(CrossSite Scripting)
- OWASP Top10와 CVE(Common Vulnerabilities and Exposures) 관련 규칙들을 자동 갱신

- 방어를 위한 시큐어 코딩, 코드 레벨에서 방어를 할 것인지 등이 애플리케이션 보안에 해당하는 내용.

    - 서버: S3가 제공하는 암호화 방식 등
- 암호화 방식
    - 대칭키: 속도가 빠르다. 데이터 컨텐츠는 대칭키를 주로 사용
    - 비대칭키: 인증에 관한 데이터는 비대칭키를 주로 사용

- Key 관리의 문제
    - 다수의 키 관리 --> 키 관리에 대한 오버헤드 발생
    - 키 저장 및 교체
    - 키 관리 권한 및 사용 권한 제어

- KMS: AWS환경에서 제공하는 키 관리 서비스
    - 암호화 키 관리 서비스
    - 키 저장 및 관리, 암호화
    - 봉투 암호화
    - 2계층 키: 마스터 키, 암복호화 키 별도로 제공하는 방식. 마스터키에게 암호화키(데이터키)를 발급 받아 암호화, 또 다른 데이터 키를 통해 복호화하는 방식
    - 정책 기반 키 보호(키에 대한 권한 설정 등)
    - 키 교체 자동, 비활성화 및 삭제, 키 사용 감사(키를 사용한 작업에 대한 로그 생성)

- KMS 암호화 지침 : 봉투 암호화
    - 암호화 한 키도 암호화하여 전송하는 방식. 
    - 마스터키(Wrapping Key)로부터 Data Key를 받아 평문을 암호화 전송, 마스터키는 대칭키인 Data Key를 암호화하여 전송 --> 마스터 키에게 키 복호화 요청 --> 권한 확인 후 복호화된 키 전송 --> 키를 받아 암호화된 데이터를 복호화

- KMS 키 종류
    - AWS 관리형 키
    - **고객 생성 마스터 키**
    - 가져온 키(Import 된 키) --> KMS 키 형식에 맞춘 후 Import 가능. AWS 관리형 키, 마스터 키는 Import 불가

    - **보안 그룹 : 인바운드 규칙 항목 삭제**
