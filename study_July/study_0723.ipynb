## 보안 그룹과 NACL

### 보안그룹
- 인스턴스 레벨의 방화벽
- 기본값
    - 인바운드 규칙 모두 허락 X
    - 아웃바운드 규칙 모두 허락
    - 기본값은 변경 가능. 주로 인바운드 값을 수정

- 규칙 추가를 통해 인바운드 룰 또는 아웃바운드 규칙 생성 가능
- 설정 후 보안 그룹 생성

- 보안 그룹: 체인 다이어그램
    - 로드 밸런싱과 연관
    - 3 Tier
        - web -> app -> DB
        리소스: 서비스의 생성 결과물. 인스턴스를 여러 개 생성하면 각각을 구분할 수 있는 리소스ID가 생성된다.
        - 소스 자리에 이전의 Security를 통과한 리소스 ID에게만 허용을 하는 티어 구조를 가진다. 0.0.0.0을 사용하는 것보다 보안 그룹을 사용하여 이전의 보안을 통과한 리소스 ID만 통과시키는 것이 권장 사항이다.



### NACL(Network Access Control List)
- 서브넷 단위의 네트워크 방화벽
- 기본값
    - 인바운드, 아웃바운드 모두 허락
    - 마찬가지로 변경 가능

    - 디폴트: * 사용. 조건에 해당하지 않는 모든 규칙을 디폴트에 포함하여 Deny할 수 있다.

    - EC2는 서브넷 안에 배포한다. 서브넷과 인스턴스에서 허락이 되어야 하기 때문에 상호 보완적이다.(보안 그룹과 NACL은 상호 보안적)

    - **루트 사용자는 일상적인 작업을 위해 사용하지 않도록 권고**

### IAM - AWS IAM과 접근 권한
- 종류를 따지지 않고 2번 이상 인증한다. - 다단계 인증. 카드 은행 등에서 박릅받아 IAM을 연결하면 봄 더 보안적이 된다.
- 두 종류 이상의 인증을 사용한다. - 다요소 인증

- User, Group, Policy
    - IAM 정책에 그룹이나 사용자를 연결한다.
    - 정책을 변경하면 연결된 그룹이나 사용자의 권한이 바뀐다.



- 정책: 어떤 서비스 기능에 어떤 조작을 할 수 있는지 정의
    - 정책 생성기로 간편하게 **JSON 형식**의 정책 문서를 만들 수 있다.
    - IAM 사용자, 그룹 및 IAM 역할, AWS 리소스에 정책을 연결
    - 필수 키워드: 
        - Effect: 허용/거부 (Allow, Deny)
        - Action: 서비스 특정 작업
        - Resource: 리소스 이름 (ARN, Amazone Resource Name)

    - 관리형 정책: 재사용 가능
        - AWS 관리형 정책: **AWS 제공** --> 사용이 편하다.
        - 고객 관리형 정책: **고객이 직접 생성** --> 세밀한 권한 설정 가능
    - inline 정책: 싱글 identity (IAM 사용자, 그룹, 역할)에 연결 (Policy:identity = 1:1)

- 정책 우선 순위 -> 정책 충돌 시 판정 우선순위
1. 명시적 상태 Deny를 체크. 존재하면 바로 거부.
2. 명식적 Allow 확인 --> 조건(Condition) 확인. 충족 시 허용, 불만족시 거부
3. 암시적 Deny --> 거부 (allo, deny가 존재하지 않으면 암시적 Deny)

    - 역할(Role): 임시 권한 부여
        - **User의 임시 권한 필요 시** --> 특정 시간동안 권한 부여

        - **서비스가 서비스를 액세스 할때** --> EC2에서 S3의 파일에 접근할 일이 있을 때, 서비스에는 정책 부여가 불가능하기 때문에 Role을 사용하여 위임한다.

        - **자격 증명 연동**

    - 역할에 부여하는 정책
        - 권한 정책
        - 신뢰 정책


    - IAM에서 권한을 부여할 수 있는 객체는 User, Groupt, Role 3가지 뿐이다. 임시로 잠깐 권한을 주고 싶을 때 정책을 부여했다가 제거한다. 또는 권한이 부여되어 있는 그룹에 유저를 넣었다가 제외하는 방법도 있다. 마지막으로 권한이 있는 유저의 ID와 PW를 공유하는 방법이 있다.

    - 가장 권장하는 방법은 Role을 이용하여 특정 유저에게 위임하여 수여받는 유저가 특정 시간(TTL)동안 권한을 행사할 수 있다.
    

- 서비스가 서비스를 액세스 할 때
- 계정간의 AWS 리소스 접근 --> 권한을 하나씩 받는 것보다 Role이 가진 권한 만큼 임시로 작업을 할 수 있는 것. 인증을 따로 받기 때문에 할 때마다 인증할 필요가 없다.
