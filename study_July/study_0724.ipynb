## IAM

### IAM - 권한 경계
- 권한 경계(Permission Boundary)
    - 아무리 권한이 허락되어 있어도 특정 범위까지만 작업할 수 있도록 하는 것.
    - IAM에서 권한의 최대 범위를 제한
    - 유저 혹은 역할에 부여할 수 있는 최대 권한을 설정
    - 자격 증명 기반 정책과 권한 경계 **모두에서 허용되는 작업**만 수행하도록 허용: **교집합 범위만 허용**

- IAM 정책으로 S3에게 모든 권한을 허락해도 권한 경계가 ReadOnly라면 객체 업로드, 삭제 등의 작업이 불가능하다. (읽기, 가져오기(다운로드)만 가능.)

### EC2 오토 스케일링
- 개념: 로드 처리를 위해 EC2의 개수 자동 조절
- 필요성: 통계를 통해 요일 별로 사용하는 EC2가 달라 피크치를 기준으로 일정한 수를 유지하면 낭비되는 EC2가 생기기 때문에 수요에 맞춰 크기를 자동으로 조정하기 위해 나온 기능이 Auto Scailing이다.

- AmazonEC2 Auto Scaling 란?
    - 애플리케이션의 부하를 처리할 수 있는 EC2 인스턴스 수량을 min과 max 수량 사이에서 자동 추가 및 제거를 통해 유지

    - 확장/축소 개념
        - Scale-out vs Scale-in (수평 확장/축소 --> 동일한 기능의 EC2 개수를 늘려 성능 향상, )
        - Scale-up vs Scale-down(수직 확장/축소 --> DB등 사용, 기존 인스턴스의 성능을 업그레이드)

- 오토 스케일링 장점
    - 향상된 가용성
    - 비용 절감

- 오토 스케일링 요소
    - ASG(Auto Scailing Group)
        - min/max/desired(최소, 최대, 수요에 맞게)
        - 기준을 가지고 정해야 한다.
        - MAX의 기준은 커버 가능한 비용과 관련된다.
        - MIN은 유지해야하는 최소 **가용성**과 관련된다. 반드시 유지해야하는 가용성 값!
        - Desired capacity: 원하는 개수. 가변 값으로, 정책에 따라 증감할 수 있다. **현재 떠 있는 EC2 개수를 의미**

    - 시작 템플릿: 인스턴스 구성 정보 지정
        - AMI, 인스턴스 유형, 보안 그룹, 키페어 등

    - 오토 스케일링 정책

- ### 오토 스케일링 정책
    - 수동: 오토 스케일링으로 설정 되어 있어도 수동 조작은 가능하다. desired 값 변경

    - 자동
        - 예약된 일정 - 

        - 초기값 유지 - min/max/desired : 2/2/2, 무조건 2개 유지

        - **동적 조정 정책** 
            - **cloudwatch** 서비스를 이용하여 크기를 조정하는 정책.
            
            - 단순 조정 정책: 단순한 기준을 두고 정책 설정.
                - 예: CPU 60% 이상이면 EC2 추가
            - 단계 조정 정책
                - 예: CPU 60% 이상이면 EC2 추가, CPU 20%미만이면 EC2 하나 제거
            - 대상 추적 조정 정책
                - 타겟을 정하고, 타겟에 맞게 min/max/desired 내에서 EC2 수를 조정하는 정책
                - 예: 타겟이 CPU Utilization 60%이면, 이를 맞추기 위해 EC2를 추가하거나 삭제한다. 현재 EC2 3개에 65%이면 하나 늘려 4개를 사용한다.

    - 예측 기반 조정 정책
        - 머신러닝 기법으로 분석하여 이를 기반으로 예측하여 설정하는 것
        - 분석 비용이 들어간다.
        - 편리함
