## AWS 개요
- AWS
    - 아마존에서 제공하는 클라우드 컴퓨팅 서비스.
    - 리소스를 필요에 따라 사용하고 사용한만큼 비용을 지불하는 플랫폼

- 클라우드 컴퓨팅 서비스
    - 인터넷
    - IT 리소스
    - **On-demand**: 필요에 따라, 요구 시 즉시 이용 가능을 의미. 빠르게 서버 배포 가능하다.
    - 종량 요금제: 사용한 만큼 청구서

    - **온프레미스**: 전산실, 데이터 센터 등, 작업을 위해 용량 산정 계획(CP) 및 구매, 도입, 설치, 설정이 필요하여 시간이 걸린다. (많은 초기 투자 필요)

- 클라우드 컴퓨팅(on-demand) vs on-premis
    - 클라우드 컴퓨팅이 유연성이 뛰어나다. on-premis 환경에서는 계획에 맞게 작업을 하기 때문에 적절한 조절을 하기가 어렵다. 서비스가 실제로 너무 오버 사이즈거나, 너무 작은 경우 문제가 될 수 있다. 그에 반해 클라우드 컴퓨팅은 사이즈가 부족하면 사이즈를 늘리고, 남는다면 리소스의 크기를 줄여 비용 절감을 할 수 있는 등 유연한 대처가 가능하다.

- 온프레미스가 클라우드보다 좋은 것

### AWS 제공 솔루션
- 온프레미스의 모든 자산을 AWS로 이전하지는 않는다.
- 보안 등의 이유로 클라우드에 올리지 않아야 하는 데이터들도 존재.
- 이런 경우 양쪽을 지원하는 하이브리드 환경이 필요한데, AWS는 이를 지원하는 옵션이 존재한다!



### AWS Global Infra
- 리전 (Region)
- 가용 영역 (Availability Zone, AZ)

- 각 리전에 여러 가용 영역이 존재한다.
    - 지역적으로 묶어서 해당 지역에 있는 데이터 센터들을 모아 논리적으로 부르는 용어가 가용 영역이다.
    - 가용 영역들의 모임을 리전이라고 한다.
    - **데이터 센터 (DCs) < 가용 영역 (AZs)< 리전 (Region)**
    - 데이터 센터는 선택할 수 없음. 사용할 리전과 가용 영역까지 선택할 수 있다.

- 리전 선택의 기준
    - 해당 지역의 데이터 센터에 서비스를 올리는 것. 
    - 지연시간: **가장 빠르게** 클라이언트에게 서비스를 제공할 수 있는 리전을 선택
    - 법적인 문제
    - 가용 서비스
    - 비용

## EC2 개요
- EC2(Elastic Compute Cloud): AWS 컴퓨팅 플랫폼 기본
    - 컴퓨팅 리소스(CPU, 메모리 등)를 가진 서버 노드 기능, 인스턴스
    - 지원 운영 체제 : Linux, Windows, Mac OS
    - 실행 중일 때만 비용이 발생

- 서버와 인스턴스
    - 인스턴스란 가상 서버를 말한다. 
    - 서버 = EC2 인스턴스

- 가상 서버를 구축하는 절차
    1. 리전을 선택하고 관리 콘솔을 연다.
        - EC2 대시보드를 연다
    2. EC2 인스턴스를 생성한다.
        - AMI(아마존 머신 이미지)를 선택한다.
        - 인스턴스 유형을 선택한다.
        - 네트워크와 IP 주소를 할당한다.
        - 스토리지를 추가한다.
        - 보안 그룹을 설정한다.
        - 키 페어를 선택(작성)한다.
    3. EC2에 접속한다.
        - SSH로 EC2에 접속한다.
    4. 소프트웨어를 설치하고 설정한다.
        - 소프트웨어 설치
        - 소프트웨어 설정


### EC2 개요 - AMI와 인스턴스 유형
- AMI(Amazon Machine Image): 루트 볼륨 템플릿. OS + a, 생성 틀과 같다.
- AMI 활용
- 고객 생성 AMI
    - AMI에서는 같은 설정의 서버를 얼마든지 생성할 수 있다.(서버 복제 가능)
    - 개인이 만든 AMI는 복제와 배포가 가능

- 인스턴스 유형: 인스턴스의 용량(크기)
    - 인스턴스 유형에 따라 인스턴스에 사용되는 호스트 컴퓨터의 하드웨어가 결정
    - 각 인스턴스 유형은 서로 다른 컴퓨팅, 메모리 및 스토리지 기능을 제공

- 인스턴스 요금 옵션
    - 온디맨드
    - 예약
    - 스팟

- 유형 선택 예시
    - t2.micro
        - t2: 인스턴스 유형(대략적인 용도, 버스트 가능한 범용 유형의 2세대)
        - micro: 인스턴스 크기(CPU나 탑재 메모리의 용량 등의 성능)
    - 오버 사이징된 경우 크기를 축소하여 비용을 절감할 수 있다.
    - 유연한 대처 가능
    - 범용, 컴퓨팅 최적화, 메모리 최적화, 가속화된 컴퓨팅 최적화, 스토리지 최적화 등 용도에 맞게 선택할 수 있다.
