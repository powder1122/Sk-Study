## EC2 개요-Amazon EBS와 스냅샷
### EBS
- EBS(Elastic Block Store) 기능: 
    - EC2에 **종속적인** 스토리지로, EC2에 연결되어 OS적인 파일 저장 스토리지이다. 
    - EBS에 저장되어있는 데이터에 접근하기 위해서는 EC2에 연결하는 것이 필수이다.
        - **연결을 위해서는 EC2와 같은 가용 영역에 있어야 한다.**

- EBS 유형
    - HDD: SSD보다 성능이 낮다. 값싼 마그네틱 스토리지. 접근 빈도가 낮을 때 사용한다.
    - SSD: HDD보다 성능이 뛰어나다. 고성능 스토리지가 필요한 경우 사용한다. 필요한 IOPS를 사용자 지정 가능하다.
    - 목적에 따라 선택하면 된다!

- EC2 생성 시 EBS 생성: 스토리지 볼륨 설정
    - 루트 볼륨(OS 볼륨) 유형과 크기
    - 데이터 볼륨 추가 가능: 루트 볼륨 외 데이터를 저장할 볼륨

- EBS 백업: 스냅샷
    - 서버를 저장하는데, 특정 주기마다 그 시점의 서버 상태를 저장한다.

## 네트워크
### VPC
- VPC(Virtual Private Cloud)
    - AWS에서 생성할 수 있는 프라이빗 가상 네트워크
    - 단일 리전에 연결
    - 사설 IP CIDR 블록을 지정. 충분한 IP 설정을 갖도록 설정
    - 사설 네트워크를 만드는 이유: 외부로 나가는 것, 또는 내부로 접근하는 것을 차단하여 보안을 향상하는 것이 목적이다. (외부와 단절)

    - 네트워크 장비에 해당하는 리소스를 VPC 안에 추가해나간다.
    - VPC를 쪼갠 단위를 서브넷으로 한다.



### 서브넷과 가용 영역
- 서브넷
    - VPC를 더 작은 단위로 분할한 하위 네트워크
        - 서브넷 CIDR 블록은 사용할 IP 개수를 고려 결정
        - 서브넷 CIDR 블록은 VPC CIDR블록의 하위 조각
    - 서브넷은 하나의 가용영역(AZ)에 속함
    - 서브넷간 IP 주소는 중복될 수 없음

- Public Subnet
    - **연결된 라우팅 테이블에 Internet gateway 항목이 있음**
        - 서브넷에 연결하여 외부 Public 통신 가능
    - 외부에 공개되어야 할 노드 배포

- Private Subnet
    - 연결된 라우팅 테이블에 Internet gateway 항목이 없음
    - 외부 접근 차단 노드 배포

- CIDR (Classless Inter Domain Routing)
    - 네트워크의 호스트 주소 개수를 계산
    - ip주소 뒤에 /숫자가 온다. 192.168.0.0/16 --> 앞의 주소는 범위 내 첫 번째 IP 주소, /16이 CIDR(2의 16제곱 개의 IP 사용)
    - 호스트 자리에 0은 대표 주소, 
    - 255는 브로드캐스트 주소, 
    - 1은 내부 게이트웨이 주소, 
    - 2는 해당 서브넷 안에서 제공하는 네임 서버
    - 3은 예약 주소로, 할당하지 않음
    - 예약된 5개를 제외한 251개가 주소로 설정된다. (CIDR /24의 경우)

    - EC2 10개를 만든다면? /28의 경우 11개 주소 지원, 10개의 IP주소까진 커버 가능하다. 그러나, 확장을 하려면 CIDR 확장이 필요한데, 한 번 생성한 서브넷의 CIDR은 변경할 수가 없기 때문에 추가로 서브넷을 생성하여 기존의 서브넷을 옮겨야 한다.(번거롭다)
    - 이러한 일이 빈번하게 일어나지 않도록 하기 위해 권장 사항이 존재
        - AWS에서 서브넷 생성 시 넉넉하게 CIDR 지정을 권장한다.
        - 사설 IP(VPC) 내에서 부여하기 때문에 아낄 필요가 없다.
        - Public IP는 따로 지정해야 한다.
        - /24 이상(24 이하의 수)의 규모를 권장



### 인터넷 게이트웨이
- 관리형 서비스 - AWS가 자동으로 관리해주는 서비스

- 인터넷 게이트웨이 서비스
    - VPC와 인터넷간 통신이 가능하게 하는 관리형 게이트웨이 서비스
    - IPv4, IPv6 지원
    - 퍼블릭 네트워크를 지원해주는 AWS 관리형 게이트웨이 서비스
    - 이 게이트웨이를 사용하도록 조치한 서브넷을 Public subnet이라고 한다.

### ENI & EIP
- EC2가 공인 IP를 받는 방법 2가지.

- ENI(Elastic Network Interface)
    - **다이나믹 IP**, 바뀔 수 있다.
    - VPC에서 가상 네트워크 카드를 나타내는 논리적 네트워킹 구성 요소
    - EC2 연결 가능
    - IP 주소, MAC 주소 부여 가능
    - EIP 연결 가능
    - 보안 그룹 부여 가능

- EIP(Elastic IP)
    - 고정 공인 IP
    - EC2나 ENI 연결

- 하나의 네트워크 카드로 여러 기능을 지원하는 경우
    - Multihomed Service. 이 경우 안정성이 떨어져 사용하지 않는다.

- 최근에는 모니터링, 보안 등의 이유로 각각 네트워크 카드를 사용한다.
    - 모니터링에서 이 기능만을 위한 네트워크 카드를 요구하는 일이 많다.
    - EC2 하나에 네트워크 카드 하나 사용. 추가적인 네트워크 카드를 요구하면, ENI를 이용하여 네트워크 카드를 추가할 수 있다.

### 실습 10번 AWS 네트워크 구성
- 서브넷과 가용 가용영역
- 네트워크의 가용 영역에 중복 배포할 수 있도록 구성하는 것을 권장

1. 기본 VPC(Default VPC) 삭제
2. 주어진 리전에서 VPC(MyVPC),Subnet(2개의 AZ에 퍼블릭 서브넷, 프라이빗 서브넷 하나씩)을 생성(위 그림 참조하여 생성)
3. IGW(myIGNAT GW(myNATGW, MyPubSubnet에 생성)W) 생성하여, VPC연결 → 라우팅 테이블 생성 → 라우팅 테이블 연결(퍼블릭 서브넷에)
4. 라우팅테이블 생성하여 프라이빗 서브넷에 연결(MyPubRT, MyPriRT)
6. 마무리: NAT,EIP 삭제(릴리스)->시간조금 걸립니다.

- NAT 게이트웨이는 Public 서브넷에 설치하여 Private 서브넷에 연결(Public 둘 중 하나에 배포)
- EIP 할당 필요
- 통신을 많이 하면 작게나마 비용이 발생. 실습 후 NAT, EIP 삭제
- 외부에서 통신하는 라우팅은 Public으로 구성하는 것이 맞지만, 보안상의 이유로 필수적이지 않으면 Private로 설정하는 것을 권장한다.
- **노출이 필요한 것만 Public 서브넷 사용!!**
- 라우팅 테이블이 제대로 작성되어 있으면 모든 EC2 서버에 적용이 된다.

    - 서브넷을 이용하여 각 영역을 설정할 수 있다.
    - 서브넷은 VPC 내의 가용 영역. AZ 내에 설정
    - 기본적으로 인바운드/아웃바운드가 차단된 Private 상태
    - Internet Gateway를 생성하고 라우팅 테이블에 연결하여 노출이 필요한 노드들(서브넷)을 Public으로 연결할 수 있다.
    - 인터넷 게이트웨이에 연결된 서브넷이 Public Subnet, 그 외에는 Privete Subnet이라고 한다.

    - Privete 서브넷은 NAT게이트웨이를 이용하여 외부와 연결할 수 있다.
    - NAT는 중간 다리 역할로 Public 서브넷에 두고 Private 라우팅 테이블에 연결하여 내부에서 외부로의 요청을 전달해주고, 내부로의 접근은 차단할 수 있도록 한다.

    - 보안 권장 상 최소 두 개의 가용 영역에 중복 배포하는 것이 좋다.
    - 한 군데에 모두 배포하지 않고, 장애 발생 시 대처할 수 있도록 구성하는 것이 좋다.

    - 변경되는 IP: Dynamic IP
    - 고정적인 IP: EIP(static한 IP)
