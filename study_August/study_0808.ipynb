### 버그헌팅과 모의해킹 
- 정의 및 차이점
- 버그헌팅 : 허가된 시스템에서 새로운 버그*를 찾아내 포상을 받는 것
    - 버그(Bug)란 소프트웨어가 예상한 동작을 하지 않고 잘못된 결과를 내거나 오류가 발생하거나 작동이 실패하는 등의 문제를 의미

- 모의해킹 : 허가된 시스템의 취약점을 찾아내고 이를 성공적으로 공격하는
    것(개념증명 포함)
    - 취약점 진단은 잠재적인 보안문제를 찾기 위해 서비스와 시스템을
    검토하는 과정
    - 모의해킹은 잠재적인 보안문제를 통해 악용될 수 있는 공격 시나리오를
    증명

- 버그 바운티(Bug Bounty)
    - 웹 서비스 또는 소프트웨어의 취약점을 찾아낸 사람에게 포상금을 지급하는 제도
    - 화이트 해커 측면 : 개인역량 및 인지도 상승, 실질적 포상
    - 기업의 측면 : 취약점 업데이트, 보안 위협 대응, 비용 절감, S/W 품질 향상, 기업 가치 제고
    - 약속한 부분이 아닌 곳을 공격하면 해킹으로 간주. 허가 없이 시도하는 것도 해킹으로 간주.

- 버그: 개발자가 졸다가 실수한 것 | 디버깅
- Fuzzing: 정상의 시스템에 과도한 스트레스 테스트를 수행하면 함수 오동작, 메모리 누출, 격리 우회 | 유발(잠재적 오류)

- 해킹이 됐다가 안되는 것은 취약점이 아니다!

- 모의해킹의 과정: 대상, 가설 수립, 가설 검증(방법론), 검증, 페이퍼 작성, 페이퍼 검증(소스코드)

### 애플리케이션 보안 위험 평가
- 위험 요인
    - 위협요소 : 위협원, 공격자(개인 또는 그룹)를 의미
    - 공격벡터 : Attack Surface, 공격방법, 공격 가능성을 의미(찾기 쉽거나 어려울 수 있음). 찾기 쉬운 것은 위험!
    - 보안취약점 : 잠재적인, 알려진 보안 취약점(확산 정도, 탐지 가능성에 따라 분류)

- 영향 요인
    - 보안통제 : 보안 취약점을 대응하기 위한 보안통제방안을 의미(우회 가능 여부)
    - 기술적영향 : 자산과 시스템 기능에 미치는 영향(영향이 미미하거나 클 수 있음)
    - 비즈니스영향 : 조직의 사업적인 영향(영향이 미미하거나 클 수 있음)

- 위험 등급
    - 공격가능성 : 쉬울수록 위험, 어려울수록 덜 위험
    - 취약점 확산정도 : 광범위할수록 위험, 확산범위가 좁을수록 덜 위험
    - 취약점 탐지 정도 : 탐지 쉬울수록(더 많이 알려진 취약점일수록) 위험, 탐지 어려울수록 덜 위험
    - 기술적 영향 : 자산/시스템에 미치는 영향이 클수록 위험, 경미할수록 덜 위험


### 10대 웹 애플리케이션 취약점
1. 인젝션
- SQL, OS, XXE, LDAP 인젝션 취약점은 신뢰할 수 없는 데이터가 명령어나 쿼리문의 일부분이 인터프리터(번역기)로 보내질 때 발생
- 공격자의 악의적인 데이터는 예상하지 못하는 명령을 실행하거나, 적절한 권한 없이 데이터에 접근하도록 인터프리터를 속일 수 있음

2. 인증 및 세션관리 취약점
- 인증 및 세션 관리와 관련된 애플리케이션 기능이 종종 잘못 구현되어 공격자에게 취약한 암호, 키 또는 세션 토큰을 제공해 다른 사용자의 권한을 (일시적으로 또는 영구적으로) 얻도록 공격 할 수 있음

3. 민감한 데이터 노출
- 대부분의 웹 애플리케이션과 API는 금융정보, 건강정보, 개인식별정보와 같은 민감정보를 제대로 보호하지 못하고 있음
- 공격자는 신용카드사기, 신분도용 또는 다른 범죄를 수행하는 취약한 데이터를 훔치거나 변경할 수 있으며 브라우저에서 중요 데이터를 저장 또는 전송할 때 암호화 필요

4. XML 외부 개체(XXE)
- XXE는 악의적인 자바스크립트를 막기 위한 필터장치를 우회하는 취약점으로 XML 문서에서 동적으로 외부 URI의 리소스를 포함시킬 수 있는 외부 엔티티(Entity)를 사용할 때 발생
- 외부 엔티티는 파일 URL 처리기, 패치되지 않은 Windows 서버의 내부 SMB 파일 공유, 내부 포트 검색, 원격 코드 실행 및 서비스 거부 공격과 내부파일 공개에 악용될 수 있음

5. 취약한 접근 통제
- 취약한 접근 통제는 인증된 사용자가 수행할 수 있는 작업에 대한 제한이 제대로 적용되지 않는 것을 의미
- 공격자는 해당 결함을 악용해 다른 사용자의 계정에 액세스하거나, 중요한 파일을 보고 다른 사용자의 데이터를 수정하거나, 접근 권한을 변경하는 등 권한 없는 기능 또는 데이터에 접근

6. 잘못된 보안 구성
- 바람직한 보안은 애플리케이션, 프레임워크, WAS, 웹 서버, DB 서버 및 플랫폼에 대해 보안 설정이 정의되고 적용
- 보안 기본(Default) 설정은 대부분 안전하지 않기 때문에 정의와 구현, 그리고 유지라는 3박자가 잘 이루어져야 함(또한 소프트웨어는 최신 버전으로 관리해야 함)

7. 크로스 사이트 스크리팅(XSS)
- XSS 취약점은 애플리케이션이 적절한 유효성 검사 또는 이스케이프 처리(URI로 데이터를 전달하기 위해서 문자열을 인코딩)없이, 새 웹 페이지에 신뢰할 수 없는 데이터를 포함하거나, JavaScript를 생성할 수 있는 브라우저 API를 사용해 사용자가 제공한 데이터로 기존 웹 페이지를 업데이트
- XSS를 사용하면 공격자가 희생자의 브라우저에서 사용자 세션을 도용하거나, 웹사이트를 변조시키거나, 악성사이트로 리다이렉션 시킬 수 있음

8. 안전하지 않은 역직렬화
- 응용 프로그램이 악의적인 직렬화된 객체를 받았을 때 불안정한 직렬화 오류가 발생
- 안전하지 않은 직렬화로 인해 원격코드가 실행(범용 API나 데이터를 변환하여 추출 시 사용)
- 공격자에 의해 직렬화된 개체가 재생, 변조, 삭제, 삽입, 공격자 권한 상승 등의 문제가 발생

9. 알려진 취약점이 있는 구성요소 사용
- 컴포넌트, 라이브러리, 프레임워크 및 소프트웨어 모듈은 애플리케이션과 같은 권한으로 실행
- 이런 취약한 컴포넌트를 악용해 공격하는 경우, 심각한 데이터 손실이 발생하거나 서버 탈취
- 알려진 취약점이 있는 구성 요소를 사용하는 애플리케이션과 API는 어플리케이션을 약화시키고 다양한 공격에 노출될 수 있음

10. 불충분한 로깅 및 모니터링
- 불충분한 로깅 및 모니터링과 사고 대응과의 통합이 누락되거나 비효율적인 경우 공격자에 의해 시스템 공격으로 이어지거나 데이터 변조와 추출, 심지어 파괴될 수 있는 취약점
- 제로데이 악성코드, 내부자 위협, APT, DDoS 공격을 탐지하기 위해 네트워크 트래픽과 네트워크 내에서 수평으로 이동하는 내부트래픽(east west traffic)의 통합 모니터링 필요

- ### OWASP 06 구성요소 취약점
- OWASP 06 취약점 - A06(취약하고 오래된 구성요소): 애플리케이션에서 사용하는 서드파티 라이브러리, 프레임워크 또는 운영체제(OS) 등에서 취약점이 발견되었지만, 최신 버전으로 업데이트되지 않을 경우 발생한다. 이는 현대 소프트웨어 개발에서 오픈소스 라이브러리와 상용 컴포넌트를 광범위하게 사용함에 따라 더욱 중요한 보안 요소로 부각됩니다. 구성요소의 취약점이 있으면 애플리케이션 전체가 취약해질 수 있기 때문입니다. 따라서 이를 주기적으로 점검하고 업데이트하는 것이 필수적입니다.

- 정리: A06은 지원 종료되었거나 알려진 취약점을 포함한 컴포넌트 사용 시 발생하는 위험이다.

- 취약한 구성요소들은 해커들에게 시스템 침입의 문을 열어준다. 해커들은 이러한 약점을 이용해 데이터베이스 조작, 악성 스크립트 삽입, 서버 장악 등 다양한 공격을 시도합니다. 특히 서버 장악 공격은 해커가 서버에서 원하는 명령을 실행할 수 있게 하여 전체 시스템이 완전히 통제당할 위험이 있습니다. 예)Log4j
