- Q. 애플리케이션 보안을 위해서는 서버리스의 어떤 부분을 이해해야 할까?
    - Stateful vs Stateless (상태 기반 방식 비교)
        - Stateful(상태): 정보 = 상태, 쿠키/세션 기반, 상태 고정(Client-Server), 부하분산/라우팅변경(세션유실) | 최소화, TCP 
        - Stateless(무상태): 정보 = 상태X, 그때그때 상태 정보 요청. 부하분산/라우팅변경(세션유지) | 최대화, UDP/HTTP 
    - RestAPI vs WebAPI: URL 타입의 인터페이스 정보 송수신 경로(RestAPI) --> Custom 방식의 상호보완적인 WebAPI
    - API 호출하는 방식: URL처럼 호출 --> 마이데이터, 오픈뱅킹, 기업간연동/연계  | 기계-기계(API), AI-AI(MCP)
        - 개인정보의 자동화된 처리 - 인공지능에 의해 자동화된 처리가 될 경우의 정보 보호 가이드라인(2020.12개정)
    - SaaS인 이유? 자산 식별의 방법이 회사마다 다르기 때문에 Argument 가 존재!
        - 보안 담당자는 '자산'으로써 식별이 어떻게 되고 있는지, '자산'이 갖는 특징이나 기술적인 동작원리는 어떤지, 이게 왜 SaaS로써 보안 관리가 되어야 하는지 설명할 수 있어야 한다.

## 토론
1. 왜 망분리가 완회되는지, 망분리가 완화됨에 따라 어떤 보안조치가 추가되었는지? 그래서 뭘 하라는 것인지?
    - 인공지능·클라우드와 같은 인터넷 기반 기술이 급격히 발전하였고, 일률적인 인터넷망 차단조치로 인해 혁신 기술의 활용이 어려워 개선이 필요하다는 의견이 지속적으로 제기되어 왔다. --> 망분리 완화 이유
    1. 첫째, 위험도에 따라 인터넷망 차단수준을 차등적으로 적용한다.
    - 개인정보처리자는 내부관리계획에 따라 위험분석을 거쳐 위험수준을 구분해 저위험·중위험 컴퓨터 등에 대하여 인터넷망 차단조치 이외에 이와 상응하는 보호조치 하에서 제한된 인터넷망 접속도 허용하도록 완화할 방침이다.
    - 개인정보 데이터베이스(DB)에 대한 접근 권한을 설정할 수 있는 고위험 컴퓨터 등은 탈취 시 대규모 개인정보 유출이 우려되므로, 기존과 같이 인터넷망 차단조치 의무를 유지할 계획
    2. 둘째, 개인정보처리자의 책임 강화를 통해 보호수준 저하를 방지한다.
    - 인터넷망 차단조치 완화로 인해 개인정보 보호 수준이 저하되지 않도록, 개인정보처리자 스스로가 책임 강화를 통해 보호수준 저하를 방지해야 한다.
    3. 셋째, 개인정보처리자에 대한 지원을 강화한다.
    - 기술지원 전담반을 구성하여, 상응 보호조치 적용을 원하는 개인정보처리자에 대해 현황진단, 애로 상담 등 맞춤형 원스톱 지원 서비스를 제공할 계획
    - 사업자가 적용할 상응 보호조치가 인터넷망 차단조치에 상응하는 조치인지 여부를 사전에 검토하여 개인정보처리자의 불확실성을 제거할 예정

2. 제로 트러스트, 보안 신기술(솔루션), 생성형 AI 활용 등의 방법을 통해서 망분리가 완화되었을 때, 내가 보안담당자라면 어떤 식으로 보안 리스크를 완화할 수 있을까?

- 3가지로 간추려서 말해보세요.
    1. 제로 트러스트 기반의 보안 아키텍처 도입
    - 기본 원칙: “아무도 신뢰하지 말고 항상 검증하라”
    - 사용자 및 디바이스에 대해 다요소 인증(MFA) 적용
    - 세분화된 접근 제어 정책 수립 (예: 업무별 최소 권한 부여)
    - 행위 기반 이상 탐지 시스템 구축으로 내부 위협까지 감시
    - 망분리 완화 환경에 적합한 이유: 내부망과 외부망의 경계가 흐려지므로, 모든 접속과 행위를 지속적으로 검증하는 제로 트러스트 모델이 필수적입니다.

    2. 다층 보안 체계(MLS) 설계 및 운
    - 핵심 개념: 정보의 중요도에 따라 보안 수준을 차등 적용하는 방식
    - 중요도 기반 정보 등급 분류 (기밀, 민감, 일반 등)
    - 각 등급에 맞는 접근 통제 및 암호화 정책 적용
    - 망연계 구간에 대한 실시간 모니터링 및 감사 로그 확보
    - 장점: 단일 방어선이 아닌 여러 단계의 보안 장치를 통해 침입 시도에 대한 저항력을 높일 수 있어요
