- ### DB
    - 클라우드 DB (NoSQL 기반)이 증가
    - 애플리케이션 보안의 타겟(주요 보호대상) = DB
    - DB랭킹으로 살펴보는 변천사
        1. RDB가 많이 사용이 되고 있음(Legacy)
        2. NoSQL이 빠르게 발전하고 있음(New)
        - 4가지 타입: 키밸류, 도큐먼트, 컬럼스토어, 그래프
        - RDB vs NoSQL: RDB(안정성, 표준화, SQL) vs NoSQL(속도, 비표준화, SQLX)
            - 블록체인은 NoSQL. RDB는 속도가 느린 대신, 충돌이 전혀 없어야 하는 안정성이 높다.
            - 블록체인은 모든 원장 데이터를 모든 노드가 동일하게 가지고 있어야 한다. (전파속도) 그나마 속도를 높여주는 NoSQL 사용
    - 최근 애플리케이션 환경의 변화
        - 실시간, 대용량 데이터, 스트리밍 데이터, 가상/증강/복합, NFT/스테이블코인/디지털화폐, 웹3.0(사용자 수익화), 5G네트워크 -> 7G네트워크 | 데이터는 커지고, 속도는 빨라지고, 사용자의 기대 수준이 높아지고 있음. --> NoSQL이 요구될 수 밖에 없다!
    
    - 미래를 위해 공부한다면?
        - MongoDB, Neo4j, Cassandram DynamoDB, BigQuery, NoSQL을 공부!

- ### 호스팅 vs 코토케이션

- 호스팅: 서버 자원을 임대
    - 웹 호스팅: 하나의 서버를  여러 회사가 다중 임차하는 방식(Multi-Tenancy) like as Cloud
    - 서버 호스팅: 하나의 서버를 단독 회사가 단독 임차하는 방식(single)
    - 컨테이너 호스팅: 하나의 서버에서 다수의 컨테이너를 만들고 -> 여러 회사에 다중 임차하는 방식(회사와 고객 모두 1개의 서버가 필요하여 윈윈할 수 있다.)

- 코로케이션: 공간을 임대(서버 호텔 = IDC)
    - 공간: 서버렉, 스토리지베이, 전원케이블, 송전, 항온항습, UPS(비상발전기), 조명, 이중바닥, 강화천장, 비산방지 창문, 생체인증/기록 출입문, 방화(불이 번지지 않는)벽, 경비원, 경비시스템, 스피드게이트, CCTV 등등(상면비용 = 1'U'nit)

- ### 서버 가상화 
- 베어메탈, 가상화, 컨테이너
- 베어메탈: 1개의 서버, 1개의 OS, 1개의 애플리케이션 (가상화하지 않은 물리서버) --> 최신 버전: 네이버 뉴로클라우드

- 가상화ㅣ 하이퍼바이저, 1개의 서버, 여러 개의 OS, 여러 개의 애플리케이션 | 하이퍼바이저 부담, 부하 단점, 독립성, 격리 장점(커널 분리)
    - 가상화: 서버 가상화, 스토리지 가상화, 데이터 가상화, 네트워크 기능 가상화, PC 가상화, 운영체제 가상화...

- 컨테이너: 도커 엔진, 프로세스 분리, 커널 공유, 1개의 서버, 여러 개의 OS/애플리케이션 | 부감 경감, 속도(보안 문제). 그러나 커널 공유 보안적 단점(자원을 공유할 수 있음)

### API 매개변수 변조
- OWASP Top 10, 주정통 해설서를 매핑하는 것은 서로 유사하거나 대응되는 취약점 항목을 연결해보는 작업을 의미한다. 즉, 같은 보안 문제를 서로 다른 관점에서 어떻게 정의하고 있는지를 비교하고 정리하는 것.

- 주정통 해설서 기준 API 매개변수 변조 관련 항목
    - 항목 13. 불충분한 인증: 인증 절차가 미흡하여 매개변수 조작으로 인증 
    - 항목 17. 불충분한 인가: 권한 검증이 제대로 이루어지지 않아 다른 사용자 데이터 접근 가능
    - 항목 16. 세션 예측: 세션 ID가 예측 가능하여 매개변수 조작과 함께 세션 탈취 가능
    - 항목 19. 세션 고정: 세션 ID를 고정시켜 공격자가 세션을 공유하거나 탈취 가능
    - 항목 25. 경로 추적: URL이나 파라미터를 통해 시스템 내부 경로가 노출되어 공격 가능
    - 항목 28. 쿠키 변조: 쿠키 내 정보 조작을 통해 인증/인가 우회 가능
    - 항목 21. 프로세스 검증 누락: 입력값에 대한 검증이 부족하여 파라미터 변조가 가능함
