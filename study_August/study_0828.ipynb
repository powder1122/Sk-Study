### 빌표 관련
- 사이버 킬체인을 활용한 내부침투 - 확산
- 주요 키워드: 파일 감염, AD, 파워쉘, Leteral Movement, C2 통신 등
- 파일: 오피스, 실행, 바로가기, 화면보호기, dll 라이브러리 등(더블클릭)
- AD: 연결성(중앙 집중 관리), 초연결 시대, 인프라 복잡, 하이브리드 멀티 클라우드, MSA(Docler - K8S, VMs - Hyperviser) | 망분리 시대에서 내부/외부
    - AD, 파워쉘은 지양하는 방향으로 발전 중(위험이 큼)
- 파일리스 악성코드에 주로 활용되는 내장 프로그래밍 도구(쉘)
- 크로스 플랫폼 악성코드
- 복잡한 구문을 지원 -> 프로그래밍(ps1) -> 스케쥴링 가능
-  랜섬웨어의 특징은 조건이 맞을 때 실행되는데, 스케쥴링을 통해 조건을 맞출 수 있다.
- 일반 사용자는 쓸 일이 없다. (특권 계층의 전유물)
    - 해커들은 파워쉘, 특권 계층을 노린다!
    - DB 접근 등 특권 계층만 가지는 권한들이 중요하기 때문
- Leteral Movement: 횡적 이동, 내부간 이동 | 외부 -> 내부, 내부 - 내부 (개부 간 구획, segmentation)
- C2(Command and Control): 외부 마스터 해커 ---원격 쉘--- 내부 좀비 악성코드 형식으로 커맨드를 이용하여 조작하는 통신

- 악성 코드는 살아 있을까? C2 통신 연결 - 악성코드는 일괄 배송되지 않는다 절대!

- AV는 무용지물인가? 이미 알려진 공격에 대해서는 대응/탐지가 가능하다.(유의미함) / 알려지지 않은 공격, 일반 소스코드/악성소스코드 구분이 불가능한 파일, 파일리스(다른 파일 호출하는 형태의 공격)은 막지 못함(무용)
    - 로그를 시간 기준으로 상관관계 분석: 시계열 분석 필요

- 악성코드는 대놓고 '악성'일까? 암호화, 형태 변조, 난독화, 분할, 은닉, 삭제, 파괴 등 숨겨둔 형태가 대부분이다.(탐지는 삭제가 가장 어려움. 로그만 남는다.)

- 악성코드는 얼마나 길까? 몇 줄 정도 되어야 악성 행위를 할 수 있을까? 내용이 중요하다. 코드 1~3줄이어도 상관이 없다.

- 악성코드는 어디서나 실행이 가능할까?(모바일, 윈도우 등) 시스템 OS 호환성이 보장되어야 한다. (운영체제에 종속성) 운영체제의 특수성, 기능, 특징, 원리를 보면 악성코드가 무엇을 하는지 알 수 있다.(시스템 보안의 영역)

- 가만히 있어도 악성코드는 스스로 감염을 시키나? 아니다! 사용자의 허락이 요구(UAC, User Access Control). 아무거나 클릭하지 않으면 예방할 수 있다. 
    - 근본적인 예방 방법: 아예 설치 가능한 형태의 파일 다운로드 금지! 사용자 권한은 다운 그레이드! (사용자 권한에 맞게 악성코드 권한도 다운된다)
    - 커널 기반의 보안 강화(시스템 콜 커널 접근 권한 차단)

- 신뢰할 수 없는 프로그램 vs 신뢰할 수 있는 프로그램: 코드 사이닝(코드 인증서)
    - 금 거래소에서 골드바를 사면 인증서를 준다! 그것과 같이 소프트웨어를 설치할 때 인증서에 게시자를 확인할 수 있다. 게시자가 신뢰할 수 있는 게시자 or 실행자일 때만 설치하는 것을 권장!
