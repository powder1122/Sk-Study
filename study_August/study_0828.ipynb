### 정보보호의 정의 및 개념
- 정보보호는 문서화! 
    - 모든 업무가 문서화가 되어야 한다!
    - 문서화 시에는 PDCE를 다 준수해야 한다!

- 보안에서는 구술적 증거는 증거가 안된다.(보안을 열심히 해도 문서,영상,사진 등 제출 자료가 있어야 인정된다!)
- PDCA
    - Plan: 악성코드 탐지 및 차단 계획서(Ex: AV 구축, 고도화, 탐지룰셋, ML 도입, EDR)
    - Do: 악성코드 탐지 및 차단 운영 **내역서**(EX 구축 운영 탐지 분석 대응 재발방지를 실제로 하는 것)
    - Check: 악성코드 탐지 및 차단 검증(Ex: AV 검증, 최적화, 튜닝, 보완, 업데이트, 교체)
    - Act: 악성코드 탐지 및 차단 개선(Ex: 제로 트러스트 AV, XDR 고도화, ML 회귀 AV)
    - 이런 증거물의 집합으로 '우리 보안 잘했어요' 도장 받는 것이 인증이다.
    - 이 중 하나라고 빵꾸나면 보안이 미흡했다고 판정을 받는 것이다.
    - 여러 항목에 대해서 각각 PDCA를 지켰는지 확인하고, 하나라도 미흡하면 보안이 미흡했다고 판단을 받는다. (더 큰 잘못을 한 것처럼 보일 수 있다)
    - 그만큼 PDCA를 중요하게 여긴다.

    - 이메일이 유일하게 망분리되 내부망의 기업 사용자를 감염시킬 수 있는 수단

    - 생성형 AI는 보안이 강해 침투하기 쉽지 않아 사칭, LLM 오픈소스 감염, 허깅페이스/깃헙에 악성코드를 첨부하는 등의 방식을 사용

    - 로그를 시간 기준으로 상관관계 분석: 시계열 분석 필요

- 악성코드는 대놓고 '악성'일까? 암호화, 형태 변조, 난독화, 분할, 은닉, 삭제, 파괴 등 숨겨둔 형태가 대부분이다.(탐지는 삭제가 가장 어려움. 로그만 남는다.)

- 악성코드는 얼마나 길까? 몇 줄 정도 되어야 악성 행위를 할 수 있을까? 내용이 중요하다. 코드 1~3줄이어도 상관이 없다.

- 악성코드는 어디서나 실행이 가능할까?(모바일, 윈도우 등) 시스템 OS 호환성이 보장되어야 한다. (운영체제에 종속성) 운영체제의 특수성, 기능, 특징, 원리를 보면 악성코드가 무엇을 하는지 알 수 있다.(시스템 보안의 영역)

- 가만히 있어도 악성코드는 스스로 감염을 시키나? 아니다! 사용자의 허락이 요구(UAC, User Access Control). 아무거나 클릭하지 않으면 예방할 수 있다. 
    - 근본적인 예방 방법: 아예 설치 가능한 형태의 파일 다운로드 금지! 사용자 권한은 다운 그레이드! (사용자 권한에 맞게 악성코드 권한도 다운된다)
    - 커널 기반의 보안 강화(시스템 콜 커널 접근 권한 차단)

- 신뢰할 수 없는 프로그램 vs 신뢰할 수 있는 프로그램: 코드 사이닝(코드 인증서)
    - 금 거래소에서 골드바를 사면 인증서를 준다! 그것과 같이 소프트웨어를 설치할 때 인증서에 게시자를 확인할 수 있다. 게시자가 신뢰할 수 있는 게시자 or 실행자일 때만 설치하는 것을 권장!

### [PKI] 공개키 기반 구조
- 코드 서명: 신뢰할 수 없는 소프트웨어 개발 공급자가 코드를 인증하고 우리에게 제공한다는 것(소프트웨어 개발 공급자에 대한 코드 인증 기관에서 발행한 인증서) -> MS
- 디지털 서명(전잦서명): 개념, 온라인 인감도장, 신뢰할 수 있는 제3자가 보증해준 서명키를 가지고 서명(인증)을 수행하는 것! --> 금융결제원
- 웹 SSL/TLS(인증서): 크롬/엣지 브라우저 공급자에(신뢰할 수 있는 제3의 기관)에 등록!
- 응용: 성적증명서(서명), 문서, 전자여권(위변조 여부), 서명(부인방지)
