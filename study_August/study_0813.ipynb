### 법 해석
- 법을 해석할 때 하나만 보는 것이 아니다
    - 적용받는 법이 굉장히 다양하다.
    - 일반법(일반적인 상황에서 기본, 온라인/오프라인) VS 특별법(전자금융거래, 전자서명, 정보통신망, 신용정보, 클라우드, 인공지능 등 특별한 경우에는 특별법 우선의 원칙)
    - 특별법에서 다루지 않는 법은 일반법에서 지킨다.
    - 특별법은 어떤 것들이 있을까?

        - 법학 전공하면 좋음(행정법, 민사(민법), 상법, 형사소송법)

- 동의
    - 필수(무조건, 법적 근거) vs 선택(재화 제공 기준 X, 법적근거 마케딩, 이득, 영리) | 박스를 구분
