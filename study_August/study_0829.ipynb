````python
# (A) SQL Injection (URI 내 일반 패턴)
alert tcp $EXTERNAL_NET any -> $HOME_NET $HTTP_PORTS (
    msg:"WEB-APP SQLi pattern";
    flow:to_server,established;
    http_uri;
    content:"'"; nocase;
    pcre:"/(\%27)|(\')|(\-\-)|(\%23)|(#)/Ui";
    classtype:web-application-attack;
    sid:1000001; rev:1;
)

# (B) SSH brute force (단기간 반복 연결 시도)
alert tcp $EXTERNAL_NET any -> $HOME_NET 22 (
    msg:"Possible SSH brute force";
    flow:to_server,established;
    detection_filter:track by_src, count 10, seconds 60;
    classtype:attempted-recon;
    sid:1000002; rev:1;
)

# (C) XSS 시그니처 (단순 스크립트 인젝션)
alert tcp $EXTERNAL_NET any -> $HOME_NET $HTTP_PORTS (
    msg:"WEB-APP XSS pattern";
    flow:to_server,established;
    http_uri;
    content:"<script>"; nocase;
    classtype:web-application-attack;
    sid:1000003; rev:1;
)

# (D) TCP SYN scan 의심 (SYN만 다수 포트)
# 정확도 향상을 위해 sfPortscan(또는 Snort3 port_scan inspector) 사용 권장.
alert tcp $EXTERNAL_NET any -> $HOME_NET any (
    msg:"TCP SYN scan suspected";
    flags:S; ack:0;
    threshold:type both, track by_src, count 20, seconds 60;
    classtype:attempted-recon;
    sid:1000004; rev:1;
)

# (E) DNS Tunneling 의심 (과도한 길이/엔트로피 레이블 - 단순화)
# Snort에서 DNS 질의 길이/패턴은 정규식으로 근사 탐지.
alert udp $HOME_NET any -> $EXTERNAL_NET 53 (
    msg:"Suspicious long/entropy DNS query";
    dns_query;
    pcre:"/([A-Za-z0-9+\/]{20,}\.){2,}/U";
    classtype:attempted-user;
    sid:1000005; rev:1;
)

# (F) 대용량 HTTP POST 업로드 (잠재적 유출)
# Content-Length는 http_header 영역 정규식으로 근사 탐지.
alert tcp $HOME_NET any -> $EXTERNAL_NET $HTTP_PORTS (
    msg:"Large HTTP POST possible exfil";
    flow:to_server,established;
    http_method; content:"POST"; nocase;
    pcre:"/Content-Length:\s?(?:[5-9]\d{5,}|\d{7,})/H";
    classtype:data-theft;
    sid:1000006; rev:1;
)
````

````python
'''
## 1. **YARA 룰셋이란?**

* **YARA**는 악성코드 분석 및 탐지에 활용되는 규칙 기반 엔진입니다.
* Snort, Suricata가 **네트워크 침입 탐지/차단(패킷 기반)** 에 특화되어 있다면,
  YARA는 **파일/메모리/프로세스 내 특정 패턴 탐지(콘텐츠 기반)** 에 주로 사용됩니다.
* 보통 **디지털 포렌식(DFIR)**, **악성코드 헌팅**, **엔드포인트 위협 탐지(EDR)** 에 활용됩니다.

즉, Snort는 **네트워크 트래픽 레벨**, YARA는 **파일 및 메모리 아티팩트 레벨**에서 탐지를 수행합니다.
따라서 “침해사고 대응”에서 Snort는 **실시간 네트워크 공격 탐지**, YARA는 **사후 분석 및 악성코드 패턴 탐지** 용도로 보완적 역할을 합니다.

---

## 2. **Snort 룰셋 (비정상 이벤트 탐지 예시)**

예: HTTP URI에서 `"/cmd.exe"` 문자열이 탐지되면 공격 시도로 간주.

```snort
alert tcp any any -> any 80 (msg:"Suspicious HTTP URI - Possible Command Execution"; \
content:"/cmd.exe"; nocase; http_uri; sid:1000001; rev:1;)
```

* **alert tcp any any -> any 80** : 모든 TCP 트래픽에서 80 포트(HTTP) 대상으로 검사
* **content:"/cmd.exe"; http\_uri;** : HTTP 요청 URI 내 `cmd.exe` 문자열 탐지
* **nocase** : 대소문자 구분 안 함
* **sid** : Snort Rule 고유 ID

---

## 3. **동일 탐지를 위한 YARA 룰셋 예시**

* 목적: 파일, 메모리, 로그 덤프에서 공격 흔적 탐지
* 예: 악성 URI 문자열(`cmd.exe`, `nc.exe`)을 포함한 파일이나 로그를 탐지

```yara
rule Suspicious_Command_URI
{
    meta:
        description = "Detects suspicious command execution attempts in logs/files"
        author = "SecurityResearcher"
        date = "2025-08-29"
        reference = "Snort to YARA translation example"

    strings:
        $cmd1 = "/cmd.exe"
        $cmd2 = "nc.exe"
        $cmd3 = "powershell -enc"

    condition:
        any of ($cmd*)
}
```

---

## 4. **비교: Snort vs YARA**

| 구분    | Snort            | YARA                     |
| ----- | ---------------- | ------------------------ |
| 대상    | 네트워크 패킷 (실시간)    | 파일, 메모리, 로그 덤프           |
| 활용 시점 | 공격 발생 시 즉각 탐지    | 사후 분석, 악성코드 헌팅           |
| 룰 형태  | 패킷 레벨 매칭         | 문자열/바이너리 패턴 매칭           |
| 적용 예  | `/cmd.exe` 요청 탐지 | 로그/파일 내 `/cmd.exe` 흔적 탐지 |

---

## 5. **추가 고려사항**

* 실제 대응 환경에서는 **Snort + YARA 병행**이 일반적입니다.

  * Snort → 네트워크 레벨에서 실시간 탐지 및 차단
  * YARA → 로그/파일/메모리 덤프에서 공격 흔적 사후 검증
* 따라서 프로젝트를 진행하실 때,

  * (1) Snort 룰은 **실시간 탐지 및 이벤트 발생**에,
  * (2) YARA 룰은 **분석 아티팩트 검증**에 적용하는 방식이 가장 현실적입니다.

---
'''
````

````python
# -*- coding: utf-8 -*-
"""
IDS 이벤트 로그와 탐지 패킷 로그의 '현실적인' 샘플을 생성하고,
정상/비정상 시나리오를 포함하여 특징을 추출하는 예제입니다.

산출물:
- /mnt/data/ids_events.csv
- /mnt/data/packet_logs.csv
- /mnt/data/features.csv

표시는 상위 일부 행만 화면에 보여주며, 전체는 CSV로 저장합니다.
"""

import pandas as pd
import numpy as np
import ipaddress
import random
import string
import math
from datetime import datetime, timedelta
from collections import Counter

rng = np.random.default_rng(42)
random.seed(42)

# -----------------------------
# 유틸 함수
# -----------------------------
def rand_ip(cidr: str) -> str:
    """지정된 CIDR 블록에서 임의 IP 생성 (v4 테스트넷 포함)"""
    net = ipaddress.ip_network(cidr)
    host_count = net.num_addresses - 2 if net.version == 4 else net.num_addresses
    if host_count <= 0:
        return str(net.network_address + 1)
    offset = int(rng.integers(1, host_count + 1))
    return str(net.network_address + offset)

def shannon_entropy(s: str) -> float:
    """문자열의 샤논 엔트로피 (bit/char) - DNS 터널링 탐지 지표로 사용"""
    if not s: return 0.0
    total = len(s); ent = 0.0
    for c in set(s):
        p = s.count(c) / total
        ent -= p * math.log2(p)
    return ent

def random_domain(common=True):
    """일반 도메인 vs 난수형(튜널링 의심) 도메인 생성"""
    common_domains = [
        "google.com","naver.com","daum.net","github.com","microsoft.com",
        "apple.com","cloudflare.com","amazon.com","kakao.com","zoom.us"
    ]
    if common or rng.random() < 0.8:
        sub = "".join(rng.choice(list(string.ascii_lowercase), size=int(rng.integers(3,8))))
        return f"{sub}.{rng.choice(common_domains)}"
    else:
        label_len = int(rng.integers(20, 60))
        label = "".join(rng.choice(list(string.ascii_letters + string.digits + "+/"), size=label_len))
        return f"{label}.exfil.example.net"

def random_uri(normal=True):
    """정상/공격적 HTTP URI 생성"""
    paths = ["/","/index.html","/login","/search","/products","/api/v1/items","/news","/about"]
    if normal:
        q = rng.choice(["","?q=test","?page=1","?lang=ko","?ref=news"])
        return rng.choice(paths) + q
    # 공격 페이로드 (SQLi/XSS 예시)
    sqli_payloads = [
        "/login?user=admin' OR '1'='1'--",
        "/search?q=%27%20OR%201%3D1--",
        "/api/v1/items?sort=price;DROP%20TABLE%20users;--"
    ]
    xss_payloads = [
        "/search?q=<script>alert(1)</script>",
        "/comment?msg=%3Cimg%20src%3Dx%20onerror%3Dalert(1)%3E"
    ]
    return rng.choice(sqli_payloads + xss_payloads)

def pick_port(proto="tcp", normal=True):
    """서비스 포트 선택"""
    if proto == "tcp":
        if normal and rng.random() < 0.9:
            return int(rng.choice([80,443,22,8080,3306,5432,6379]))
        return int(rng.integers(1, 65535))
    if proto == "udp":
        if normal and rng.random() < 0.9:
            return int(rng.choice([53,123,500,4500,161]))
        return int(rng.integers(1, 65535))
    return int(rng.integers(1, 65535))

def tcp_flags(normal=True, syn_only=False):
    """TCP 플래그 패턴"""
    if syn_only: return "S"
    if normal:
        return rng.choice(["PA","PA","PA","FA","RA","A"])
    return rng.choice(["S","PA","RA","FA"])

def gen_time_series(start_dt: datetime, n: int, max_jitter_sec=3):
    """증분 시간 + 지터로 현실감 확보"""
    times = []; t = start_dt
    for _ in range(n):
        t += timedelta(milliseconds=int(rng.integers(1, 300)))
        t += timedelta(seconds=float(rng.random() * max_jitter_sec))
        times.append(t)
    return times

# -----------------------------
# 파라미터 & 시나리오 정의
# -----------------------------
START = datetime(2025, 8, 29, 9, 0, 0)  # Asia/Seoul 09:00
HOME_NET = "10.0.0.0/24"
EXTERNAL_NET = "203.0.113.0/24"  # TEST-NET-3
SENSOR = "sensor-ids-01"

N_NORMAL = 800
ATTACKS = {
    "SSH_BRUTE": 90,       # 1 IP가 60s 내 다량 시도
    "SQLI_HTTP": 60,       # SQLi URI/패턴
    "XSS_HTTP": 50,        # XSS 스크립트 패턴
    "PORT_SCAN": 80,       # SYN-only 다수 포트
    "DNS_TUNNEL": 40,      # 긴/고엔트로피 쿼리
    "HTTP_POST_EXFIL": 30  # 대용량 POST 업로드
}

# -----------------------------
# 1) 정상 트래픽 생성
# -----------------------------
packet_rows, event_rows = [], []
packet_id = 1

times_normal = gen_time_series(START, N_NORMAL)
for ts in times_normal:
    proto = rng.choice(["tcp","udp","icmp","tcp","tcp","udp"])
    if proto == "icmp":
        src_ip, dst_ip = rand_ip(HOME_NET), rand_ip(EXTERNAL_NET)
        src_port = dst_port = 0; flags = ""; ttl = int(rng.integers(50,64))
        win = 0; payload_len = int(rng.integers(32,128))
        http_method=http_uri=dns_query=dns_type=tls_sni=""
    else:
        # 내부→외부 60%, 외부→내부 40%
        if rng.random() < 0.6:
            src_ip, dst_ip = rand_ip(HOME_NET), rand_ip(EXTERNAL_NET)
        else:
            src_ip, dst_ip = rand_ip(EXTERNAL_NET), rand_ip(HOME_NET)
        src_port = int(rng.integers(1024,65535))
        dst_port = pick_port(proto, normal=True)
        flags = tcp_flags(normal=True) if proto=="tcp" else ""
        ttl = int(rng.integers(50,128))
        win = int(rng.integers(1024,65535)) if proto=="tcp" else 0
        payload_len = int(rng.integers(50, 1500 if dst_port in [80,443,8080] else 400))
        http_method=http_uri=dns_query=dns_type=tls_sni=""

        # HTTP(S)
        if proto=="tcp" and dst_port in [80,8080,443]:
            http_method = rng.choice(["GET","GET","GET","POST"])
            http_uri = random_uri(normal=True)
            if dst_port==443 and rng.random()<0.7:
                tls_sni = random_domain(common=True)

        # DNS
        if proto=="udp" and dst_port==53:
            dns_query = random_domain(common=True)
            dns_type = rng.choice(["A","AAAA","CNAME","MX"])

    # 패킷 로그
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": src_ip, "src_port": src_port, "dst_ip": dst_ip, "dst_port": dst_port,
        "proto": proto, "tcp_flags": flags, "ttl": ttl, "win": win, "payload_len": payload_len,
        "http_method": http_method, "http_uri": http_uri,
        "dns_query": dns_query, "dns_type": dns_type, "tls_sni": tls_sni,
        "label": "normal"
    })

    # 드물게 정보성 이벤트
    if rng.random() < 0.02:
        event_rows.append({
            "timestamp": ts.isoformat(), "sensor": SENSOR,
            "sig_id": 2000000, "sig_name": "Allowed traffic (informational)",
            "category": "protocol-activity", "severity": 1,
            "src_ip": src_ip, "src_port": src_port, "dst_ip": dst_ip, "dst_port": dst_port,
            "proto": proto, "action": "allow", "packet_id": packet_id
        })
    packet_id += 1

# -----------------------------
# 2~3) 비정상 시나리오 생성
# -----------------------------

# SSH brute force
ts = times_normal[-1] + timedelta(seconds=1)
attacker_ssh = rand_ip(EXTERNAL_NET); victim_ssh = rand_ip(HOME_NET)
for _ in range(ATTACKS["SSH_BRUTE"]):
    ts += timedelta(milliseconds=int(rng.integers(50,300)))
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": attacker_ssh, "src_port": int(rng.integers(1024,65535)),
        "dst_ip": victim_ssh, "dst_port": 22,
        "proto": "tcp", "tcp_flags": ("S" if rng.random()<0.6 else "PA"),
        "ttl": int(rng.integers(40,64)), "win": int(rng.integers(1024,65535)),
        "payload_len": int(rng.integers(0,60)),
        "http_method":"", "http_uri":"", "dns_query":"", "dns_type":"", "tls_sni":"",
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000002, "sig_name": "Possible SSH brute force",
        "category": "attempted-recon", "severity": 2,
        "src_ip": attacker_ssh, "src_port": 0, "dst_ip": victim_ssh, "dst_port": 22,
        "proto": "tcp", "action": "alert", "packet_id": packet_id
    })
    packet_id += 1

# SQLi
ts += timedelta(seconds=1)
victim_web = rand_ip(HOME_NET)
for _ in range(ATTACKS["SQLI_HTTP"]):
    ts += timedelta(milliseconds=int(rng.integers(30,200)))
    http_uri = random_uri(normal=False)
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": rand_ip(EXTERNAL_NET), "src_port": int(rng.integers(1024,65535)),
        "dst_ip": victim_web, "dst_port": 80,
        "proto":"tcp","tcp_flags":"PA","ttl": int(rng.integers(40,64)),
        "win": int(rng.integers(1024,65535)),"payload_len": int(rng.integers(200,900)),
        "http_method":"GET", "http_uri": http_uri,
        "dns_query":"", "dns_type":"", "tls_sni":"",
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000001, "sig_name": "WEB-APP SQLi pattern",
        "category": "web-application-attack", "severity": 3,
        "src_ip": "EXTERNAL", "src_port": 0, "dst_ip": victim_web, "dst_port": 80,
        "proto":"tcp", "action":"drop", "packet_id": packet_id
    })
    packet_id += 1

# XSS
ts += timedelta(seconds=1)
for _ in range(ATTACKS["XSS_HTTP"]):
    ts += timedelta(milliseconds=int(rng.integers(30,200)))
    http_uri = random_uri(normal=False)
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": rand_ip(EXTERNAL_NET), "src_port": int(rng.integers(1024,65535)),
        "dst_ip": victim_web, "dst_port": 80, "proto":"tcp","tcp_flags":"PA",
        "ttl": int(rng.integers(40,64)),"win": int(rng.integers(1024,65535)),
        "payload_len": int(rng.integers(200,700)),
        "http_method":"GET", "http_uri": http_uri,
        "dns_query":"", "dns_type":"", "tls_sni":"",
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000003, "sig_name": "WEB-APP XSS pattern",
        "category": "web-application-attack", "severity": 2,
        "src_ip": "EXTERNAL", "src_port": 0, "dst_ip": victim_web, "dst_port": 80,
        "proto":"tcp", "action":"alert", "packet_id": packet_id
    })
    packet_id += 1

# Port scan (SYN-only)
ts += timedelta(seconds=1)
scanner = rand_ip(EXTERNAL_NET); target = rand_ip(HOME_NET)
for _ in range(ATTACKS["PORT_SCAN"]):
    ts += timedelta(milliseconds=int(rng.integers(5,30)))
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": scanner, "src_port": int(rng.integers(1024,65535)),
        "dst_ip": target, "dst_port": int(rng.integers(1,65535)),
        "proto": "tcp", "tcp_flags": "S",
        "ttl": int(rng.integers(30,60)),"win": int(rng.integers(1024,65535)),
        "payload_len": 0,
        "http_method":"", "http_uri":"", "dns_query":"", "dns_type":"", "tls_sni":"",
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000004, "sig_name": "TCP SYN scan suspected",
        "category": "attempted-recon", "severity": 2,
        "src_ip": scanner, "src_port": 0, "dst_ip": target, "dst_port": 0,
        "proto": "tcp", "action": "alert", "packet_id": packet_id
    })
    packet_id += 1

# DNS tunneling (긴/고엔트로피 레이블)
ts += timedelta(seconds=1)
client = rand_ip(HOME_NET); resolver = rand_ip(EXTERNAL_NET)
for _ in range(ATTACKS["DNS_TUNNEL"]):
    ts += timedelta(milliseconds=int(rng.integers(50,120)))
    label_len = int(rng.integers(40,70))
    label="".join(rng.choice(list(string.ascii_letters + string.digits + "+/"), size=label_len))
    domain=f"{label}.{''.join(rng.choice(list(string.ascii_lowercase), size=6))}.exfil.example.net"
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": client, "src_port": int(rng.integers(1024,65535)),
        "dst_ip": resolver, "dst_port": 53,
        "proto":"udp", "tcp_flags":"", "ttl": int(rng.integers(40,64)), "win": 0,
        "payload_len": int(rng.integers(50,200)),
        "http_method":"", "http_uri":"", "dns_query": domain, "dns_type":"TXT", "tls_sni":"",
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000005, "sig_name": "Suspicious long/entropy DNS query",
        "category": "attempted-user", "severity": 2,
        "src_ip": client, "src_port": 0, "dst_ip": resolver, "dst_port": 53,
        "proto":"udp", "action":"alert", "packet_id": packet_id
    })
    packet_id += 1

# HTTP POST exfiltration (대용량 업로드)
ts += timedelta(seconds=1)
exfil_client = rand_ip(HOME_NET); exfil_server = rand_ip(EXTERNAL_NET)
for _ in range(ATTACKS["HTTP_POST_EXFIL"]):
    ts += timedelta(milliseconds=int(rng.integers(60,180)))
    packet_rows.append({
        "packet_id": packet_id, "timestamp": ts.isoformat(), "sensor": SENSOR,
        "src_ip": exfil_client, "src_port": int(rng.integers(1024,65535)),
        "dst_ip": exfil_server, "dst_port": 80,
        "proto":"tcp","tcp_flags":"PA","ttl": int(rng.integers(40,64)),
        "win": int(rng.integers(1024,65535)), "payload_len": int(rng.integers(300000,1200000)),
        "http_method":"POST","http_uri":"/upload",
        "dns_query":"", "dns_type":"", "tls_sni": random_domain(common=False),
        "label": "abnormal"
    })
    event_rows.append({
        "timestamp": ts.isoformat(), "sensor": SENSOR,
        "sig_id": 1000006, "sig_name": "Large HTTP POST possible exfil",
        "category": "exfiltration", "severity": 3,
        "src_ip": exfil_client, "src_port": 0, "dst_ip": exfil_server, "dst_port": 80,
        "proto":"tcp", "action":"alert", "packet_id": packet_id
    })
    packet_id += 1

# -----------------------------
# DataFrame 및 특징 추출
# -----------------------------
packets = pd.DataFrame(packet_rows).sort_values("timestamp").reset_index(drop=True)
events  = pd.DataFrame(event_rows).sort_values("timestamp").reset_index(drop=True)

# 패킷 단위 파생 특징
packets["is_http"] = ((packets["proto"]=="tcp") & (packets["dst_port"].isin([80,8080,443]))).astype(int)
packets["is_dns"]  = ((packets["proto"]=="udp") & (packets["dst_port"]==53)).astype(int)
packets["is_ssh"]  = ((packets["proto"]=="tcp") & (packets["dst_port"]==22)).astype(int)
packets["syn_only"] = (packets["tcp_flags"]=="S").astype(int)
packets["dns_q_len"] = packets["dns_query"].fillna("").apply(len)
packets["dns_q_entropy"] = packets["dns_query"].fillna("").apply(shannon_entropy)
packets["uri_len"] = packets["http_uri"].fillna("").apply(len)
# URI 내 위험 패턴(간단화)
packets["susp_uri_pattern"] = packets["http_uri"].fillna("").str.contains(
    r"(\%27)|(\')|(\-\-)|(\%23)|(#)|(<script>)", case=False, regex=True
).fillna(False).astype(int)

# 60초 창 기반 트래픽 지표 (src_ip)
packets["timestamp_dt"] = pd.to_datetime(packets["timestamp"])
packets.set_index("timestamp_dt", inplace=True)
window = "60s"
grp = packets.groupby("src_ip")
packets["rate_by_src_60s"] = grp["packet_id"].apply(lambda s: s.rolling(window=window).count()).reset_index(level=0, drop=True)
packets["dst_ports_unique_60s"] = grp["dst_port"].apply(
    lambda s: s.rolling(window=window).apply(lambda x: len(pd.unique(x)), raw=False)
).reset_index(level=0, drop=True)
packets.reset_index(drop=False, inplace=True)

# 최종 특징 표
feature_cols = [
    "is_http","is_dns","is_ssh","syn_only","payload_len","ttl","win",
    "dns_q_len","dns_q_entropy","uri_len","susp_uri_pattern",
    "rate_by_src_60s","dst_ports_unique_60s"
]
features = packets[["packet_id","timestamp","src_ip","dst_ip","dst_port","proto","label"] + feature_cols].copy()

# 저장
packets.to_csv("/mnt/data/packet_logs.csv", index=False)
events.to_csv("/mnt/data/ids_events.csv", index=False)
features.to_csv("/mnt/data/features.csv", index=False)
````
