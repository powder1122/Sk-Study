### 사이버 킬체인
1. Reconnaissance: 이메일 주소, 컨퍼런스 정보 등을 수집한다. 대상을 선택하여 그 대상에 대해서만 수집. ROI를 따져서 대상 물색. 주로 중소기업을 타깃. 특정 국가 혹은 국가의 지원을 받는 경우는 대기업이나 국가를 대상으로도 함. 소버린AI
2. Weaponization: Exploit과 백도어를 결합하여, Payload에 삽입한다.
3. Delivery: 무기화된 파일을 이메일이나, 웹 또는 USB 등을 경유하여 Victim 시스템에 유포한다.
4. Exploitation: Victim 세스템에서 코드를 실행하기 위해 취약점을 이용한다. 메일을 클릭하도록 유도하는 등의 방법
5. Installation: 공격 대상 자산에 악성 프로그램을 설치한다.
6. Command & Control: Victim을 원격 조작하기 위해 채널을 열어 지휘한다.(쉘을 딴다! --> SSL-VPN을 사용하면 쉽다!)
7. Action on Objectives: 키보드를 직접 다루는 것과 같은 접근이 가능하게 되면, 침입자는 본연의 목적을 달성한다. 쉘 히스토리, WORM, 쓰기 방지 전용 메모리 등으로 대처해야 한다.
