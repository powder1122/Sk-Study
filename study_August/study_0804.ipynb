1. 인터넷이란?
- 전세계 초연결 하는 매체
- 일반: 연결, 편의성, 속도, 품질 | 5G 네트워크
- 보안: 연결(취약점), 편의성(보안성 반비례), 악, 원격 조작, 속도/품질/암호화(해커들도 사용)

- 표층웹: Surface(표층, 표면) --> Attack Surface | 공개된 웹 서버
    - 공개된: 구글봇에 접근이 허용된 웹 서버! 이를 차단한 사이트는 보안 강조된 사이트라고 판단할 수 있다.
    - 공개된 표층웹 모의해킹 포인트 발굴: ASM(Attack Surface Management)
- 딥웹: 구글봇의 접근 차단 웹 서버 | 그룹웨어, 인사시스템, 보안솔루션, 내부 인하우스 시스템

- 다크웹: 일반적인 브라우저/네트워킹 접속 불가, 거래 시에만 잠시 존재하는 웹(호스팅) like as 서버리스

- 현재 해커들은 특정 회사/기관/개인보다 훨씬 더 많은 정보를 수집해서 공격을 수행한다.
- 제로 트러스트 보안이 강조되는 이유(SKT, 랩서스해킹)

- **어택 서페이스 매니지먼트** --> 웹 보안의 중추는 어택 서페이스 관리이다.

    2. 레벨2. 이해는 되는데, 실제 무슨 내용인지 잘 모른다.
        - 가이드, 안내서, 해설서, 벤더 블로그, 보안 뉴스를 읽고 이해한 바를 정리해야 함

    3. 레벨3. 이해도 되고 무슨 내용인지도 안다. 그러나 실제로 해보진 않았다.
        - 회사에 입사하면, 솔루션 Deep Dive 해볼 수 있는 경험을 쌓아야 함(필수)
        - Q. 깊게 공부해야 하나요? VS 넓게 공부해야 하나요?
            - 개인의 선택! 장기적으로는 둘 다 필요
        - Q. 외산 솔루션을 공부해야 하나요? (비용 과다, 기능 다수, 성능 및 품질) VS 국내 솔루션을 공부해야 하나요? (비용 절감, 기본 기능, 기술 자원 및 유지보수)
            - 개인 선택! 외산을 주로 사용하는 곳은 스타트업, 대기업, 금융, 글로벌 / 국산을 사용하는 곳은 공공, 중소기업, 국내기업 (어디로 가고 싶은지에 따라 달라짐)
    
    4. 레벨4. 기업의 보안 담당자가 되기 위해서는 어떤 것을 추가로 공부해야 하나요?
        - 규제 및 표준의 준수가 필요: 암호 알고리즘(K-CMVP), CC 인증, 클라우드 보안 인증(CSAP) 등
        - 제안 요청서 작성 및 보안 솔루션 PoC/BMT(입착 및 기능, 성능 검토)
        - 회사 환경에 맟춰 도입 시 고려사항 발굴, 실제 도입 시 성능 검토, 보안성 검토

- 앞으로는 업무 환경에 노트북/데스크탑/모바일/태블릿/VM/컨테이너 접근 - **CNAPP**(묶어서 관리)

- 파일 악성코드: PC 사용자가 더블클릭을 파일(.EXE, .scr, .LNK, .hta, .py, .ps1 등 다양한 확장자)을 실행하면 감염. 
- 파일리스 악성코드: 파일이 없다ㅏ. 오피스 파일(.xlsx, .pptx, .hwpx 확장 제한)을 실행하면 외부 파일 참조(OLE) 실행 --> 매크로, 프로시저, ActiveX 구동 -> .dll파일 불러오기 실행 --> 오피스 파일 삭제 --> 쉘코드(파워쉐르 wmic, cmd) 실행

- 이러한 파일들을 차단하기 위한 것이 망분리이다.

- 국내는 여러 망으로 분리가 되어 있다. 다른 망으로는 접근할 수 없다.

- 전문성은 업에 대한 전문성과 보안에 대한 전문성이 합쳐지면서 나온다. 이직을 해도 업을 바꾸면 전문성이 떨어진다.


## IP와 DNS
- IP: 공인 IP VS 사설 IP
    - 공인 IP: inter networking 목적
    - 사설 IP: intra networking 목적

- 공인 IP를 사용하는 것을 인터넷
- 사설 IP를 사용하는 것을 인트라넷

- 공인IP - 사설IP 사이를 변환해주는 것이 NAT
- DNS: IP --> URL

- NAT와 DNS는 모두 매핑 테이블을 관리한다.
- DNS의 재귀적 호출 구조 방식
    1. 로컬 호스트: hosts파일 'DNS 포이즈닝, 파밍에 활용' - 쓰기 금지(대응)
    2. 내부 DNS 서버: A record 값, 변조되지 않도록 보안 필요 - 쓰기 제한(대응) 보안 DNS 서버
    3. 외부 DNS 서버
        - 3-1: [.]ICANN - 루트 도메인에 쿼리. 
        - 3-2: [.com] ISP - TLD 도메인
        - 3-3: [naver.com] - 도메인 주소 판매업자(IP)
    4. 얻은 값을 캐싱(다음 접근 시에 바로 연결 가능하도록)
- HTTP vs HTTPS: SSL/TLS

- SSL Proxy(SSL 복호화)
    - 외부 ---HTTPS---Proxy---HTTP---내부
    - SSL 핸드쉐이크에 개입해서, 암호화된 트래픽을 복호화하고, 트래픽 내의 이상징후(내부정보유출, 악성코드 유입)를 Inspection 하기 위해서 --> 보안 솔루션 전달!
