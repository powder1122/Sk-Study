    1. 보안 관련된 주제: 탐지, 분석, 모니터링, 컨설팅
    - 최종 프로젝트: 모의해킹, 취약점 진단(관련성) <-- Security for AI, AI for Security
    2. 생성형 AI or 인공지능(머신러닝) 가능하면 사용

    - 개발, 등록, 웹, 도구(솔루션), CS 에이전트, 대시, API --> 완성도 챙기기

    - 1개에서 2개정도 대안을 가지고 진행할 것. 충분한 협의
    5. 쓸모가 있으려면?

    - 다만, 만든 솔루션을 무료로 사용할 수 있는 중소기업 --> 쓸모 있음
    - 너무 복잡하면 쓸모가 없음 --> 더블클릭, URL,
