## 1교시 
### 금융 보안(중요 시스템의 주요 특징)
- 높은 보안성과 신뢰성 요구
    - 인증 및 제어 구조
    - 100%의 인증 및 인가 시스템이 가능하면, 세상에서 가장 안전한 시스템

- 24시간 무중단 운영 필요
    - 인터넷 뱅킹, 모바일 뱅킹 등 항시 운영이 기본
    - 시스템 점검 시에도 대체 경로 제공 필수
    - 재해복구(DR) 시스템 운영이 일반적

- 정밀한 트랜젝션 처리
    - 수많은 거래 내역이 초단위로 발생
    - 오류 발생 시 회계 처리 오류로 연결되며, 법적 분쟁 가능성

- 다단계 인증 및 권한 제어 구조
    - 민감한 개인-기업 금융정보를 다루므로 강력한 사용자 인증, 접근 제어, 로깅 시스템 필요

- 엄격한 규제 및 법률 적용
    - 전자금융감독규정, 개인정보보호법, 자본시장법 등 다양한 법률 적용
    - 감독기관 감사 및 인증(ISMS-P 등) 대응 필수

- 내부자 위협 가능성 존재
    - 내부 개발자, 운영자, 외주 업체 등도 보안 위협이 될 수 있음
    - 보안 로그 분석, 권한 최소화, 작업 모니터링 필요

- 외부 연계 시스템과 복잡한 인터페이스
    - 카드사, PG사, 공인인증기관, 정부기관 등과의 다양한 연계 인터페이스
    - 외부 API 호출 시 암호화, 인증, 검증 체계가 필수
    - MSA API를 연계하여 복잡한 단계로 작업을 수행하는 형식으로 발전



### 금융 시스템 개발 시 고려할 주요 보안 이슈
1. 입력값 검증
    - SQL Injection, XSS, Command Injection 등 기본 취약점 방지
    - 중요 정보를 하나도 안가지는 등의 조건을 만들 수 있다면 안전한 시스템이다!
    - 사용자 입력값에 대한 화이트리스트 기반 검증 필수
    - 허락된 입력값 이외의 입력값을 모두 무시하는 방식 - 완벽한 입력값 검증. 예시: ID에 숫자만 입력할 수 있도록 정책 설정(화이트리스트)

2. 인증 및 권한 관리
    - 로그인, 이체 인증 등 다중 인증(MFA) 적용
    - 세션 관리: 타임아웃 설정, 세션 고정 방지
    - 권한 상승 방지: 관리자/사용자/외부 연동 구분 철저
    - 인증이 통과가 되면 자동적으로 인가가 되는 경우가 많다. 이는 보안 허점이 된다!

3. 데이터 보호
    - 개인정보 및 금융정보(주민등록번호, 계좌번호, 신용카드번호 등) 저장 시 반드시 암호화
    - 통신 시 TLS 1.2 이상 강제 적용
    - 민감정보 조회/수정 이력 로깅

4. 로깅 및 강사 추적
    - 누가, 언제, 어떤 데이터를 조회/변경했는지 기록
    - 로그 조작 방지를 위한 무결성 보장
    - 일정 기간 이상 보관 및 위변조 감지 기능 필요

5. 모바일 API 보안
    - 앱 위변조 방지(루팅 탐지, 난독화 등)
    - 금융 API 요청 시 서명, 토큰 기반 인증
    - OpenAPI 연동 시 CORS, Rate, limiting 설정

6. 외부 연계 시스템 보안
    - 공공기관, 카드사, PG사 등과의 연동 보안
    - API 인증키 관리, IP 허용 목록 제한
    - 외부 시스템의 응답 데이터라도 검증 필요

7. 개발환경 및 코드 보안
- 개발환경 및 코드 보안, 보안 테스트 및 점검, 규제 및 인증 대응
    - **소프트웨어 공급망 보안**: 공급망 라인에 대해서 공격이 발생하는 것을 공급망 공격이라 한다. 개발부터 배포까지를 공급망 라인으로 묶는다. 리소스가 공급되는 라인 전체에 대해서 보안 설정 필요!
    - 배포 서비스 전, 또는 운영 중 인증/신규 규제/심사 등의 점검이 필요
    - 보안 컨설팅은 개발 전 개발 환경에 대해 보안 아키텍처 구축

### 2025년 보안 이슈
- 명확한 유출 사고라고 판명되지 않으면 신고하지 않기 때문에 유출 사고 신고 수는 그리 많지 않다.
- 개인정보보호위원회에서 최근 3년간 유출 신고 현황, 사고 원인 등의 보고서를 확인할 수 있음

- 보안 약점(weakness) vs 보안취약점(vulunervility)
    - 소스코드에서 버그가 발생할 수 있음. 소스코드 보안 약점 진단이 필요
    - 소스코드보안약점진단원에서 진단 실시하여 버그를 잡아낸다. 그러나 미처 잡지 못한 버그가 보안약점으로 남아있게 된다.(잠재적 위험) 이 버그가 발견된다면 보안 취약점이 된다.
    - CWE: 소프트웨어가 가질 수 있는 소스코드의 보안 약점을 정리한 데이터베이스
    - 개발자는 보안 전문가가 아니기에 취약점이 많이 나올 수 있다.
        - 서비스가 정상적으로 동작될 수 있도록 테스트를 함
        - 보안 담당자 입장에서는 비정상 동작에 대해서 테스트를 함

- 지능형 지속 공격 (APT) 고도화
    - 공격 대상 및 방식의 정교화(공급망 공격, 제로데이 취약점 활용)
    - 탐지 우회 기술 발전 (난독화, 행위 기반 분석 회피)
    - 금융 시스템 내부 침투 후 장기간 은닉 및 중요 정보 탈취 시도 증가

- 랜섬웨어 공격의 진화
    - 데이터 암호화 넘어 데이터 유출 후 협박(이중 협박) 일반화
    - APT와 마찬가지로 금전적 요구를 주로 목적으로 하는 공격
    - 특정 금융 기관을 타겟으로 하는 맞춤형 랜섬웨어 공격 증가
    - 클라우드 환경 및 백업 시스템을 대상으로 하는 공격 위협 증대
    - 랜섬웨어 감염 시: 이미 공격자가 암호화를 진행한 상황. 암호키를 받지 않는 이상 백업 데이터를 이용하여 복구해야 함. 백업 외에는 방법이 없기 때문에 백업 시스템도 공격하는 경우가 증가

- API 보안 취약점 악용 증가
    - 인증 및 권한 부여 미흡으로 인한 데이터 유출 및 변조 위험
    - API 엔드포인트 공격을 통한 서비스 거부(DoS/DDoS) 시도 증가
    - 안전하지 않은 API 설계 및 구현으로 인한 보안 허점 발생

- 클라우드 환경의 새로운 위협
    - 잘못된 클라우드 설정 및 관리 부실로 인한 정보 유출(API 키 노출 등)
    - 클라우드 서비스 제공 업체의 보안 취약점 악용
    - 컨테이너 및 쿠버네티스 환경(자동 배포)의 보안 취약점 증가 (IaC)
    - IAM 정책 실수, 중요 정보 노출 등으로 보안 사고가 지속적으로 발생

- 소셜 엔지니어링 및 내부자 위협 증가
    - 피싱, 스미싱 등 지능적인 사회 공학 기법을 통한 정보 탈취 시도 증가
    - 악의적인 내부자에 의한 정보 유출 및 시스템 파괴 위협 지속
    - 퇴직자 계정 관리 소홀로 인한 보안 사고 발생 가능성

- 인공지능(AI) 및 머신러닝(ML) 기반 공격 등장
    - AI를 활용한 지능적인 피싱 및 악성코드 생성
    - ML 모델의 취약점을 이용한 새로운 공격 방식 출현 가능성

- 사회 공학적 공격 기법!!
    - 사람을 속이는 기법.(사기)
    - 크로스 사이트 스크립트(XSS) 공격이 예시. 클릭을 하도록 유도해야 함


### 사고사례를 통한 개발보안 이해
- 시큐어한 소프트웨어란? 
    - 안전한 소프트웨어?
- 조건
    1. 코드 개발 단계
        - 실수나 에러가 없어야 함
    2. 개발 후 테스트 단계
        - 결함이나 결점, 버그가 없어야 함
    3. 배포 후 서비스 단계(운영)
        - 장애나 문제가 없어야 함

- 개발자의 사소한 실수 - 시큐어코딩이 적용되지 않은 스마트 컨트랙트
    - 블록체인 뉴스 - SMT 사태
    - 오버플로우(unsigned int)
        - unsigned int는 8비트 중에서 첫 비트가 0으로 고정인데, 최대치 이상으로 증가하면 0이 되는 현상을 말함
    - OSS(Open Source Software) 취약점 공격
        - Log4j 취약점
        - 국내 상용 소프트웨어의 코드 베이스 중 70~80%는 오픈소스, 그 중 약 50%는 고위험군의 보안 취약점을 포함

- 공급망 공격 대응 방안
    - SBOM + 지속적인 모니터링 + 자동화된 취약점 관리(예방단계 + 탐지단계)
    - SBOM: 소프트웨어를 빌드하기 위해 사용된 여러 컴포넌트에 대한 내용과 공급망 관계를 형식에 맞춰 기록하는 것. 소프트웨어의 모든 요소에 대한 명세서로 한 곳에 모아 관리하기 위한 목적.
    1. 예방단계(공급망 보안 강화)
        - 코드 및 빌드 보안, 코드 서명, SAST/DAST/SCA 도구 활용, 서드파티 및 오픈소스 관리, SBOM 관리, 서드파티 벤더 보안 평가, 의존성 최소화
 
    2. 탐지단계(공격 조기 식벽)
        - 무결성 모니터링, 레지스트리 및 저장소 모니터링, 위협 인텔리전스(CTI) 활용, SIEM 연동

    3. 대응단계(공격발생시)
        - 악성 업데이트 차장, 의존성 롤백, 사고대응프로세스(IRP), 법적/규제 보고

    4. 복구 및 사후 관리
        - 포렌식 분석, 보안 아키텍처 개선, 정기 보안 훈련, 정책 및 표준화
