### 보안관제의 정의와 개념
- 보안관제(Security Monitoring & Control)

- 네트워크 트래픽 분석도구를 이용하여 24시간 서버와 네트워크를 통해 통신한 방대한 데이터에서 잠재적인 침입자의 공격 시도를 규명
- 이 과정에서 분석된 내용을 토대로 불명확했던 침입시도를 규명하는 일련의 행위

- 정책 위반 또는 침입으로부터 시스템과 네트워크 자원의 손상을 막기 위해 모든 시스템을 실시간으로 모니터링하여 즉각 대응할 수 있도록 하는 일련의 활동

- 사이버 공격을 실시간으로 즉시 탐지 및 분석, 대응하는 일련의 활동

- 요즘에는 모니터링에 탐지-대응이 추가된 개념으로 정착. 정리하면, 정보통신망 및 정보시스템에서 발생되는 사이버 공격 및 보안 위협 이벤트를 24시간 365일 모니터링하여 위협을 탐지하고 이를 분석하여 대응하는 일련의 활동

### 보안관제의 목표
- 컴퓨터 및 정보통신망을 통해 처리-유통되거나 저장되는 정보에 대한 기밀성, 무결성, 가용성을 보장하는 것

- 기밀성: 
    - 처리-유통 중이거나 저장된 자료/정보가 비인가자에게 유출되지 않도록 하는 것
    - 유출되더라고 빠르게 탐지하고 대응할 수 있도록 하는 것
    - 어떤 방법으로든 정보가 유출되기 때문에 신속한 대응이 중요
        - 부적절한 공개-노출로부터 보호하는 것
- 무결성:
    - 비인가자에 의해 저장되거나 변경 또는 삭제되지 않도록 하는 것
    - 인가자라 하더라도 부당한 방법으로 저장되거나 변경 또는 삭제되지 않도록 하는 것
        - 변조-파괴로부터 보호하는 것
- 가용성:
    - 인가잦가 정보자산에 접근 또는 사용하려는 시점이 허용된 범위 이내라면 지체없이 접근-사용할 수 있도록 정상적인 상태로 유지하는 것
        - 지연,거부,마비로부터 보호하는 것


### 보안관제의 주요 활동
- 탐지
    - 탐지를 못하면 아무런 대처를 하지 못한다.
    - 실시간으로 탐지하는활동
    - 소관 정보자산이 해킹 경유지로 악용되지 않도록 보안 취약점을 발굴하는 활동 포함
    - 내부(내부자)의 보안 위협을 탐지하는 것도 중요
        - 외부 공격에 의한 유출보다 내부 유출이 더 빈번하다!

- 분석
    - 이벤트, 기록, 침해 증거 토대
    - 공격자 정보, 공격 시간, 공격 방법 등을 발견
    - 피해 규모와 영향성 등을 파아

- 대응
    - 피해 사실 보고 및 통보: 피해 사실을 관련 조직 및 인원에게 보고 및 통보
    - 전문기술 제공: 분석 과정에서 파악된 공격자 정보와 취약점 정보를 활용하여 피해시스템이 정상적으로 운영될 수 있도록 신속하게 전문기술 제공
    - 침해사고 방지: 공격에 사용된 해킹도구 및 해킹 기법을 토대로 해킹 공격 탐지 기술을 개발하고 보안관제 업무에 활용하여 동일한 침해사고 방지


### 보안관제의 원칙
- 기본원칙
1. 무중단의 원칙
    - 사이버 공격은 시간과 장소에 구애받지 않고 수시로 발생하므로 정보통신망에 대한 보안관제 업무를 24시간 중단 없이 수행해야 함
2. 전문성의 원칙
    - 정보통신망에 대한 보안관제 업무를 책임 있게 수행하기 위해 전문 기술력을 갖춘 보안관제 인력과 첨단 시설을 갖추어야 함
    - 기술력만 있다고 보안관제 업무를 할 수 있는 것이 아님. 법에서 정한 요건을 충족해야 인정된다.
3. 정보공유의 원칙: 사이버 공격으로 인해 피해가 다른 곳으로 확산되는 것을 방지하기 위해 법령에 위배되지 않는 범위에서 보안관제 관련 정보를 공유해야 함

- 주의사항
    - 정보통신망 및 정보시스템을 실시간 모니터링하는 업무의 특성상 인권, 사생활 등을 침해하여 개인의 기본권을 제약할 여지가 있으므로 법령에 저촉되지 않도록 주의
    - 반드시 관계법령에 대한 이해를 바탕으로 법령 범위 내에서 업무 수행
    - 보안 이벤트 이외 일반적인 통신 데이터를 모니터링하지 않도록 주의(통신비밀보호법, 형사소송법, 군사법원법 위반 소지)
    - 업무를 통해 지득한 정보를 유출하지 않도록 주의

### 사이버 위기경보
- 개념: 사이버 공격 및 위협에 체계적으로 대응하고 대비하기 위해 파급영향과 피해 규모 등을 고려하여 단계별 경보를 발령하는 체계
- 관련 근거(경보 발령): 국가사이버안전관리규정 제11조, 사이버안보 업무규정 제 15조

- 발령권자
    - 국가-공공: 국가정보원장
    - 국방: 국방부장관
    - 민간: 과학기술정보통신부장관
- 보통 한 쪽이 울리면 뒤따라 울리고, 한 쪽이 내리면 뒤따라 내리는 형식
- 보통 국가정보원장이 발령한다.

- 사이버 위기경보의 단계
    - 심각도에 따라 5단계로 구성
    - 정상 - 관심 - 주의 - 경계 - 심각
    - 일상화된 사이버 위협 환경에 대응하는 취지로 2022년 2월 15일부터 상시 '관심'단계 유지 중
        - 최소한 관심 단계 유지.
        - 정상으로 돌아가긴 힘들 수 있다.(공격 심화)
    - ncsc 웹사이트에서 확인 가능



### 국가사이버안보센터
- 국가사이버안보센터(국가정보원)
    - 국가 차원에서 주요 정보통신망에 대한 상시 보안관제를 수행하는 국가 보안운영센터
    - 사이버 위협을 실시간으로 탐지하고 단계별 사이버 위기경보 발령

- 주요 업무
    - 국가사이버안보 정책 집행
        - 정책의 기획,조율 지원 & 관련 제도, 지침 수립
        - 민,관,군 정보공유체계 및 국가사이버보안협력센터 구축 및 운영
        - 해외 유관기관과의 협력체계 구축
    - 사이버공격 탐지 활동
        - 주요 정보통신망에 대한 24시간 365일 상시 보안관제
        - 부문별 사이버안전센터(부문 보안운영센터) 총괄
        - 단계별 사이버 위기경보 발령
        - 신종 해킹 탐지기술 개발 및 지원
    - 사이버안보 위기 예방 활동
        - 보안조사 및 컨설팅
        - 위기 대응 훈련
        - 보안성 검토 및 보안 측정
        - 보안 관리 실태 평가
        - 정보보호제품 평가,인증 및 보안 적합성 검증
    - 사고 조사 및 복구지원 활동
        - 해킹사고 발생 시 사고조사 및 원인규명
        - 민관군 합동조사 총괄
        - 시스템 복구 지원 및 공격 기법 분석


## 우리나라의 보안관제 관련 법,제도와 전문기업
- 보안관제 전문기업 제공 서비스

### 보안관제의 관계법령
- 보안은 결국 법령과 밀접한 관계가 있다.
- 주요 정보통신기반(인프라)
    - 정보통신기반 보호법(법률, 과학기술정보통신부)
        - 제16조(정보공유 및 분석센터)

- 국가-공공 분야
    - 사이버안보 업무규정(대통령령, 국가정보원)
        - 제14조(사이버공격, 위협의 탐지 및 대응)
    - 국가사이버안전관리규정(대통령훈령, 국가정보원)
        - 제10조의2(보안관제센터의 설치 및 운영)

- 국방 분야
    - 국방정보화 기반조성 및 국방정보자원관리에 대한 법률(법률, 국방부)
        - 제21조(국방정보침해에 대한 대응 및 복구)

- 민간 분야
    - 정보통신망 이용촉진 및 정보보호 등에 관한 법률(법률, 과학기술정보통신부)
        - 제48조2(침해사고의 대응 등)
        - 제52조(한국인터넷진흥원)




### 보안관제 전문기업
- 보안관제 전문기업 지정 제도
    - 보안관제 업무를 안전하고 신뢰성 있게 수행할 능력이 있다고 인정되는 기술인력과 업무수행능력을 갖춘 업체를 국가에서 지정하는 제도
    - 국가사이버안전과니 규정에 근거하여 보안관제 전문기업이 아닌 경우 관제서비스 제공 불가

    - 인력요건: 기술인력 15명 이상(고급 3, 중급 6명 이상)
    - 자본요건: 자기자본 20억 원 이상
    - 수행요건: 보안관제 업무수행 능력 평가 결과 70점 이상

- KISA, 정보보호산업진흥포털에서 보안관제 전문기업 확인 가능

## 보안운영센터(SOC)와 정보공유-분석센터(ISAC)
- 보안운영센터(Security Operatiions Center, SOC)
    - 보안관제센터라고도 함
    - 보안관제를 포함하여 보안운영 전반에 대한 업무를 수행하는 조직

- 국가,공공 보안운영센터
- 국방 분야 보안운영센터
- 민간 분야 보안운영센터(KISA)

### 정보공유-분석센터(ISAC)
- 실시간으로 정보공유를 통해 빠른 대응 가능
- 유사 업무 분야별로 사이버 공격과 침해사고에 대해 효과적으로 공동대응하기 위한 수행 체계 및 조직
- 정보통신기반 보호법
- 독자적으로 개별 대응하는 것보다 인력 운영과 비용 측면에서 훨씬 효율적

- 의료 ISAC같은 경우 가입률이 적음
- 주요 ISAC: 정보통신, 금융, 의료, 지자체, 방산

## 보안관제 거버넌스의 이해
### 거버넌스
- 조직이 조직의 목표를 추구하는 데 의사결정을 내리고 그 의사결정을 실행하는 체계(ISO26000)
    - 조직이 어떤 목표를 달성하기 위해 제반 활동을 조정,통제,관리하는 체계와 프로세스
    - 조직의 전략,정ㅈ책,절차를 설정하는 것과 이를 이행하는 방식에 관한 것을 포함

- 거버넌스는 오늘날 조직의 운영 및 관리에 중요한 개념으로 작용
    - ESG 경영: Environmental Social Governacne
    - 정보보호 GRC: Governance Risk Compliance

- 보안관젱제 대한 의사결정 체계


### 보안관제 거버넌스의 필요성
1. 효과적인 위험관리
    - 조직에 대한 사이버보안 위험을 체계적으로 식별,평가,관리하며, 신속하고 효과적인 대응을 위한 기반 마련
2. 일관된 보안 프로세스와 정책
    - 프로세스와 정책이 일관성 있고 체계적으로 실행되도록 보장하여 보안관제의 효율성을 높이고 보안사고에 대한 대응력 강화
3. 책임 및 투명성 확보
    - 조직 내 역할과 책임을 명확히 하여 보안사고 발생 시 투명한 대응을 가능하게 하며 조직 내-외부의 신뢰 구축에 기여
4. 법적 요구사항 충족 및 규제 준수
    - 관계 법령 및 방침에서 정하는 요구사항 및 규제 충족을 위해 필수적


## 보안관제 유형과 보안운영 조직 구성
- 수행 방식에 따른 분류
    - 자체(인하우스) 보안관제
        - 직접 보안관제: 자체 시설과 자체 인력 활용
    - 외주(아웃소싱) 보안관제
        - 파견 보안관제: 자체 시설을 두되, 보안관제 전문 업체 인력 활용
        - 원격 보안관제: 보안관제 전문 업체의 시설과 인력 활용

### 자체 보안관제
- 조직 규모가 크거나 많은 정보 자산과 중요 정보를 취급하는 경우 자체 보안관제가 유리
- 자체 보안관제를 인하우스 보안관제라고도 하며, 직접 보안관제가 대표적인 수행 방식

- 이점
    - 조직읮 요구사항을 고려한 맞춤식 보안관제 체계 구축 가능
    - 보안 이벤트와 대응에 대한 직접적인 관리 및 통제 강화
    - 조직 내 데이터가 외부로 유출되지 않도록 기밀성 유지 가능
    - 내부의 기밀 유지와 신속한 사고 처리에 초점
    - 보안관제 기술을 자체 보유할 수 있고 업무의 연속성 보장
- 한계    
    - 자체 조직과 설비를 구축하고 유지하는데 많은 비용 발생
    - 전문 보안 인력을 채용하고 지속적인 커리어 개발 요구
    - 최신 보안 위협에 대응하기 위한 기술 업데이트 및 자원 투입 등 지속적인 유지 노력 필요

- 보안관제 취업을 할 때 전문성을 키울 수 있는 환경을 제공해주는 회사를 들어가라!


### 외주 보안관제
- 조직 규모가 작거나 비용 투자 및 인력 확보 등 부담을 낮추는 경우 외주 보안관제가 유리
- 외주 보안관제를 아웃소싱 보안관제라고도 하며, 파견 보안관제와 원격 보안관제로 구분

- 작은 규모의 기업은 보안관제가 사실 필요 없다!

- 파견 보안관제
    - 자체 보안관제 시설을 구축하되, 조직은 보안관제 전문 업체의 인력을 파견받아 운영하는 방식
    - 보안관제 전문 업체의 보유 기술력과 노하우 활용 가능
- 원격 보안관제
    - 자체 보안관제 시설을 구축하지 않고 보안관제 전문 업체의 보안관제 체계를 그대로 사용하는 방식
    - 최소의 비용으로 보안관제 전문 업체의 보안관제 체계와 인력 활용 가능

- 이점
    - 자체 조직과 설비 구축에 비해 저렴할 수 있고, 고정된 예산으로 관리 가능하여 효울적인 비용 운영
    - 전문 조직 및 전문가의 지식과 경험 등 전문성을 즉각 활용 가능
    - 조직 상황과 관계없이 항상 일정 수준의 보안관제 서비스 담보
- 한계
    - 보안관제 전문 업체의 자원 및 접근방식을 따르므로 조직의 요구사항에 대한 맞춤화 제안(추가요금)
    - 조직 내 데이터가 일정 부분 보안관제 전문 업체와 공유되므로 데이터 유출 우려
    - 보안관제 전문 업체와의 의사소통 문제 발생 가능성
    

## 보안관제 방법론과 프로세스
### 주요 활동
- 기본 활동
    - 탐지 - 분석 - 대응
    - 모니터링을 기본으로 하는 활도
- 확장된 활동
    - 예방 - 탐지 - 분석 - 대응 - 보고
    - 예방과 탐지는 정보수집
    - 분석은 정보분석
    - 대응과 보고는 정보공유로 이어진다.

- 보호해야 할 정보자산을 식별하고 관리하는 것이 굉장히 중요하다! --> 잘 안지켜지는 조직이 굉장히 많다.


### 예방
- 자산 식별 및 관리: 정보통신망 및 정보시스템을 구성하는 정보자산을 식별하고 관리하는 업무
- 취약점 진단: 정보통신망 및 정보시스템에 대한 알려진 보안취약점을 파악하는 업무
- 모의해킹: 정보통신망 및 정보시스템에 대한 공격표면을 확인하고, 내재되어 있을 가능성이 있는 잠재적 보안취약점을 식별하는 업무
- 보안 패치: 정보자산에 대한 알려진 보안취약점을 제거하는 업무
- 보안운영 정책 관리: 방화벽, IDS/IPS 등 정보보호체계에 보안운영 정책(규칙)을 적용하는 업무


### 탐지
- 위협 이벤트 탐지: 해킹, 침해시도, 침해사고 등 보안위협 이벤트를 탐지하는 업무
- 장애 이벤트 탐지: 통신마비, 시스템 오류, 성능 저하 등 장애 이벤틀르 탐지하는 업무
- 관리적 이벤트 탐지: 기타 관리에 필요한 각종 이벤트를 탐지하는 업무


### 분석
- 초기 분석: 보안관제 수준에서 이벤트를 빠르게 분석하여 정탐/오탐 여부와 추가적인 조사-분석 필요성 등 대응 방책을 판단하는 업무
    - 오탐 --> 자체 종결
    - 정탐 --> 경중에 따라 대응 방책 판단
    - 초기 단계에 너무 많은 힘을 들이지 않는다.
- 침해사고 조사: 발생된 침해사고에 대해 정보자산으로부터 침해사고 흔적을 패증하거나 디지털 포렌식 등의 방법으로 조사하는 업무
- 악성코드 분석: 채증된 악성코드 샘플을 역공학 등의 수단을 활용하여 정밀하게 분석하는 업무

### 대응
- 상황전파: 피해 정보자산과 관련된 조직에 관련 사항을 통보하는 업무
- 인터뷰: 피해자를 대상으로 경위(사용 행위 등) 확인 등 인터뷰를 시행하는 업무
- 차단 및 격리: 피해 정보자산을 정보통신망에서 분리하고 격리 조치하는 업무
- 보안운영 정책 개발: 정보보호체계에 적용하기 위한 보안운영 정책(규칙)을 개발하는 업무
- 복구: 피해 정보자산을 정상 상태로 복구하여 조직의 업무 연속성 보장

### 보고
- 조직 내 보고: 조직 내 보안 관련 의사결정권자에게 관련 사항을 보고하는 업무
- 유관기관 신고: 경찰 등 정보수사기관에 침해사고 및 피해 사실을 신고하는 업무
- 조직 내 공유(공지): 조직 내 인식수준을 맞추고 향후 재발방지 등을 위한 전사 공지
- 사이버위협정보(CTI): 정보공유 업무와 연계하여 알려진 위협으로 생산 및 공유

### 정보수집-정보분석-정보공유
- 정보수집
    - 내부 위협정보 수집(내부 위협 헌팅): 조직 내 정보통신망 및 정보시스템에 내재될 수 있는 보안 위협정ㅈ보를 수집하는 업무
    - 외부 위협정보 수집(외부 위협 헌팅): 인터넷을 통해 광범위하게 유포되고 있는 보안 위협정보를 수집하는 업무

- 정보분석
    - 자동화 분석 시스템 등을 활용하여 네트워크 이벤트, 악성코드 샘플 등에 대한 기본 정보, 평판 정보, 행위 등을 분석하는 업무

- 정보공유
    - 위협정보 공유: 유관기관, 전문업체에서 생산,배포하는 위협정보를 받거나 내부에서 생산된 위협정보를 제공하는 업무

### 기술지원 및 운영
- 기술지원
    - 보안관제에 필요한 기술 및 도구를 지원하거나 정책 개발/관리 등의 업무를 수행하는 업무
    - 보안관제팀에서 자체적으로 수행할 수도 있으나, 보안관제팀이 모니터링에만 집중하는 경우 모니터링 이외의 업무는 기술지원 업무로 분류

- 운영
    - 보안관제에 필요한 인프라 및 시스템의 상시 운영을 수행하는 업무
    - 인프라 및 시스템에 대한 유지관리와 하자보수를 포함

## 보안관제 관련 프레임워크
- 보안관제 업무의 표준
    - 보안관제 업무가 표준화된 방식으로 수행될 수 있도록, 보안관제 조직과 보안관제 서비스 조직은 국제적으로 활용되는 표준 프레임워크를 토대로 보안관제 방법론을 개발하여 활용 중
    - 특히, 보안관제 전문 기업 지정 요건에 보안관제 방법론이 포함되어 있기 때문에 보안관제 조직과 보안관제 서비스 업체는 업무 특성을 고려, 반영한 보안관제 방법론 확보

- 보안관제 프레임워크
    - 정책, 절차, 위험관리, 모니터링, 분석, 사고대응-관리, 기술-인프라, 성과 측정, 보고

- 프레임워크의 이점
    - 체계적인 위험관리 접근법
    - 일관된 보안 프로세스와 정책 수립
    - 신속한 사고 대응 및 복구 지원
    - 법적 요구사항 및 규제 준수
    - 효과적인 자원 배분

### NIST Cybersecurity Framework(CSF)
- 미국 국립표준기술연구소(미연방 규격)에서 개발한 사이버보안 프레임워크는 조직이 사이버보안 위험을 관리하는데 필요한 표준, 지침, 모범 사례를 제공

- 미연방의 독특한 특성이 포함되어 있는 경우가 많아 국내에 맞게 개량해야 할 필요가 있다.
- 미국 기준에 맞춰진 프레임워크!

- ESM과 SIEM

## 보안관제시스템의 이해
### 보안관제시스템
- 보안관제 업무를 원활하게 수행할 수 있도록 수집, 분석, 가시화/시각화 등의 기능을 통합적으로 제공하는 시스템
    - 수집: 보관된 보안 정보 및 이벤트 기록을 모아서 보관
    - 분석: 이벤트 기록을 분석
    - 가시화/시각화: 관제사가 이해하기 쉽게 그림, 그래프 등의 형태로 표현

- 외부의 침해시도 뿐 아니라 내부의 보안위협을 모니터링하는 데에도 활용
- 최근에는 법규/방침 등 콤플라이언스 준수 여부 모니터링 기능을 포함
- 보안관제시스템 도입 시 보안관제사는 개별 정보보호시스템을 각각 들여다볼 필요 없이 보안관제시스템만 보면 되기 때문에 업무 효율 향상 가능


### 보안관제시스템 세대 구분
- ESM --> SIEM
- 지능형 의사결정: SOAR 사용하여 Playbook 기반 위협 대응(정형화 반복 자동화)
- 지능형 능동대응: 지능인지, 인지 상호작용, 의사결정 과정을 통해 더욱 발전된 대응이 가능할 것이다.

1. 1세대: 단위보안관제
    - 개별 정보보호시스템 또는 개별 네트워크 장비에 접속하여 개별적으로 모니터링 화면을 확인하거나 로그를 직접 확인하는 방식으로 수행
    - 시스템의 크기와 복잡도가 증가하고, 침해사고가 급증하면서 한계 도래

2. 2세대: 통합보안관제
    - ESM을 활용하여 통합된 보안 정보 및 이벤트 기록을 모니터링
    - ESM: 다양한 정보보호시스템 및 네트워크 장비에서 생성된 보안 정보 및 이벤트 기록을 수집/분석하는 기능을 제공하는 보안관제시스템
    - 기술 수준의 한계로 APT 공격이나 알려지지 않은 위협을 식별하기에는 제한

3. 3세대: 빅데이터보안관제
    - SIEM을 활용하여 장기간에 걸쳐 보관된 대용량의 방대한 데이터를 빠르게 분석하고 시각화 함으로써 체계적인 보안관제 수행
    - 상관관계의 연관성을 분석하면 단순 시그니처뿐만 아니라 비정ㅈ상적인 이상 행위까지 탐지할 수 있어 APT공격이나 알려지지 않은 위협 식별 가능

4. 4세대: 인공지능보안관제
    - SIEM이 도입된 상태에서 SOAR를 추가로 활용하여 사이버위협정도(CTI)를 연계하고 인공지능 기술에 기반을 둔 이상 행위 탐지와 플레이북에 의한 능동적이 자동화 대응 방식의 보안관제 수행

- 현재 4세대를 향해 발전 중


### 보안관제시스템의 효용 사례
- 한국과학기술정보연구원 개발 - 사이버 공격 실시간 추적 가시화 시스템
    - 실제 보안관제 현장에서 시스템 활용
    - 61개 공공/연구기관으로부터 수집되는 대규모 보안 정보 가시화
    - 보안관제사들이 분석하는 방식으로 보안관제 업무 수행

## 정보보호시스템의 이해
### 정보보호시스템의 필요성
- 정보보호시스템: 정보의 접근통제 등을 위해 사용하는 어플라이언스 또는 소프트웨어
- 사이버 공격 및 침해시도를 탐지하거나 차단하고 보안 정보 및 이벤트로 보관

- 종류: NAC, AV, IDS/IPS, 방화벽 등

- 인터넷 공간은 수많은 침해 시도가 끊임없이 발생되고 있는 매우 위험한 공간이다.
- 정보보호시스템이 구비되지 않은 경우 해킹 공격, 내부자 위협 등 침해사고 위험에 상시 노출 상태!

- 네트워크 정보보호시스템
    - 사이버 공격이 들어오는 경계 또는 통로에서 침해 시도를 탐지하여 차단하기 위해 네트워크 정보보호시스템이 마련되어야 할 필요
    - 방화벽(침입차단시스템), IDS(침입탐지시스템), IPS(침입방지시스템), NAC(네트워크접근통제), WAF(웹 방화벽), UTM(통합위협관리),WIPS(무선침입방지) NDR(네트워크 감지 및 대응) 등
- 호스트 정보보호시스템
    - 네트워크 보호에 실패한 경우 호스트 자체에서 침해시도를 탐지하여 차단 가능
    - 호스트 보호에도 실패한 경우 호스트 정보보호시스템은 호스트를 신속히 격리하고 원인 규명을 지원 가능
    - AV(안티바이러스), 호스트 방화벽, DLP(자료유출방지), DRM(디지털저작권관리), EDR(엔드포인트 탐지 및 대응) 등

### 네트워크 정보보호시스템
- 방화벽

- IDS

- IPS

- NAC
    - 내부에 위치한 컴퓨팅 장치의 정보통신망 접근을 통제하는 정보보호시스템
    - 국가/공공에서 사용하는 공식 용어도 네트워크 접근통제
    - 비인가 컴퓨팅 장치의 정보통신망 연결을 통제
        - 등록되지 않은 MAC주소를 가진 컴퓨팅 장치는 비인가 컴퓨팅 장치로 간주
    - 우리나라에서는 내부 보안 정책의 준수 여부까지 확인한다.

- WAF
    - 방화벽은 네트워크 경계 보호가 목적이기 때문에 응용 계층을 검사하더라도 헤더(응용서비스 유형)까지만 확인한다. 
        - 확인 안할 수도 있음(포트번호까지만 확인)
    - 웹 방화벽
    - 웹 응용프로그램에 대한 웹 해킹을 방어하기 위해 응용 계층의 본문 내용(메시지)까지 확인
    - 웹 응용프로그램이 탑재된 웹 서버의 앞 단에 설치.
        - 단, 웹 서버 내 설치하는 웹 방화벽(소프트웨어)의 경우 호스트 정보보호시스템으로 분류

- UTM
    - 통합위협관리 시스템
    - 정보통신망의 규모가 큰 경우: 방화벽, IDPS, WAF 등을 별도 설치하여 운용하는 것이 이득이다.
    - 정보통신망의 규모가 작은 중소 조직의 경우 UTM 사용이 유리하다.
- WIPS
    - 무선침입방지
    - 와이파이 등 무선 환경에서의 비인가 컴퓨팅 장치의 정보통신망 접근을 통제하는 정보보호시스템
    - NAC와 유사한 기능도 포함하지만, 무선 환경에 특화된 기능 제공
    - 조직 내 무선 AP에 대한 공격 트래픽 차단
    - 조직 내 무선 AP에 연결을 시도하는 경우 인가된 컴퓨팅 장치인지 여부와 인가된 사용자인지 여부를 확인
    - 인가되지 않은 핫스팟/테더링 등 비인가 무선 신호의 발생 또는 접속을 차단(인증 해제)

- NDR
    - 컴퓨팅 능력과 저장장치의 용량 증가 덕분에 모든 트래픽을 장기간 보관했다 심층 분석할 수 있게 해주는 NDR의 필요성 대두
    - 모든 트래픽을 통째로 보관하기 때문에 개인정보 논란이 있을 수 있음
    - 막대한 양의 저장장치가 필요함
    

### 호스트 정보보호시스템
- AV
    - 컴퓨터 바이러스와 악성코드를 탐지하고 방어하는 호스트 기반 정보보호시스템
- Host Firewall
    - 일반적으로 지칭하는 방화벽(침입차단시스템)은ㅇ 네트워크 정보보호시스템으로, 네트워크 경계를 보호하거나 내부 통신을 통제하기 위한 목적으로 사용되는 네트워크 방화벽
    - 현대적인 운영체제는 개별 호스트를 보호하기 위해 호스트 방화벽 기능을 내장하는 경향
- DLP
    - 자료유출방지
    - 내부자 위협을 방지하기 위함
    - 내부 자료의 외부 반출을 통제하고, 자료/정보의 흐름을 모두 기록으로 보존
    - DLP만으로는 부족한 감이 있다.
- DRM
    - 디지털저작권관리
    - 자료유출방지에는 우회할 방법이 많기 때문에 충분하지 못하다.
    - 이를 보완하기 위해 사용하는 기술로, 조직에서 생산되는 파일에 디지털 서명을 하고 암호화 함으로써 조직 외부로 자료/정보가 유출되더라도 열 수 없게 조치할 수 있다.
- EDR
    - NDR과 마찬가지로 각 호스트에서 발생되는 모든 이벤트를 실시간으로 모니터링하며 수집, 분석할 수 있는 EDR의 필요성 대두
    - 안티 바이러스에서 진단하지 못하는 신종 바이러스/악성코드가 존재할 수 있고, 이외에도 사용자가 악의적인 목적으로 내부 정보를 유출하는 등의 행위 가능성 존재
    - 컴퓨터 운영체제 내 모든 이벤트와 네트워크 연결에 관한 정보를 모니터링
    - 한계: 
        - 모든 트래픽을 통째로 보관하기 때문에 개인정보 논란이 있을 수 있음
        - 막대한 양의 저장장치가 필요함

### 정보보호시스템 인증/평가의 필요성
- 국가 안보에 영향을 미칠 수 있기 때문에 일정 수준 이상의 보안 기능 품질이 담보되어야 함
- CC평가(과학기술정보통신부)
- 정보보호제품성능평가(과학기술정보통신부)
- 보안 기능 시험(국가정보원) -- 시험이 되어 경쟁력이 떨어짐
- 보안적합성 검증(국가정보원)
- 암호 모듈 검증(국가정보원)


### CC평가/인증
- 일반적인 CC인증/평가는 국제 CC인증을 의미
- 국내용 CC를 별도 운영중
- 정책기관: 과학기술정보통신부
- 인증기관: 국가보안기술연구소(NSR)의 IT보안인증사무국(ITSCC)
- 평가기관: 한국인터넷진흥원(KISA) 등 7개 기관(업체)



### 정보보호제품 성능평가
- 정보보호시스템이 운영환경에서 네트워크 트래픽에 얼마나 적절히 대응하는지 성능을 평가하여 결과를 제공하는 제도
- 성능 위주의 지표를 고려

- 정책기관: 과학기술정보통신부
- 심의/관리기관: KISA
- 성능평가기관: 한국정보통신협회(TTA) 등 5개 기관(업체)

### 보안 기능 시험
- CC 평가/인증은 국제적인 공통평가기준과 공통평가방법론을 적용하고 있는 만큼 안전성과 신뢰성이 담보된다. 그러나, 많은 시간과 비용이 투입되는 단점이 존재
- 보안 기능이 탑재된 IT제품에 대한 국가용 보안요구사항 만족 여부를 시험하여 안전성을 확인하는 제도

### 보안적합성 검증
- 보안기능이 탑재된 IT제품의 안전성을 검증하는 제도
- CC 평가/인증을 받았거나 보안 기능 시험을 통과한 검증필 제품은 검증절차 생략(간소화) 가능
 

### 암호모듈 검증
- 비밀이 아닌 비공개 업무자료를 보호하기 위해 국가/공공 분야에서 도입하는 사용 암호모듈의 안전성과 구현 적합성을 검증하는 제도


## 정보보호 로그
- 로그: 
    - 컴퓨터 시스템에서 발생되는 각종 이벤트에 대한 정보를 담고 있는 기록
    - 파일 형태로 보관되거나 데이터베이스 형태로 보관
    - 별도의 분석 필요

- 로그 활용의 예
    - 시스템 성능 이슈 확인
    - 시스템 장애 원인 확인
    - 시스템 취약점 분석

### Syslog
- 컴퓨팅 장치마다 로그를 생성하는 형식이 다르면 매 장치에 대한 로그 분석 메커니즘을 다르게 구현해야 하므로 비효율성이 야기될 가능성이 있다.
- 로그의 표준!
- 구조
    - 우선순위(PRI): 장비유형 * 8 + 심각도
    - 헤더
        - 시간, 호스트명을 비롯하여 로그 생ㅈ성자의 식별에 필요한 정보 포함
        - 헤더와 메시지의 구분을 위해 헤더의 마지막에는 콜론과 띄어쓰기 한 칸 삽입
        - 메시지: 로그의 세부 내용





### 로그의 보관
- 모든 정보자산은 로그를 생성하여 자체적으로 보관하는 기능 보유
- 로그의 손실 가능성 때문에 별도의 로그 서버를 유지하여 따로 보관할 필요성 존재
- 로그의 무결성을 담보하기 위한 보안 대책 필요

### 네트워크 정보보호시스템
- 방화벽, IDPS 등 대부분의 네트워크 정보보호시스템은 POSIX운영체제를 기반으로 하고 있어 로그 규격으로 Syslog를 채택

- 전통적인 방화벽: 최소한의 기억장치만으로도 충분
- 대부분의 네트워크 정보보호시스템: 자체 통계/분석 기능을 제공하기 위해 일정 수준에서 로그를 보관


### 호스트
- 운영체제 종류에 따라 로그를 생성하여 보관하는 방식이 상이하다.

- 윈도우
    - 중앙집중화된 이벤트 형태로 로그를 저장 --> 관리 용이
    - 생성/유지할 로그의 유형을 보안 정책에서 설정 후 이벤트 뷰어를 통해 확인
    - 공격자가 로그를 삭제하기에 용이

- POSIX
    - Syslog 채택
    - 각 유형별로 산발적인 위치에 저장 --> 관리에 있어 운영체제에 대한 기본 지식 요구
    - 공격자가 로그를 모두 삭제하기가 어려움

### 호스트 정보보호 시스템
- 호스트에서 실행되는 호스트 정보보호시스템은 응용 프로그램인만큼 개발사에 따라 로그를 생성/보관하는 방식이 상이

- 파일 형태: txt, log 등 텍스트 파일 형태로 생성하며, 압축하거나 암호화하는 경우 임의로 해독 불가

- 데이터베이스 형태: 자체 정의한 파일 형태로 생성하며 임의로 해독 불가

- 원격 보관: 호스트에는 남겨두지 않고 개발사 서버로 모두 전송

## ESM과 SIEM
- ESM
    - 보안 정보 및 이벤트 기록을 수집/분석하는 기능을 제공하는 보안관제시스템
    - 로그를 얼마나 잘 파싱에서 레파지토리로 넣어주느냐가 관건이었음
    - 관계형 데이터베이스 기반으로 개발
    - 방대한 양의 데이터 처리 및 보관에 한계
    - Agent에서 데ㅣ터를 일정 수준 정제해서 Manager에 보내야 했기 때문에 정보보호시스템, 네트워크 장비 위주로 로그를 수집할 수밖에 없었고, 모든 종류의 정보자산을 들여다보지는 못하는 한정적인 수준의 수집만 가능했다.

- SIEM
    - ESM에 비해 보다 다양한 형태의 컴퓨팅 장치 및 소프트웨어를 포괄할 수 있는 빅데이터 기반 보안관제시스템
    - 장기간에 걸쳐 보관된 대용량의 방대한 데이터를 빠르게 분석 가능
    - Agentless 방식 채택으로 로그 정제과정 없이 그대로 보내도 변환하여 보관하는 기능 제공
        - 단, SIEM 관리 기능에서 정제/변환을 위한 규칙을 정의해서 넣어주어야 함
    - APT 공격이나 알려지지 않은 위협을 식별할 수 있도록 상관 관계의 연관성 분석 기능을 제공
        - 단순 시그니처 뿐만 아니라 비정상적인 이상 행위까지 탐지 가능
        - 대시보드를 통해 시각화된 형태의 분석 결과 제공

### 보안관제 가시화/시각화
- 차이점
    - 가시화: 보안 정보 및 이벤트 기록을 볼 수 있게 하는 것
    - 시각화: 단순히 보여주는 수준을 넘어 쉽게 이해할 수 있도록 시각적으로 표현하는 것

- 보안관제 시각화
    - 방대한 양의 보안 정보 및 이벤트 기록을 가시화 할 뿐 아니라 차트, 그래프, 맵, 애니메이션 등 시각적 요소를 사용하여 쉽게 이해할 수 있게 표현하는 것

- 대시보드: 전체적인 현황을 일목요연하고 직관적으로 볼 수 있게 구성된 화면
    - 외부 공격 중심 구성: 외부로부터 유입되는 침해시도 및 위협 등 공격 흐름을 중심으로 구성하는 형태
    - 내부 위협 중심 구성: 내부자에 의한 위협 행위를 중심으로 구성하는 형태
    - 정보자산 상태 중심 구성: 보안관제 대상 정보자산의 상태를 중심으로 구성하는 형태
