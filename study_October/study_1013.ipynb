## 정보보호컨설팅 개요
- 정보보호컨설팅: 조직의 정보자산을 보호하기 위해 법, 제도, 기술, 관리적 측면에서 위험을 진단 및 분석하고 대응방안을 제시하여 보안 수준을 향상시키는 전문 서비스.
- 자산의 개념이 중요하다. 
- 사람에 대한 진단도 진행. 보안서약서, 권한, 계정이 부여되어 있는지 등을 초점으로 진단.
- 시설은 IDC와 같은 시설을 지칭

- 법/제도 준수, 기술적 진단, 관리적 개선
    - 우리나라는 법/제도 준수 측면이 강함.
    - 연 1회 진행(금융은 2회): 법에 명시되어 있음.
    - ISMS 인증은 연 1회 진행하도록 되어 있음. 이 외에도 국가기반시설에 대해서는 연 1회 진행하도록 명시되어 있음. (법에 근거)

- 컨설팅은 예방 차원.
 



### 리스트 기반 접근 사이클
- 단순 점검이 아닌 위험 식별/분석/평가를 통해 우선순위를 정하고, 효과적인 개선을 추진하는 체계적 방법론.
- 리스크를 줄이기 위해 사전에 식별된 위험을 점검.
    1. 리스크 식별: 자산에 대한 식별과 이에 위협이 되는 요소를 도출
    2. 리스크 분석: 위험 요소의 발생 가능성과 영향도 평가
    3. 리스크 평가: 분석된 위험을 조직의 기준과 비교

- 정보보호는 정보의 비즈니스 가치 보호, 정보보안은 정보시스템 기반에서의 위협 관리.
- 정보보호 3요소. 상당히 중요하다. 자산을 평가할 때도 사용함. 대부분의 사고는 CIA로 설명이 가능함.

- 컨설팅을 받는 것은 건강검진을 받는 것. 기술의 발전 속도가 매우 빠르기 때문에 새로운 취약점이 계속 생겨남.

- ISMS는 작년매출 100억 이상, 5천명 이상의 병상을 가진 중형 병원 등에서는 필수이나, 필수가 아닌 기업들도 인증을 받을 시 침해사고 발생 시 법원에서 참작하여 감형받을 수 있어 대부분 받는 편이다.



- 컨설팅은 고객과 함께 만들어가는 공동 작업의 성격이다. 또한, 협업이 중요한 작업이다.
