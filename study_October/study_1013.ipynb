### 질의응답
1. 컨설팅으로 시작하는 것을 추천하는가?
- 컨설팅은 많은 분야가 있다. 
- 신입을 뽑으면 기술진단부터 시작하는 것이 첫번째.
- 수탁사 점검. 개인정보를 한 기업에서 다 처리할 수 없기 때문에 위탁,수탁에 대한 관리가 필요하다. 이러한 업무부터 시작한다.

2. 모의해킹, 취약점 진단, 컨설팅의 차이?
- 어디에 집중하는지가 가장 큰 차이점이다. 셋 다 개발부터 시작하는데, 각자 깊이의 차이가 존재한다. 집중하는 부분이 다름. 모의해킹은 시나리오 기반 진단으로 좀 더 실력이 있음. 컨설팅은 좀 더 전체적인 범위로 넓게 하는 직무. 최종 프로젝트가 OJT를 겸한다. 

3. 면접
- 이력서에서 질문. 최근 보안 이슈에 대해 질문. 

## 정보보호컨설팅 개요
- 정보보호컨설팅: 조직의 정보자산을 보호하기 위해 법, 제도, 기술, 관리적 측면에서 위험을 진단 및 분석하고 대응방안을 제시하여 보안 수준을 향상시키는 전문 서비스.
- 자산의 개념이 중요하다. 
- 사람에 대한 진단도 진행. 보안서약서, 권한, 계정이 부여되어 있는지 등을 초점으로 진단.
- 시설은 IDC와 같은 시설을 지칭

- 법/제도 준수, 기술적 진단, 관리적 개선
    - 우리나라는 법/제도 준수 측면이 강함.
    - 연 1회 진행(금융은 2회): 법에 명시되어 있음.
    - ISMS 인증은 연 1회 진행하도록 되어 있음. 이 외에도 국가기반시설에 대해서는 연 1회 진행하도록 명시되어 있음. (법에 근거)

- 컨설팅은 예방 차원.
 



### 리스트 기반 접근 사이클
- 단순 점검이 아닌 위험 식별/분석/평가를 통해 우선순위를 정하고, 효과적인 개선을 추진하는 체계적 방법론.
- 리스크를 줄이기 위해 사전에 식별된 위험을 점검.
    1. 리스크 식별: 자산에 대한 식별과 이에 위협이 되는 요소를 도출
    2. 리스크 분석: 위험 요소의 발생 가능성과 영향도 평가
    3. 리스크 평가: 분석된 위험을 조직의 기준과 비교

- 정보보호는 정보의 비즈니스 가치 보호, 정보보안은 정보시스템 기반에서의 위협 관리.
- 정보보호 3요소. 상당히 중요하다. 자산을 평가할 때도 사용함. 대부분의 사고는 CIA로 설명이 가능함.

- 컨설팅을 받는 것은 건강검진을 받는 것. 기술의 발전 속도가 매우 빠르기 때문에 새로운 취약점이 계속 생겨남.

- ISMS는 작년매출 100억 이상, 5천명 이상의 병상을 가진 중형 병원 등에서는 필수이나, 필수가 아닌 기업들도 인증을 받을 시 침해사고 발생 시 법원에서 참작하여 감형받을 수 있어 대부분 받는 편이다.



- 컨설팅은 고객과 함께 만들어가는 공동 작업의 성격이다. 또한, 협업이 중요한 작업이다.

## 컨설팅 프로세스 흐름
- 제안요청서RFP를 작성하기 전에 RFI가 먼저 작성되어야 함. 연계되어 있는 작업.
- 제안 요청 내용이 들어감. 제안서 작성 방안, 견적 및 금액에 대한 내용이 존재. 
- 수주를 위한 경쟁이 있음. 실주가 되면 인력이 놀게 됨.

- 국내 서면 점검 업체: 44개
- 현장 점검 업체: 32개
    - 서면 점검은 ISMS, ISMS-P 이외에도 여러가지 조건을 통과한 업체만이 가능하여 믿을만한 업체라는 의미이다. 보안수준이 인정된 업체

- 해외 수탁사 서면 점검의 경우 문서 번역 작업도 필요하기 때문에 영어 가능자 우대함.

## 컨설턴트 역량 & 윤리
- 논리적 사고 필요
- 근본 원인을 찾아 해결하는 것이 중요
- 개발자의 감에 의존하는 것이 아닌 가이드에 따라 절차적인 진행을 할 수 있어야 올바른 접근이다.


- 3대 역량: 분석 역량, 문서화 역량, 커뮤니케이션 역량

- 보안 컨설턴트는 조직의 기밀 정보와 보안 현황을 직접 다루는 사람으로 윤비적 책임과 신뢰 확보가 컨설팅 성과보다 먼저 전제되어야 한다.



## 정보보호컨설팅 분류 및 유형
- 3대 영역으로 구분. 각 영역별 특성에 맞는 진단/개선 활동을 통해 조직의 보안 체계를 종합적으로 강화함.

- 관리영역: 조직 차원의 보안 정책/프로세스 체계화
- 기술영역: IT 인프라 및 애플리케이션의 취약점 식별/제거
- 물리영역: 시설 및 자산에 대한 물리적 접근 통제 및 보호

- ISMS 등 종합으로 들어가면 관리 기술 물리를 종합적으로 보기 때문에 알아만 두자.

### 컨설팅 유형
- ISMS, ISMS-P: 조직의 정보보호 수준을 정량적으로 평가하고, 법령/기준에 따른 체계 구축 및 인증 획득을 지원하는 종합 서비스

- 개인정보 영향평가: 시스템 도입 시 개인정보 처리 위험을 사전에 평가하는 유형

- 클라우드 기반: CSAP 인증 컨설팅. 공공기관이 클라우드를 활용하고자 할 떄 해당 인증/인가를 받은 업체만 허용함.

- ISMS: 80개 항목. 정보보호 관리체계 수립 및 운영 인증. 일정 규모 이상 공공기관/기업 대상. 정보통신망법 근거.
- ISMS-P: 80개 + 개인정보 21항목 총 101개 항목. 정보보호 + 개인정보보호 통합 인증. 개인정보를 처리하는 사업자 대상. 개인정보보호법 근거.




### 기술진단 컨설팅
- 기준명
    - OWASP Top 10: 웹/앱 중심의 국제 표준 취약점 분류 (예: SQL, XSS, Broken Auth 등)
    - CVE/CWE: 국제 공개 취약점 식별 번호 체계 및 분류 체계
    - CCE: 보안설정항목
    - CVSS: 취약점 심각도 등급 산정 체계(Base /Temporal / Environmental Score)
    - ISMS / CSAP / NIST: 국내외 인증 기준 및 점검 기준서(KISA, NIST SP800-115 등)

- 진단 프로세스
    1. 진단 및 Gap 분석
        - 현재 운영 중인 정보보호 활동 분석
        - 인증 기준과의 차이점 도출
    2. 정책 및 절차 문서화
        - 정보보호/개인정보보호 정책서, 지침, 운영 절차 문서 작성
    3. 위험분석 및 보호조치 설계
        - 자산 식별, 위협/취약점 분석, 위험도 평가 및 대응
    4. 증적자료 수집 및 관리
        - 이행증적(로그, 시스템 설정, 회의록 등) 확보 및 보관
    5. 내부 점검 및 모의심사 훈련
        - 인증 대응 리허설 및 질의응답 시뮬레이션


### 개인정보보호 컨설팅
- 조직이 개인정보를 수집 이용 보유 파기 등 전 과정에서 적법하고 안전하게 처리하고 있는지를 진단하고 개선을 지원하는 컨설팅 서비스

- 일반 개인정보보호 진단 컨설팅
- PIA 컨설팅(개인정보보호 영향평가)
- 수탁사 관리 컨설팅
- 개인정보관리체계(ISMS-P 대응) 컨설팅
- 국외 법규 대응 컨설팅(GDPRm CBPR) 등
